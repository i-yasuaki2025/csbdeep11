```r
# ==============================================================================
# SVM Radial 解析：完全版
# (12ファイル対応 + NA完全除去 + モデル・プロットデータ保存 + 指定Grid)
# ==============================================================================

library(caret)
library(kernlab)
library(dplyr)
library(Metrics)

# --- 設定 ---
set.seed(42)

# データパス
base_path <- "/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/20220127_PLS 202201 ALL/20251216_for_making_collect_data_nm_baseall_OHFP_csv/rebuilt_merged_data/"

# 出力先
output_path <- "./" 
if (!dir.exists(output_path)) dir.create(output_path)

# 12ファイル対応
file_names <- c(
    "n_base.csv", "n_base_OH_rebuilt.csv", "n_base_FP_rebuilt.csv",
    "n_all.csv",  "n_all_OH_rebuilt.csv",  "n_all_FP_rebuilt.csv",
    "m_base.csv", "m_base_OH_rebuilt.csv", "m_base_FP_rebuilt.csv",
    "m_all.csv",  "m_all_OH_rebuilt.csv",  "m_all_FP_rebuilt.csv"
)

target_vars <- c("Jsc", "Voc", "FF", "PCEmax")
today <- format(Sys.Date(), "%Y%m%d")

# --- 関数定義 ---

# OOD分割関数
create_ood_split <- function(df, feature_cols, ood_ratio = 0.1) {
  X <- df[, feature_cols, drop = FALSE]
  centroid <- colMeans(X)
  dists <- apply(X, 1, function(row) sqrt(sum((row - centroid)^2)))
  num_ood <- max(1, floor(nrow(df) * ood_ratio))
  ood_indices <- order(dists, decreasing = TRUE)[1:num_ood]
  return(list(train = df[-ood_indices, ], test = df[ood_indices, ]))
}

# スケーリング関数 (-1 to 1)
scale_neg1_to_1 <- function(train_df, test_df, feature_cols) {
  pp <- preProcess(train_df[, feature_cols, drop=FALSE], method = "range")
  train_01 <- predict(pp, train_df[, feature_cols, drop=FALSE])
  test_01  <- predict(pp, test_df[, feature_cols, drop=FALSE])
  train_df[, feature_cols] <- train_01 * 2 - 1
  test_df[, feature_cols]  <- test_01 * 2 - 1
  return(list(train = train_df, test = test_df))
}

# Permutation Importance 計算関数
calc_permutation_importance <- function(model, data, target_col, feature_cols, base_r2) {
  importance_list <- list()
  for (col in feature_cols) {
    temp_data <- data
    temp_data[[col]] <- sample(temp_data[[col]])
    
    # 特徴量のみ渡す
    preds <- predict(model, temp_data[, feature_cols, drop=FALSE])
    
    new_r2 <- cor(data[[target_col]], preds)^2
    if(is.na(new_r2)) new_r2 <- 0
    importance_list[[col]] <- base_r2 - new_r2
  }
  return(importance_list)
}

# --- メイン処理 ---
summary_data <- data.frame()
feature_importance_data <- data.frame()

cat(paste0("--- SVM Radial Analysis Started: ", today, " ---\n"))
cat("Sigma Grid: 2^seq(-15, 3, length = 7)\n")

for (file in file_names) {
  filepath <- file.path(base_path, file)
  if (!file.exists(filepath)) {
    cat(sprintf("Skipping: %s (Not found)\n", file))
    next
  }
  
  cat(sprintf("\n=== Processing SVM Radial: %s ===\n", file))
  
  tryCatch({
    df_raw <- read.csv(filepath, row.names = 1, check.names = FALSE)
  }, error = function(e) {
    cat(paste("Error reading file:", e$message, "\n"))
    next
  })
  if ("X" %in% colnames(df_raw)) df_raw$X <- NULL
  
  for (target in target_vars) {
    if (!target %in% colnames(df_raw)) next
    
    # 1. データ整理 (NA完全除去)
    numeric_cols <- sapply(df_raw, is.numeric)
    df_temp <- df_raw[, numeric_cols, drop=FALSE]
    df_temp <- na.omit(df_temp)
    
    if(nrow(df_temp) < 20) next
    
    features <- setdiff(colnames(df_temp), target)
    vars <- sapply(df_temp[, features, drop=FALSE], var, na.rm=TRUE)
    features <- names(vars)[vars > 0 & !is.na(vars)]
    
    # 重複削除 (相関行列NA対策済み)
    if(length(features) > 1) {
      cor_mat <- cor(df_temp[, features], use = "pairwise.complete.obs")
      cor_mat[is.na(cor_mat)] <- 0
      high_corr <- findCorrelation(cor_mat, cutoff = 0.99999)
      if (length(high_corr) > 0) features <- features[-high_corr]
    }
    if(length(features) == 0) next
    
    # 2. OOD分割
    split_res <- create_ood_split(df_temp, features, ood_ratio = 0.1)
    train_set <- split_res$train
    test_set  <- split_res$test
    
    # 3. スケーリング (-1 to 1)
    scaled_res <- scale_neg1_to_1(train_set, test_set, features)
    train_scaled <- scaled_res$train
    test_scaled  <- scaled_res$test
    
    # 4. モデル構築 (svmRadial)
    # GPRと同じsigma範囲 + Costパラメータ(C)の組み合わせ
    tune_grid <- expand.grid(
      sigma = 2^seq(-15, 3, length = 7),
      C = c(1, 10, 100)
    )
    
    train_ctrl <- trainControl(method = "cv", number = 10, savePredictions = "final")
    
    set.seed(42)
    model <- train(
      x = train_scaled[, features, drop=FALSE],
      y = train_scaled[[target]],
      method = "svmRadial",
      metric = "RMSE",
      trControl = train_ctrl,
      tuneGrid = tune_grid
    )
    
    # 5. 評価
    # 【重要】predictには特徴量のみ渡す
    pred_train <- predict(model, train_scaled[, features, drop=FALSE])
    r2_train <- cor(train_scaled[[target]], pred_train)^2
    rmse_train <- rmse(train_scaled[[target]], pred_train)
    mae_train <- mae(train_scaled[[target]], pred_train)
    rpd <- sd(train_scaled[[target]]) / rmse_train
    
    pred_test <- predict(model, test_scaled[, features, drop=FALSE])
    r2_ood <- cor(test_scaled[[target]], pred_test)^2
    rmse_ood <- rmse(test_scaled[[target]], pred_test)
    
    summary_row <- data.frame(
      File = file, Target = target,
      R2 = r2_train, RMSE = rmse_train, MAE = mae_train, RPD = rpd,
      n_samples = nrow(df_temp), n_features = length(features),
      OOD_R2 = r2_ood, OOD_RMSE = rmse_ood,
      Best_Params = paste0("sigma=", round(model$bestTune$sigma, 5), ", C=", model$bestTune$C)
    )
    summary_data <- rbind(summary_data, summary_row)

    # === 保存処理 ===
    
    # A. モデル本体 (.rds)
    model_data_bundle <- list(model = model, training_data = train_scaled, ood_test_data = test_scaled)
    rds_file <- paste0("fixed_", today, "_SVM_Radial_model_", tools::file_path_sans_ext(file), "_", target, ".rds")
    saveRDS(model_data_bundle, file = file.path(output_path, rds_file))

    # B. プロット用データ (.csv)
    df_pred_all <- rbind(
      data.frame(Observed = train_scaled[[target]], Predicted = as.numeric(pred_train), Type = "Train"),
      data.frame(Observed = test_scaled[[target]],  Predicted = as.numeric(pred_test),  Type = "OOD_Test")
    )
    pred_file <- paste0("fixed_", today, "_SVM_Radial_", tools::file_path_sans_ext(file), "_", target, "_pred.csv")
    write.csv(df_pred_all, file.path(output_path, pred_file), row.names = FALSE)

    # 6. 変数重要度
    cat(sprintf("  Computing importance for %s...\n", target))
    imps <- calc_permutation_importance(model, train_scaled, target, features, r2_train)
    
    for (feat_name in names(imps)) {
      imp_val <- imps[[feat_name]]
      if (imp_val > 1e-5) {
        feature_importance_data <- rbind(feature_importance_data, data.frame(
          File = file, Target = target, Feature = feat_name, Importance_Mean = imp_val
        ))
      }
    }
    cat(sprintf("  Target: %s | R2: %.3f | OOD_R2: %.3f\n", target, r2_train, r2_ood))
  }
}

# --- サマリー保存 ---
sum_file <- file.path(output_path, paste0("fixed_", today, "_SVM_Radial_R_summary.csv"))
write.csv(summary_data, sum_file, row.names = FALSE)
imp_file <- file.path(output_path, paste0("fixed_", today, "_SVM_Radial_R_feature_importance.csv"))
write.csv(feature_importance_data, imp_file, row.names = FALSE)

cat(sprintf("\nSVM Radial Analysis Finished. Summary: %s\n", sum_file))
```

```r
# ==============================================================================
# データ診断ツール：NA (欠損値) と 分散0 (定数値) の特定
# ==============================================================================

base_path <- "/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/20220127_PLS 202201 ALL/"
output_path <- "./"

file_names <- c(
    "n_base.csv", "n_base_OH.csv", "n_base_FP.csv",
    "n_all.csv", "n_all_OH.csv", "n_all_FP.csv",
    "m_base.csv", "m_base_OH.csv", "m_base_FP.csv",
    "m_all.csv", "m_all_OH.csv", "m_all_FP.csv"
)

# 結果格納用データフレーム
diagnosis_report <- data.frame(
  File = character(),
  Problem_Type = character(), # "NA_Found" or "Zero_Variance"
  Column_Name = character(),
  Count = integer(),          # NAの数 または ユニークな値の数(1なら分散0)
  stringsAsFactors = FALSE
)

cat("--- Data Diagnosis Started ---\n")

for (file in file_names) {
  filepath <- file.path(base_path, file)
  if (!file.exists(filepath)) {
    cat(sprintf("Skipping: %s (Not found)\n", file))
    next
  }
  
  # 読み込み
  df <- read.csv(filepath, row.names = 1, check.names = FALSE)
  if ("X" %in% colnames(df)) df$X <- NULL
  
  # 数値列のみ対象
  numeric_cols <- sapply(df, is.numeric)
  df_num <- df[, numeric_cols, drop=FALSE]
  
  cat(sprintf("\nChecking: %s (Rows: %d, Numeric Cols: %d)\n", file, nrow(df), ncol(df_num)))
  
  # -------------------------------------------------------
  # 1. NA (欠損値) のチェック
  # -------------------------------------------------------
  na_counts <- colSums(is.na(df_num))
  na_cols <- na_counts[na_counts > 0]
  
  if (length(na_cols) > 0) {
    cat(sprintf("  [!] Found NA in %d columns!\n", length(na_cols)))
    print(head(na_cols, 5)) # 最初の5つを表示
    
    # レポートに追加
    for (col_name in names(na_cols)) {
      diagnosis_report <- rbind(diagnosis_report, data.frame(
        File = file,
        Problem_Type = "NA_Found",
        Column_Name = col_name,
        Count = na_cols[[col_name]]
      ))
    }
  } else {
    cat("  [OK] No NAs found.\n")
  }
  
  # -------------------------------------------------------
  # 2. 分散0 (すべての値が同じ) のチェック
  #    ※これが相関行列でNAを生む主な原因です
  # -------------------------------------------------------
  # NAを除外して分散を計算
  vars <- sapply(df_num, var, na.rm = TRUE)
  
  # 分散が0、または計算不能(NA)になっている列を探す
  zero_var_cols <- names(vars)[vars == 0 | is.na(vars)]
  
  if (length(zero_var_cols) > 0) {
    cat(sprintf("  [!] Found %d Zero-Variance columns (Constant values)!\n", length(zero_var_cols)))
    print(head(zero_var_cols, 5))
    
    # レポートに追加
    for (col_name in zero_var_cols) {
      diagnosis_report <- rbind(diagnosis_report, data.frame(
        File = file,
        Problem_Type = "Zero_Variance",
        Column_Name = col_name,
        Count = 0 # 分散0という意味で0
      ))
    }
  } else {
    cat("  [OK] No Zero-Variance columns found.\n")
  }
  
  # -------------------------------------------------------
  # 3. 無限大 (Inf) のチェック
  # -------------------------------------------------------
  inf_counts <- sapply(df_num, function(x) sum(is.infinite(x)))
  inf_cols <- inf_counts[inf_counts > 0]
  
  if (length(inf_cols) > 0) {
    cat(sprintf("  [!] Found Infinite values in %d columns!\n", length(inf_cols)))
    for (col_name in names(inf_cols)) {
      diagnosis_report <- rbind(diagnosis_report, data.frame(
        File = file,
        Problem_Type = "Infinite_Value",
        Column_Name = col_name,
        Count = inf_cols[[col_name]]
      ))
    }
  }
}

# --- 結果の保存 ---
if (nrow(diagnosis_report) > 0) {
  output_csv <- file.path(output_path, "data_diagnosis_report.csv")
  write.csv(diagnosis_report, output_csv, row.names = FALSE)
  cat(sprintf("\nDiagnosis Finished. Problems found.\nDetailed report saved to: %s\n", output_csv))
  
  # 簡易サマリー表示
  cat("\nSummary of problems:\n")
  print(table(diagnosis_report$File, diagnosis_report$Problem_Type))
  
} else {
  cat("\nDiagnosis Finished. NO PROBLEMS FOUND in any file.\n")
}
```

```r
import pandas as pd
import os

# --- 設定 ---
base_path = "/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/20220127_PLS 202201 ALL/"
output_path = "./"

# Rの診断で問題が見つかったファイルを重点的にチェック
target_files = ["n_all_FP.csv", "m_all_OH.csv"] 

print("--- Detailed Data Diagnosis Started ---\n")

for file in target_files:
    filepath = os.path.join(base_path, file)
    if not os.path.exists(filepath):
        print(f"Skipping {file} (Not found)")
        continue

    print(f"=== Checking File: {file} ===")
    
    # データを読み込む
    try:
        df = pd.read_csv(filepath, index_col=0)
        # ID列(X)があれば削除
        if "X" in df.columns:
            df = df.drop(columns=["X"])
    except Exception as e:
        print(f"Error reading file: {e}")
        continue

    # --- 1. NA（欠損値）の犯人探し ---
    # どの「列」が悪いのか？
    na_counts = df.isnull().sum()
    bad_cols = na_counts[na_counts > 0]
    
    if len(bad_cols) > 0:
        print(f"\n[Problem 1] Found NAs in {len(bad_cols)} columns:")
        print(bad_cols)
        
        # どの「行（ID）」が悪いのか？（最初の10件を表示）
        # 行ごとの欠損数をカウント
        rows_with_na = df[df.isnull().any(axis=1)]
        print(f"\n  -> Total {len(rows_with_na)} rows contain NAs.")
        print("  -> First 10 Problematic Sample IDs:")
        print(rows_with_na.index.tolist()[:10])
        
        # A or B の判定ヒント
        # もし列名に "Jsc", "Voc", "Lay2thickness" などが含まれていれば -> ベースデータ(A)の問題
        # もし列名が "FP_..." や "Bit..." などであれば -> 特徴量データ(B)の問題
    else:
        print("\n[OK] No NAs found.")

    # --- 2. 分散0（値が同じ）の犯人探し ---
    # 数値列のみ抽出
    df_num = df.select_dtypes(include=['number'])
    variances = df_num.var()
    zero_var_cols = variances[variances == 0].index.tolist()
    
    if len(zero_var_cols) > 0:
        print(f"\n[Problem 2] Found {len(zero_var_cols)} Zero-Variance columns (All values are same):")
        print(zero_var_cols)
        # 値を確認（どうせ全部同じなので最初の1つを表示）
        first_val = df[zero_var_cols[0]].iloc[0]
        print(f"  -> Value is constant: {first_val}")
    else:
        print("\n[OK] No Zero-Variance columns found.")
    
    print("-" * 50)

print("\nDiagnosis Finished.")
```
