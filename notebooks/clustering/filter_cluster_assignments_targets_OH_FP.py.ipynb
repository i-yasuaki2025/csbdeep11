```python
"""
filter_cluster_assignments_targets_OH_FP.py

STEP3.3_U25_20251027_correct / [OH|FP] / 04_assignments にある
ClusterAssign_*.csv から、指定した材料名（OH）／変数名（FP）の
行だけを抽出して、05_filtered_rows_YYYYMMDD_HHMMSS 以下に
フィルタ済みCSVとサマリCSVを出力するスクリプト。
"""

import pandas as pd
import numpy as np
from pathlib import Path
from datetime import datetime
import re

# =========================
# Common parameters
# =========================
BASE = Path("/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/cal_cluster_No/20251026_under_25clusters")
RUN_ROOT = "STEP3.3_U25_20251027_correct"
MODES = ["OH", "FP"]   # 両方まとめて処理

# =========================
# Target name lists
# （※OH側の長いリストは元コードからそのまま貼り付けてください）
# =========================

# ---- OH: SMILES / 材料名リスト ----
OH_TARGET_NAMES_STR = """
SMILESsnameinM_PC61BM,SMILESsnameinM_PC71BM,SMILESsnameip1M_AtD,SMILESsnameip1M_BP,SMILESsnameip1M_BP.C60,SMILESsnameip1M_Bu.BP.C60,SMILESsnameip1M_C4DBTA,SMILESsnameip1M_C6DBTA,SMILESsnameip1M_C8DBTA,SMILESsnameip1M_DTA,SMILESsnameip1M_EBDBTA,SMILESsnameip1M_EH.DBTA,SMILESsnameip1M_f.BP.C60,SMILESsnameip1M_r.BP.C60,SMILESsnameip2M_BP,SMILESsnamenM_C60,SMILESsnamenM_DBFIEDOT,SMILESsnamenM_DPc.T,SMILESsnamenM_DPc.mP,SMILESsnamenM_ICEM,SMILESsnamenM_ICMA,SMILESsnamenM_ICMM,SMILESsnamenM_MOPFP,SMILESsnamenM_NC70BA,SMILESsnamenM_NCMA,SMILESsnamenM_NCMM,SMILESsnamenM_PC61BM,SMILESsnamenM_PC71BM,SMILESsnamenM_PCBM,SMILESsnamenM_PFP,SMILESsnamenM_PNDTI.BT,SMILESsnamenM_PNDTI.DTT,SMILESsnamenM_PNDTI.T,SMILESsnamenM_PNDTI.TT,SMILESsnamenM_PTBI2T,SMILESsnamenM_PTBI2TR,SMILESsnamenM_PTBI2Tz,SMILESsnamenM_PTBITT,SMILESsnamenM_TC61BM,SMILESsnamenM_TC61PM,SMILESsnamenM_TFP,SMILESsnamenM_THPc,SMILESsnamep1M_2.2..bithiopheneP1,SMILESsnamep1M_3hexylthieno.3.2b.thiopheneP3,SMILESsnamep1M_3hexylthieno.3.2b.thiopheneP4,SMILESsnamep1M_9triarylamino9HcarbazoleP3,SMILESsnamep1M_ADT,SMILESsnamep1M_AtD,SMILESsnamep1M_BBTzBTIDG,SMILESsnamep1M_BBTzalkylthiopheneBTIDG,SMILESsnamep1M_BDTC6,SMILESsnamep1M_BDTDTKP,SMILESsnamep1M_BDTH1,SMILESsnamep1M_BDTH11,SMILESsnamep1M_BDTH3,SMILESsnamep1M_BDTH6,SMILESsnamep1M_BDTH7,SMILESsnamep1M_BDTH8,SMILESsnamep1M_BDTH9,SMILESsnamep1M_BDTT6,SMILESsnamep1M_BP,SMILESsnamep1M_BPC60,SMILESsnamep1M_BPc,SMILESsnamep1M_BPcpre1,SMILESsnamep1M_BPcpre2,SMILESsnamep1M_BTAnt,SMILESsnamep1M_BTIBDTP2,SMILESsnamep1M_BuBPC60,SMILESsnamep1M_C10DPPBP,SMILESsnamep1M_C2DPPBP,SMILESsnamep1M_C3DPPTTSe,SMILESsnamep1M_C3DPPTTT,SMILESsnamep1M_C3DPPTTTe,SMILESsnamep1M_C4DPPBP,SMILESsnamep1M_C6DPPBP,SMILESsnamep1M_C8DPPBP,SMILESsnamep1M_CPDTDTKP,SMILESsnamep1M_DTA,SMILESsnamep1M_DTAnt,SMILESsnamep1M_DTBT,SMILESsnamep1M_DTPDPP2TBDT,SMILESsnamep1M_DTPDPP2TDBT,SMILESsnamep1M_DTPDPP2TTT,SMILESsnamep1M_DTPDPP3T,SMILESsnamep1M_DTPDPP4T,SMILESsnamep1M_DTPDPPTPT,SMILESsnamep1M_DTSC4,SMILESsnamep1M_DTSC6,SMILESsnamep1M_DTSC6L,SMILESsnamep1M_DTSC8,SMILESsnamep1M_DTSCEH,SMILESsnamep1M_DTSCh,SMILESsnamep1M_DTSChy,SMILESsnamep1M_DTSDC,SMILESsnamep1M_DTST6,SMILESsnamep1M_DTSTCh,SMILESsnamep1M_DTSTChy,SMILESsnamep1M_DTTPTP1,SMILESsnamep1M_DTTPTP2,SMILESsnamep1M_DTTPTP3,SMILESsnamep1M_DTTPTP4,SMILESsnamep1M_DTTPTP5,SMILESsnamep1M_DTTPTP6,SMILESsnamep1M_DTfBT,SMILESsnamep1M_DTffBT,SMILESsnamep1M_EHDBTA,SMILESsnamep1M_F0F0,SMILESsnamep1M_F0F2,SMILESsnamep1M_F1PDPP2TBP,SMILESsnamep1M_F2F0,SMILESsnamep1M_F2F2,SMILESsnamep1M_F2PDPP2TBP,SMILESsnamep1M_FLODTKP,SMILESsnamep1M_P.BDTFBT.,SMILESsnamep1M_P.BDTTBT.,SMILESsnamep1M_P.BDTTTBO.,SMILESsnamep1M_P.BDTTTBT.,SMILESsnamep1M_P.BDTTTFBT.,SMILESsnamep1M_P.BDTTTHBT.,SMILESsnamep1M_P.T32EHiI.,SMILESsnamep1M_P.T3C10iI.,SMILESsnamep1M_P.T3C12iI.,SMILESsnamep1M_P.T3C6iI.,SMILESsnamep1M_P.T3C8iI.,SMILESsnamep1M_P3HNT,SMILESsnamep1M_P3HT,SMILESsnamep1M_P4TBT,SMILESsnamep1M_P4TDPP,SMILESsnamep1M_P4TDTBT,SMILESsnamep1M_P4TPDTQ,SMILESsnamep1M_PBDFDODTBT,SMILESsnamep1M_PBDFDTBO,SMILESsnamep1M_PBDFDTBT,SMILESsnamep1M_PBDFDTBTz,SMILESsnamep1M_PBDFFBT,SMILESsnamep1M_PBDT.DTBSe,SMILESsnamep1M_PBDT.DTBT,SMILESsnamep1M_PBDTBDD,SMILESsnamep1M_PBDTBO,SMILESsnamep1M_PBDTBT,SMILESsnamep1M_PBDTBTz,SMILESsnamep1M_PBDTDTBTm,SMILESsnamep1M_PBDTDTBTo,SMILESsnamep1M_PBDTDTBTp,SMILESsnamep1M_PBDTDTBTz,SMILESsnamep1M_PBDTDPT1,SMILESsnamep1M_PBDTDPT2,SMILESsnamep1M_PBDTDTTTST,SMILESsnamep1M_PBDTFBO,SMILESsnamep1M_PBDTFBT,SMILESsnamep1M_PBDTFQT,SMILESsnamep1M_PBDTFQTT,SMILESsnamep1M_PBDTHDODTHBTffP3,SMILESsnamep1M_PBDTPBO,SMILESsnamep1M_PBDTPDTBT,SMILESsnamep1M_PBDTPODTBO,SMILESsnamep1M_PBDTPODTBT,SMILESsnamep1M_PBDTPOFBT,SMILESsnamep1M_PBDTT10TPD,SMILESsnamep1M_PBDTT12TPD,SMILESsnamep1M_PBDTT8TPD,SMILESsnamep1M_PBDTTBO,SMILESsnamep1M_PBDTTBT,SMILESsnamep1M_PBDTTDPP,SMILESsnamep1M_PBDTTDTBO,SMILESsnamep1M_PBDTTDTBT,SMILESsnamep1M_PBDTTDTBTz,SMILESsnamep1M_PBDTTDTTBT,SMILESsnamep1M_PBDTTEHDTEHBTffP2,SMILESsnamep1M_PBDTTEHDTHBTffP1,SMILESsnamep1M_PBDTTFBSe,SMILESsnamep1M_PBDTTFDPP,SMILESsnamep1M_PBDTTFTTE,SMILESsnamep1M_PBDTTPD,SMILESsnamep1M_PBDTTPDP1,SMILESsnamep1M_PBDTTPDP2,SMILESsnamep1M_PBDTTPDP3,SMILESsnamep1M_PBDTTS1,SMILESsnamep1M_PBDTTS2,SMILESsnamep1M_PBDTTS3,SMILESsnamep1M_PBDTTSeDPP,SMILESsnamep1M_PBDTTTBT,SMILESsnamep1M_PBDTTTC,SMILESsnamep1M_PBDTTTCF,SMILESsnamep1M_PBDTTTE,SMILESsnamep1M_PBDTTTS,SMILESsnamep1M_PBDTTTST,SMILESsnamep1M_PBPT12,SMILESsnamep1M_PBQ1,SMILESsnamep1M_PBQ2,SMILESsnamep1M_PBQ3,SMILESsnamep1M_PBQ4,SMILESsnamep1M_PBT0F,SMILESsnamep1M_PBT1F,SMILESsnamep1M_PBT2F,SMILESsnamep1M_PBT3F,SMILESsnamep1M_PBTPyDPP,SMILESsnamep1M_PBTff4T2OD,SMILESsnamep1M_PBnDTFTAZ,SMILESsnamep1M_PBnDTHTAZ,SMILESsnamep1M_PCBTDPP,SMILESsnamep1M_PCDTBT,SMILESsnamep1M_PCDTBX,SMILESsnamep1M_PCDTDPP,SMILESsnamep1M_PCDTPT,SMILESsnamep1M_PCDTPX,SMILESsnamep1M_PCDTQx,SMILESsnamep1M_PCDTTPP,SMILESsnamep1M_PCPDTBT,SMILESsnamep1M_PCPTDO,SMILESsnamep1M_PCTDPP,SMILESsnamep1M_PCV,SMILESsnamep1M_PCVT,SMILESsnamep1M_PCVTT,SMILESsnamep1M_PCVTTTT,SMILESsnamep1M_PCz,SMILESsnamep1M_PDPP2FTC12,SMILESsnamep1M_PDPP2FTC14,SMILESsnamep1M_PDPP2FTC16,SMILESsnamep1M_PDPP2T,SMILESsnamep1M_PDPP2TBP,SMILESsnamep1M_PDPP3T,SMILESsnamep1M_PDPP3TaltTPT,SMILESsnamep1M_PDPPBBT,SMILESsnamep1M_PDPPBT,SMILESsnamep1M_PDPPTPT,SMILESsnamep1M_PDPPTVT,SMILESsnamep1M_PDTPDTDPP,SMILESsnamep1M_PDTPDTDPPBu,SMILESsnamep1M_PDTST,SMILESsnamep1M_PDTSTPD,SMILESsnamep1M_PEN,SMILESsnamep1M_PFBTBDT,SMILESsnamep1M_PFDTBT,SMILESsnamep1M_PFDTDPP,SMILESsnamep1M_PGFDTBT,SMILESsnamep1M_PGFDTDPPC4S25,SMILESsnamep1M_PGFDTDPPC8,SMILESsnamep1M_PICDTBT,SMILESsnamep1M_PIDTTFBSe,SMILESsnamep1M_PNNT12DT,SMILESsnamep1M_PNNT12HD,SMILESsnamep1M_PNNTDT,SMILESsnamep1M_PNOz4T,SMILESsnamep1M_PNPz4T,SMILESsnamep1M_PNT4T2OD,SMILESsnamep1M_PNTz4T,SMILESsnamep1M_PNTz4TF2,SMILESsnamep1M_PNTz4TF4,SMILESsnamep1M_PPDT2FBT,SMILESsnamep1M_PPDTBT,SMILESsnamep1M_PPDTFBT,SMILESsnamep1M_PQT12,SMILESsnamep1M_PSBTBT,SMILESsnamep1M_PSEHTT,SMILESsnamep1M_PTB1,SMILESsnamep1M_PTB2,SMILESsnamep1M_PTB3,SMILESsnamep1M_PTB4,SMILESsnamep1M_PTB5,SMILESsnamep1M_PTB6,SMILESsnamep1M_PTB7,SMILESsnamep1M_PTB7Th,SMILESsnamep1M_PTBDTDTBT,SMILESsnamep1M_PTBDTffDTBT,SMILESsnamep1M_PTBF0,SMILESsnamep1M_PTBF1,SMILESsnamep1M_PTBF2,SMILESsnamep1M_PTBF3,SMILESsnamep1M_PTBFTDTBT,SMILESsnamep1M_PTBI2T,SMILESsnamep1M_PTBI2TR,SMILESsnamep1M_PTBI2Tz,SMILESsnamep1M_PTBITT,SMILESsnamep1M_PTzBT10HD,SMILESsnamep1M_PTzBT10OD,SMILESsnamep1M_PTzBT12HD,SMILESsnamep1M_PTzBT12OD,SMILESsnamep1M_PTzBT14HD,SMILESsnamep1M_PTzBT14OD,SMILESsnamep1M_PTzBTBOBO,SMILESsnamep1M_PTzBTBOHD,SMILESsnamep1M_PTzBTBOOD,SMILESsnamep1M_PTzBTEHHD,SMILESsnamep1M_PTzBTEHOD,SMILESsnamep1M_PTzBTHD,SMILESsnamep1M_PTzQT14,SMILESsnamep1M_PeNPz2T,SMILESsnamep1M_PeNPz3T,SMILESsnamep1M_PeNPz4T,SMILESsnamep1M_PffBT4T2OD,SMILESsnamep1M_PhBADT,SMILESsnamep1M_Poly.N9.fheptadecanyl2.7carbazolealt5.5.4.f.7.fdi2.thienyl..1.2.5.oxadiazolo.3.4d.pyridazine.P1,SMILESsnamep1M_Poly.N9.fheptadecanyl2.7carbazolealt5.5.4.f.7.fdi2thienyl.2.3dioctylpyrazino.2.3d.pyridazine.P2,SMILESsnamep1M_Poly.N9.fheptadecanyl2.7carbazolealt5.5.4.f.7.fdi2thienyl.pyrazino.2.3d.pyridazine.P3,SMILESsnamep1M_RanPBDPPT,SMILESsnamep1M_RegPBDPPT,SMILESsnamep1M_TDPPBBT,SMILESsnamep1M_TPDBr16,SMILESsnamep1M_Thieno.3.4c.pyrrole4.6dioneP1,SMILESsnamep1M_Thieno.3.4c.pyrrole4.6dioneP2,SMILESsnamep1M_Thieno.3.4c.pyrrole4.6dioneP3,SMILESsnamep1M_TzTzBTIDG,SMILESsnamep1M_TzTzalkylthiopheneBTIDG,SMILESsnamep1M_VacdepoBPc,SMILESsnamep1M_fBPC60,SMILESsnamep1M_pBTTT,SMILESsnamep1M_rBPC60,SMILESsnamep1M_ranPDPPTPDalt2T,SMILESsnamep1M_regPDPPTPDalt2T,SMILESsnamep1M_thieno.3.2b.thiopheneP2,SMILESsnamep2M_PBDTTFTTE
"""

# ---- FP: 変数名リスト ----
FP_TARGET_NAMES_STR = """
X100.1,X100.3,X100.4,X101.3,X101.4,X101.5,X102.4,X105,X105.1,X105.3,X105.4,X107.4,X108.1,X108.3,X108.4,X109.1,X109.3,X109.4,X110.3,X110.4,X111.3,X111.4,X112.1,X112.3,X112.4,X113.3,X113.4,X114.1,X114.3,X114.4,X115.1,X115.3,X115.4,X116.3,X116.4,X117.4,X118.1,X118.3,X118.4,X119.4,X120.3,X120.4,X121.1,X121.3,X121.4,X122.3,X122.4,X123.3,X123.4,X124.4,X125.3,X125.4,X126.3,X127.3,X127.4,X128.1,X128.3,X128.4,X129.1,X129.3,X129.4,X130.4,X131.1,X131.3,X131.4,X132.3,X132.4,X133.4,X134.4,X135.4,X136.4,X137.3,X137.4,X138.3,X138.4,X139.4,X140.4,X141.1,X141.4,X142.3,X142.4,X143.3,X143.4,X144.3,X144.4,X145.3,X145.4,X146.3,X146.4,X147.1,X147.3,X147.4,X148.4,X149.1,X149.3,X149.4,X150.1,X150.3,X150.4,X151.1,X151.3,X151.4,X152.3,X152.4,X153.1,X153.3,X153.4,X154.3,X154.4,X155.1,X155.3,X155.4,X157.3,X157.4,X158.1,X158.3,X158.4,X159.3,X159.4,X160.1,X160.3,X160.4,X161.1,X161.3,X161.4,X162.4,X163.3,X163.4,X164.1,X164.3,X164.4,X165,X165.1,X165.2,X165.3,X165.4,X165.5,X166.4,X17.1,X17.4,X19.3,X19.4,X20.4,X22.3,X22.4,X26.1,X26.3,X26.4,X3.4,X30.4,X33.4,X34.4,X36.1,X36.3,X36.4,X44.3,X44.4,X45.4,X46.4,X47.3,X47.4,X50.1,X50.3,X50.4,X51.4,X52.4,X57.4,X58.4,X59.4,X62.1,X62.3,X62.4,X64.4,X65.3,X65.4,X66.4,X67.4,X70.4,X71.4,X72.4,X73.4,X74.4,X76.4,X77.4,X79.3,X79.4,X80.4,X83.3,X85.4,X86.3,X86.4,X88.1,X88.3,X88.4,X89.4,X90.4,X91.4,X93.3,X94.4,X95.4,X96.3,X97.4,X98.3,X98.4,X99.3,X99.4
"""

OH_TARGET_SET = {x.strip() for x in OH_TARGET_NAMES_STR.split(",") if x.strip()}
FP_TARGET_SET = {x.strip() for x in FP_TARGET_NAMES_STR.split(",") if x.strip()}

# =========================
# Helpers
# =========================
def infer_match_column(df: pd.DataFrame, candidates: set) -> str:
    """
    データフレーム df 内で、候補値 candidates と最も多く一致する '単一の列' を推定する。
    """
    scores = {}
    for c in df.columns:
        try:
            s = df[c].astype(str)
        except Exception:
            continue
        score = s.isin(candidates).sum()
        scores[c] = int(score)
    # 最大一致列を返す（0の場合は None）
    best_col = max(scores, key=scores.get) if scores else None
    return best_col if best_col and scores[best_col] > 0 else None

def read_csv_safely(path: Path) -> pd.DataFrame:
    for enc in (None, "utf-8", "utf-8-sig", "cp932"):
        try:
            return pd.read_csv(path, encoding=enc)
        except Exception:
            continue
    # 最後の手段
    return pd.read_csv(path, engine="python")

# =========================
# Main: OH / FP をまとめて処理
# =========================
stamp = datetime.now().strftime("%Y%m%d_%H%M%S")
OUT_BASE = BASE / RUN_ROOT

all_summary = []

for mode in MODES:
    if mode == "OH":
        IN_DIR = OUT_BASE / "OH" / "04_assignments"
        TARGET_SET = OH_TARGET_SET
    else:
        IN_DIR = OUT_BASE / "FP" / "04_assignments"
        TARGET_SET = FP_TARGET_SET

    OUT_DIR = OUT_BASE / mode / f"05_filtered_rows_{stamp}"
    OUT_DIR.mkdir(parents=True, exist_ok=True)

    in_files = sorted(IN_DIR.glob("**/*.csv"))

    print(f"\n[MODE] {mode}")
    print(f"[IN   ] {IN_DIR}")
    print(f"[OUT  ] {OUT_DIR}")
    print(f"[FILES] {len(in_files)} found")

    for f in in_files:
        try:
            df = read_csv_safely(f)
        except Exception as e:
            print(f"  - SKIP (read error): {f.name} -> {e}")
            all_summary.append({
                "mode": mode,
                "file": str(f.relative_to(IN_DIR)),
                "rows": None,
                "kept": None,
                "status": f"read_error: {e}"
            })
            continue

        # 列推定
        match_col = infer_match_column(df, TARGET_SET)

        # 「name / feature / variable / label」系列のフォールバック
        if match_col is None:
            name_like_cols = [c for c in df.columns
                              if re.search(r"(name|feature|variable|var|label)", c, re.I)]
            for c in name_like_cols:
                if df[c].astype(str).isin(TARGET_SET).any():
                    match_col = c
                    break

        if match_col is None:
            print(f"  - WARN (no matched column): {f.name}")
            all_summary.append({
                "mode": mode,
                "file": str(f.relative_to(IN_DIR)),
                "rows": len(df),
                "kept": 0,
                "status": "no_match_column"
            })
            # 空でもファイルは出す（追跡のため）
            out_path = OUT_DIR / f"{f.stem}__filtered.csv"
            pd.DataFrame(columns=df.columns).to_csv(out_path, index=False)
            continue

        filtered = df[df[match_col].astype(str).isin(TARGET_SET)].copy()
        out_path = OUT_DIR / f"{f.stem}__filtered.csv"
        filtered.to_csv(out_path, index=False)

        all_summary.append({
            "mode": mode,
            "file": str(f.relative_to(IN_DIR)),
            "rows": len(df),
            "kept": len(filtered),
            "status": f"ok:{match_col}"
        })
        print(f"  - OK: {f.name} | col='{match_col}' | kept={len(filtered)}/{len(df)} -> {out_path.name}")

# まとめ（OH/FP混在のサマリ）
sum_df = pd.DataFrame(all_summary)
sum_df.to_csv(OUT_BASE / f"summary_filter_counts_{stamp}.csv", index=False)

print("\n✅ Done.")
print(f"Summary -> {OUT_BASE / f'summary_filter_counts_{stamp}.csv'}")
```
