```r
#!/usr/bin/env Rscript

############################################################
# STEP3.2_02_mds.R  （eigen_contributions 統合版）
#
# - 入力: STEP3.2_01_distance.R の出力
#   sub/01_distance_STEP3.2_signlessCorr/run_<run_id>/
#       variables/OH/distance_corr_variables_OH_<run_id>.csv
#       variables/FP/distance_corr_variables_FP_<run_id>.csv
#       samples/A_OH_plus_others/distance_corr_samples_A_OH_plus_others_<run_id>.csv
#       ...
#
# - 出力: MDS 固有値＋座標
#   sub/02_mds_STEP3.2_signlessCorr/run_<run_id>/
#     variables/<dataset>/  （OH / FP）
#       MDS_eigen_variables_<dataset>_<run_id>.csv
#       MDS_eigen_contributions_variables_<dataset>_<run_id>.csv  ★追加
#       MDS_linear_top3_<dataset>_<run_id>.csv
#       MDS_linear_cumeig_<dataset>_<run_id>.csv
#       MDS_nonlinear_top3_<dataset>_<run_id>.csv
#       MDS_nonlinear_cumeig_<dataset>_<run_id>.csv
#       MDScoords_linear_top3_variables_<dataset>_<run_id>.csv
#       MDScoords_linear_cumeig_variables_<dataset>_<run_id>.csv
#       MDScoords_nonlinear_top3_variables_<dataset>_<run_id>.csv
#       MDScoords_nonlinear_cumeig_variables_<dataset>_<run_id>.csv
#
#     samples/<dataset>/    （A_OH_plus_others, B_FP_plus_others, C_OH_only, D_FP_only）
#       MDS_eigen_samples_<dataset>_<run_id>.csv
#       MDS_eigen_contributions_samples_<dataset>_<run_id>.csv  ★追加
#       MDS_linear_top3_<dataset>_<run_id>.csv
#       MDS_linear_cumeig_<dataset>_<run_id>.csv
#       MDS_nonlinear_top3_<dataset>_<run_id>.csv
#       MDScoords_linear_top3_samples_<dataset>_<run_id>.csv
#       MDScoords_linear_cumeig_samples_<dataset>_<run_id>.csv
#       MDScoords_nonlinear_top3_samples_<dataset>_<run_id>.csv
#
#   ※ samples の nonlinear_cumeig (isoMDS) は重いため、
#      別スクリプト STEP3.2_02b_mds_samples_nonlinear_cumeig.R で処理
############################################################

suppressPackageStartupMessages({
  libs <- c("MASS", "dplyr")
  for (p in libs) {
    if (!requireNamespace(p, quietly = TRUE)) install.packages(p)
    library(p, character.only = TRUE)
  }
})

set.seed(42)

############################################################
# 0. パス設定 & 基本設定
############################################################

root_base <- "/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/cal_cluster_No/20251026_under_25clusters_program_and_result/for_checking_20251127"
sub_base  <- file.path(root_base, "sub")

step01_base <- file.path(sub_base, "01_distance_STEP3.2_signlessCorr")
step02_base <- file.path(sub_base, "02_mds_STEP3.2_signlessCorr")
dir.create(step02_base, showWarnings = FALSE, recursive = TRUE)

# 対象データセット
datasets_variables <- c("OH", "FP")
datasets_samples   <- c("A_OH_plus_others",
                        "B_FP_plus_others",
                        "C_OH_only",
                        "D_FP_only")

# MDS 次元選択用の閾値
cumeig_thr <- 0.80

############################################################
# 1. run_id の取得
############################################################

# ★ Jupyter から使うときはここを指定
override_run_id <- "20251130_153500"

get_run_id <- function() {
  # 1) override_run_id があればそれを優先
  if (!is.na(override_run_id) &&
      is.character(override_run_id) &&
      grepl("^\\d{8}_\\d{6}$", override_run_id)) {
    message("[INFO] Using override_run_id = ", override_run_id)
    return(override_run_id)
  }
  # 2) Rscript 実行時の引数
  args <- commandArgs(trailingOnly = TRUE)
  if (length(args) >= 1 && grepl("^\\d{8}_\\d{6}$", args[1])) {
    return(args[1])
  }
  stop("Usage: Rscript STEP3.2_02_mds.R <run_id (YYYYMMDD_HHMMSS)>\n",
       "  - or set override_run_id <- \"YYYYMMDD_HHMMSS\" at the top of this script.")
}

run_id <- get_run_id()
cat(">>> STEP3.2_02_mds — run_id:", run_id, "\n")

step01_run_dir <- file.path(step01_base, paste0("run_", run_id))
step02_run_dir <- file.path(step02_base, paste0("run_", run_id))
dir.create(step02_run_dir, showWarnings = FALSE, recursive = TRUE)

cat(">>> STEP3.2_02_mds input dir:   ", normalizePath(step01_run_dir), "\n")
cat(">>> STEP3.2_02_mds output dir:  ", normalizePath(step02_run_dir), "\n")

############################################################
# 2. ユーティリティ関数
############################################################

# 距離行列の読み込み
read_distance_matrix <- function(path) {
  if (!file.exists(path)) {
    stop("[ERROR] Input file not found: ", path)
  }
  df <- read.csv(path, row.names = 1, check.names = FALSE)
  dmat <- as.matrix(df)
  if (nrow(dmat) != ncol(dmat)) {
    stop("[ERROR] Distance matrix is not square: ", path)
  }
  as.dist(dmat)
}

# 固有値から「正の固有値の寄与率テーブル」を作成
compute_eigen_contributions <- function(eig_vals, thr = 0.80) {
  # eig_vals: cmdscale の $eig (ベクトル)
  eig_vals <- as.numeric(eig_vals)
  pos_idx  <- which(eig_vals > 0)
  if (length(pos_idx) == 0) {
    return(list(
      contrib_df = data.frame(
        axis        = integer(0),
        eigenvalue  = numeric(0),
        percent     = numeric(0),
        cum_percent = numeric(0)
      ),
      k_top3      = 0L,
      k_cumeig    = 0L
    ))
  }
  pos_vals <- eig_vals[pos_idx]
  total_pos <- sum(pos_vals)
  if (!is.finite(total_pos) || total_pos <= 0) {
    percent <- rep(0, length(pos_vals))
  } else {
    percent <- pos_vals / total_pos
  }
  cum_percent <- cumsum(percent)
  
  contrib_df <- data.frame(
    axis        = pos_idx,
    eigenvalue  = pos_vals,
    percent     = percent,
    cum_percent = cum_percent
  )
  
  # k_top3: 正の固有値が 3 個未満ならその個数まで
  k_top3 <- as.integer(min(3L, length(pos_idx)))
  
  # k_cumeig: cum_percent >= thr となる最小の軸数
  k_cumeig <- which(cum_percent >= thr)[1]
  if (is.na(k_cumeig)) {
    k_cumeig <- length(pos_idx)
  }
  k_cumeig <- as.integer(k_cumeig)
  
  list(
    contrib_df = contrib_df,
    k_top3     = k_top3,
    k_cumeig   = k_cumeig
  )
}

# isoMDS を安全に実行（初期値は cmdscale の座標）
run_isoMDS_safe <- function(dist_obj, k, init_coords = NULL,
                            unit, dataset, mode_dim_label) {
  if (k <= 0) {
    warning("[WARN] isoMDS requested with k <= 0 (", unit, ", ", dataset,
            ", ", mode_dim_label, "); skip.")
    return(NULL)
  }
  n <- attr(dist_obj, "Size")
  k_eff <- min(k, n - 1L)
  if (k_eff <= 0) {
    warning("[WARN] isoMDS k_eff <= 0 (", unit, ", ", dataset,
            ", ", mode_dim_label, "); skip.")
    return(NULL)
  }
  message("[INFO] ---- isoMDS (", mode_dim_label, ") 開始: ",
          dataset, " k=", k_eff, " ----")
  if (!is.null(init_coords)) {
    # init_coords は n × p の行列を想定 → 先頭 k_eff 列を初期値として渡す
    init_k <- init_coords[, seq_len(min(ncol(init_coords), k_eff)), drop = FALSE]
    res <- MASS::isoMDS(dist_obj, y = init_k, k = k_eff, maxit = 50)
  } else {
    res <- MASS::isoMDS(dist_obj, k = k_eff, maxit = 50)
  }
  message("[INFO] ---- isoMDS (", mode_dim_label, ") 終了: ",
          dataset, " ----")
  res$points
}

############################################################
# 3. 1データセットの MDS 処理
############################################################

process_one_dataset_mds <- function(unit, dataset) {
  message("\n--------------------------------------------------")
  message("[INFO] Processing MDS for unit=", unit, ", dataset=", dataset)
  message("--------------------------------------------------")
  
  # 距離ファイルのパス
  dist_dir_unit <- file.path(step01_run_dir, unit, dataset)
  if (!dir.exists(dist_dir_unit)) {
    stop("[ERROR] Distance dir not found: ", dist_dir_unit)
  }
  
  if (unit == "variables") {
    dist_file <- file.path(
      dist_dir_unit,
      sprintf("distance_corr_variables_%s_%s.csv", dataset, run_id)
    )
  } else if (unit == "samples") {
    dist_file <- file.path(
      dist_dir_unit,
      sprintf("distance_corr_samples_%s_%s.csv", dataset, run_id)
    )
  } else {
    stop("[ERROR] Unknown unit: ", unit)
  }
  
  message("[INFO] Reading distance file: ", dist_file)
  dist_obj <- read_distance_matrix(dist_file)
  dmat     <- as.matrix(dist_obj)
  n        <- nrow(dmat)
  
  # 出力ディレクトリ
  out_unit_dir <- file.path(step02_run_dir, unit)
  out_ds_dir   <- file.path(out_unit_dir, dataset)
  dir.create(out_ds_dir, showWarnings = FALSE, recursive = TRUE)
  
  ##########################################################
  # 3-1. classical MDS (cmdscale) & 固有値寄与テーブル
  ##########################################################
  
  message("[INFO] ---- 固有値計算 開始: ", dataset, " ----")
  # k = n-1 まで計算（cmdscale の仕様上）
  mds_linear_full <- cmdscale(dist_obj, eig = TRUE, k = n - 1)
  eig_vals <- as.numeric(mds_linear_full$eig)
  message("[INFO] ---- 固有値計算 終了: ", dataset, " ----")
  
  # 固有値ベクトルを保存（互換用）
  if (unit == "variables") {
    eig_file <- file.path(
      out_ds_dir,
      sprintf("MDS_eigen_variables_%s_%s.csv", dataset, run_id)
    )
  } else {
    eig_file <- file.path(
      out_ds_dir,
      sprintf("MDS_eigen_samples_%s_%s.csv", dataset, run_id)
    )
  }
  eig_df <- data.frame(axis = seq_along(eig_vals),
                       eigenvalue = eig_vals)
  write.csv(eig_df, eig_file, row.names = FALSE)
  message("[INFO] Saved eigenvalues → ", eig_file)
  
  # ★ 寄与率テーブル（正の固有値のみ）を作成・保存
  contrib_info <- compute_eigen_contributions(eig_vals, thr = cumeig_thr)
  contrib_df   <- contrib_info$contrib_df
  k_top3       <- contrib_info$k_top3
  k_cumeig_raw <- contrib_info$k_cumeig
  
  if (unit == "variables") {
    eig_contrib_file <- file.path(
      out_ds_dir,
      sprintf("MDS_eigen_contributions_variables_%s_%s.csv", dataset, run_id)
    )
  } else {
    eig_contrib_file <- file.path(
      out_ds_dir,
      sprintf("MDS_eigen_contributions_samples_%s_%s.csv", dataset, run_id)
    )
  }
  write.csv(contrib_df, eig_contrib_file, row.names = FALSE)
  message("[INFO] Saved eigen contributions → ", eig_contrib_file)
  message("[INFO] k_top3 = ", k_top3, ", k_cumeig_raw = ", k_cumeig_raw)
  
  # MDS 座標（全次元）
  coords_linear_full <- mds_linear_full$points   # n × p
  
  ##########################################################
  # 3-2. linear_top3 / linear_cumeig 座標の切り出し
  ##########################################################
  
  get_coords_subset <- function(coords_all, k) {
    if (k <= 0) return(NULL)
    p <- ncol(coords_all)
    k_eff <- min(k, p)
    coords_all[, seq_len(k_eff), drop = FALSE]
  }
  
  coords_linear_top3   <- get_coords_subset(coords_linear_full, k_top3)
  coords_linear_cumeig <- get_coords_subset(coords_linear_full, k_cumeig_raw)
  
  # 保存（共通フォーマット）
  if (!is.null(coords_linear_top3)) {
    # summary 用（互換的、ID 行名付きテーブル）
    lin_top3_file <- file.path(
      out_ds_dir,
      sprintf("MDS_linear_top3_%s_%s.csv", dataset, run_id)
    )
    df_lin_top3 <- as.data.frame(coords_linear_top3)
    df_lin_top3$ID <- rownames(coords_linear_top3)
    df_lin_top3 <- df_lin_top3[, c(ncol(df_lin_top3), 1:(ncol(df_lin_top3)-1))]
    write.csv(df_lin_top3, lin_top3_file, row.names = FALSE)
    
    # clustering 用（MDScoords_...）
    coords_lin_top3_file <- file.path(
      out_ds_dir,
      sprintf("MDScoords_linear_top3_%s_%s_%s_%s.csv",
              unit, dataset, run_id) # 後で sprintf で整形するので分解
    )
    # ここは 03_clustering の期待形式に合わせる → MDScoords_linear_top3_<unit>_<dataset>_<run_id>.csv
    coords_lin_top3_file <- file.path(
      out_ds_dir,
      if (unit == "variables") {
        sprintf("MDScoords_linear_top3_variables_%s_%s.csv", dataset, run_id)
      } else {
        sprintf("MDScoords_linear_top3_samples_%s_%s.csv", dataset, run_id)
      }
    )
    write.csv(coords_linear_top3, coords_lin_top3_file, row.names = TRUE)
    
    message("[INFO] Saved linear_top3 → ",
            lin_top3_file, " , ",
            coords_lin_top3_file)
  }
  
  if (!is.null(coords_linear_cumeig)) {
    lin_cumeig_file <- file.path(
      out_ds_dir,
      sprintf("MDS_linear_cumeig_%s_%s.csv", dataset, run_id)
    )
    df_lin_cumeig <- as.data.frame(coords_linear_cumeig)
    df_lin_cumeig$ID <- rownames(coords_linear_cumeig)
    df_lin_cumeig <- df_lin_cumeig[, c(ncol(df_lin_cumeig), 1:(ncol(df_lin_cumeig)-1))]
    write.csv(df_lin_cumeig, lin_cumeig_file, row.names = FALSE)
    
    coords_lin_cumeig_file <- file.path(
      out_ds_dir,
      if (unit == "variables") {
        sprintf("MDScoords_linear_cumeig_variables_%s_%s.csv", dataset, run_id)
      } else {
        sprintf("MDScoords_linear_cumeig_samples_%s_%s.csv", dataset, run_id)
      }
    )
    write.csv(coords_linear_cumeig, coords_lin_cumeig_file, row.names = TRUE)
    
    message("[INFO] Saved linear_cumeig → ",
            lin_cumeig_file, " , ",
            coords_lin_cumeig_file)
  }
  
  ##########################################################
  # 3-3. isoMDS (nonlinear_top3 / nonlinear_cumeig)
  ##########################################################
  # - variables: top3, cumeig 両方ここで計算
  # - samples : top3 だけここで計算
  #             cumeig は STEP3.2_02b で別途計算（重いので分割）
  ##########################################################
  
  # 共通: nonlinear_top3
  if (k_top3 > 0) {
    coords_nl_top3 <- run_isoMDS_safe(
      dist_obj,
      k = k_top3,
      init_coords = coords_linear_full,
      unit = unit,
      dataset = dataset,
      mode_dim_label = "nonlinear_top3"
    )
  } else {
    coords_nl_top3 <- NULL
  }
  
  if (!is.null(coords_nl_top3)) {
    nl_top3_file <- file.path(
      out_ds_dir,
      sprintf("MDS_nonlinear_top3_%s_%s.csv", dataset, run_id)
    )
    df_nl_top3 <- as.data.frame(coords_nl_top3)
    df_nl_top3$ID <- rownames(coords_nl_top3)
    df_nl_top3 <- df_nl_top3[, c(ncol(df_nl_top3), 1:(ncol(df_nl_top3)-1))]
    write.csv(df_nl_top3, nl_top3_file, row.names = FALSE)
    
    coords_nl_top3_file <- file.path(
      out_ds_dir,
      if (unit == "variables") {
        sprintf("MDScoords_nonlinear_top3_variables_%s_%s.csv", dataset, run_id)
      } else {
        sprintf("MDScoords_nonlinear_top3_samples_%s_%s.csv", dataset, run_id)
      }
    )
    write.csv(coords_nl_top3, coords_nl_top3_file, row.names = TRUE)
    
    message("[INFO] Saved nonlinear_top3 → ",
            nl_top3_file, " , ",
            coords_nl_top3_file)
  }
  
  # nonlinear_cumeig:
  if (unit == "variables") {
    # 変数側はここで cumeig も計算
    if (k_cumeig_raw > 0) {
      coords_nl_cumeig <- run_isoMDS_safe(
        dist_obj,
        k = k_cumeig_raw,
        init_coords = coords_linear_full,
        unit = unit,
        dataset = dataset,
        mode_dim_label = "nonlinear_cumeig"
      )
    } else {
      coords_nl_cumeig <- NULL
    }
    
    if (!is.null(coords_nl_cumeig)) {
      nl_cumeig_file <- file.path(
        out_ds_dir,
        sprintf("MDS_nonlinear_cumeig_%s_%s.csv", dataset, run_id)
      )
      df_nl_cumeig <- as.data.frame(coords_nl_cumeig)
      df_nl_cumeig$ID <- rownames(coords_nl_cumeig)
      df_nl_cumeig <- df_nl_cumeig[, c(ncol(df_nl_cumeig), 1:(ncol(df_nl_cumeig)-1))]
      write.csv(df_nl_cumeig, nl_cumeig_file, row.names = FALSE)
      
      coords_nl_cumeig_file <- file.path(
        out_ds_dir,
        sprintf("MDScoords_nonlinear_cumeig_variables_%s_%s.csv", dataset, run_id)
      )
      write.csv(coords_nl_cumeig, coords_nl_cumeig_file, row.names = TRUE)
      
      message("[INFO] Saved nonlinear_cumeig (variables) → ",
              nl_cumeig_file, " , ",
              coords_nl_cumeig_file)
    }
  } else {
    # samples 側の nonlinear_cumeig はここでは実行しない（02b に任せる）
    message("[INFO] Skip nonlinear_cumeig for samples (handled in STEP3.2_02b)")
  }
  
  message("[DONE] MDS finished for unit = ", unit,
          ", dataset = ", dataset)
}

############################################################
# 4. メインループ
############################################################

cat("\n===== STEP3.2_02_mds: Variables (OH/FP) =====\n")
for (ds in datasets_variables) {
  process_one_dataset_mds("variables", ds)
}

cat("\n===== STEP3.2_02_mds: Samples (A/B/C/D) =====\n")
for (ds in datasets_samples) {
  process_one_dataset_mds("samples", ds)
}

cat("\n✅ STEP3.2_02_mds finished. run_id = ", run_id, "\n")
cat("   Output: ", normalizePath(step02_run_dir), "\n")
```

```r
#!/usr/bin/env Rscript

########################################################################
## STEP3.2_02b_mds_samples_nonlinear_cumeig.R
##
## 目的:
##   ・STEP01 で計算済みの sample-wise 距離から
##     各 samples データセット (A/B/C/D) について
##     「nonlinear_cumeig」の isoMDS 座標だけを計算・保存する。
##
## ポイント:
##   ・k_cumeig_raw (>=0.8, min 3) を決めたあと、
##     isoMDS では k_eff = min(k_cumeig_raw, 50) を使う
##     → 高次元すぎて kernel が止まるのを回避
########################################################################

suppressWarnings({
  library(MASS)      # isoMDS
})
set.seed(42)
#### 0. パス設定 #######################################################
root_base <- "/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/cal_cluster_No/20251026_under_25clusters_program_and_result/for_checking_20251127"
sub_base  <- file.path(root_base, "sub")

step01_base <- file.path(sub_base, "01_distance_STEP3.2_signlessCorr")
step02_base <- file.path(sub_base, "02_mds_STEP3.2_signlessCorr")

datasets_samples <- c("A_OH_plus_others",
                      "B_FP_plus_others",
                      "C_OH_only",
                      "D_FP_only")

#### 1. run_id の取得 ##################################################
## ★ 必ず 01/02/03/04 で同じ値にそろえる ★
override_run_id <- "20251130_153500"  # ← 自分の run_id に書き換え

get_run_id <- function() {
  if (!is.na(override_run_id) &&
      is.character(override_run_id) &&
      grepl("^\\d{8}_\\d{6}$", override_run_id)) {
    message("[INFO] Using override_run_id = ", override_run_id)
    return(override_run_id)
  }
  stop("ERROR: Please set override_run_id in the script as YYYYMMDD_HHMMSS.")
}

run_id <- get_run_id()
cat(">>> STEP3.2_02b_mds_samples_nonlinear_cumeig — run_id:", run_id, "\n\n")

#### 2. 出力ベースフォルダ ############################################
out_base <- file.path(step02_base, paste0("run_", run_id))
dir.create(out_base, recursive = TRUE, showWarnings = FALSE)

#### 3. k_cumeig の決定関数 ###########################################
decide_k_cumeig <- function(eig_vals, thr = 0.8, k_min = 3) {
  pos_vals <- eig_vals[eig_vals > 0]
  if (length(pos_vals) == 0) return(k_min)
  contrib <- pos_vals / sum(pos_vals)
  k <- which(cumsum(contrib) >= thr)[1]
  if (is.na(k) || k < k_min) k <- k_min
  return(k)
}

#### 4. isoMDS 用の安定化付きラッパ ##################################
compute_isoMDS_safe <- function(dist_obj, k, tag = "") {

  message("[INFO] ---- isoMDS (nonlinear_cumeig) 開始: ", tag,
          "  k=", k, " ----")

  mat <- as.matrix(dist_obj)
  n   <- nrow(mat)

  diag(mat) <- 0

  off_diag <- mat[upper.tri(mat)]
  pos_vals <- off_diag[off_diag > 0]

  if (length(pos_vals) == 0) {
    warning("[WARN] All off-diagonal distances are zero or NA for ", tag,
            " → isoMDS をスキップします。")
    return(NULL)
  }

  min_pos <- min(pos_vals, na.rm = TRUE)
  eps <- min_pos * 1e-4
  if (!is.finite(eps) || eps <= 0) eps <- 1e-6

  for (i in 1:(n - 1)) {
    for (j in (i + 1):n) {
      if (!is.na(mat[i, j]) && mat[i, j] <= 0) {
        mat[i, j] <- eps
        mat[j, i] <- eps
      }
    }
  }

    dist_fixed <- as.dist(mat)
  
  # ★★★ 追加: 初期値 (cmdscale) を計算 ★★★
  # ランダムスタートを防ぎ、02_mds.R と挙動を合わせる
  init_coords <- cmdscale(dist_fixed, k = k)
  # ★★★★★★★★★★★★★★★★★★★★★★★

  res <- tryCatch(
    {
      # ★★★ 変更: y = init_coords を追加 ★★★
      isoMDS(dist_fixed, y = init_coords, k = k, maxit = 50)
    },
    error = function(e) {
      warning("[WARN] isoMDS failed for ", tag, ": ", e$message,
              " → nonlinear_cumeig をスキップします。")
      return(NULL)
    }
  )

  if (is.null(res)) {
    message("[INFO] ---- isoMDS (nonlinear_cumeig) スキップ: ", tag, " ----")
  } else {
    message("[INFO] ---- isoMDS (nonlinear_cumeig) 終了: ", tag, " ----")
  }

  return(res)
}

#### 5. 1 データセット分の処理 #######################################
process_one_sample_ds <- function(ds) {

  message("==================================================")
  message("[INFO] Processing samples / dataset = ", ds)
  message("==================================================")

  ## 距離ファイル
  dist_file <- file.path(step01_base,
                         paste0("run_", run_id),
                         "samples", ds,
                         paste0("distance_corr_samples_", ds, "_", run_id, ".csv"))

  if (!file.exists(dist_file)) {
    warning("[WARN] Distance file not found → SKIP: ", dist_file)
    return(NULL)
  }

  message("[INFO] Reading distance file: ", dist_file)
  dmat <- read.csv(dist_file, row.names = 1, check.names = FALSE)
  dmat <- as.matrix(dmat)
  dist_obj <- as.dist(dmat)

  ## 固有値から k_cumeig_raw を決定（再計算）
  message("[INFO] ---- 固有値計算 (for k_cumeig) 開始: ", ds, " ----")
  eig_res  <- cmdscale(dist_obj, eig = TRUE, k = nrow(dmat) - 1)
  eig_vals <- eig_res$eig
  k_cumeig_raw <- decide_k_cumeig(eig_vals, thr = 0.8, k_min = 3)

  ## samples の isoMDS 用には上限をかける（例: 50 次元）
  k_max_samples_nonlinear <- 50
  k_eff <- min(k_cumeig_raw, k_max_samples_nonlinear)

  message("[INFO] ---- 固有値計算 (for k_cumeig) 終了: ", ds,
          "  → k_cumeig_raw = ", k_cumeig_raw,
          ", k_eff_for_isoMDS = ", k_eff, " ----")

  ## isoMDS(nonlinear_cumeig) with k_eff
  tag <- paste0(ds, " nonlinear_cumeig (k_eff=", k_eff, ")")
  iso_res <- compute_isoMDS_safe(dist_obj, k = k_eff, tag = tag)
  if (is.null(iso_res)) {
    warning("[WARN] No isoMDS result for ", ds, " (nonlinear_cumeig).")
    return(NULL)
  }

  ## 出力フォルダ & 保存
  out_dir <- file.path(out_base, "samples", ds)
  dir.create(out_dir, recursive = TRUE, showWarnings = FALSE)

  out_file <- file.path(out_dir,
                        paste0("MDS_nonlinear_cumeig_", ds, "_", run_id, ".csv"))
  write.csv(iso_res$points, out_file)

  message("[INFO] Saved nonlinear_cumeig coords → ", out_file)
}

#### 6. A〜D を順番に処理 #############################################
for (ds in datasets_samples) {
  process_one_sample_ds(ds)
}

message("\n>>> STEP3.2_02b samples_nonlinear_cumeig 完了 (run_id = ", run_id, ")\n")
```
