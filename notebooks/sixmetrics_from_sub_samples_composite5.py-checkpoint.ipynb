```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

"""
sixmetrics_from_sub_samples_composite5.py
----------------------------------------
sub/03_clustering_STEP3.2_signlessCorr/run_<RUN_ID>/samples/
  ├─ A_OH_plus_others/cluster_labels_matrix_samples_A_OH_plus_others_<RUN_ID>.csv
  └─ B_FP_plus_others/cluster_labels_matrix_samples_B_FP_plus_others_<RUN_ID>.csv

と、

for_MDS_STEP2/preprocessed_features_{OH,FP}.csv
  （行=サンプル, 列=変数）

を用いて、サンプル方向の six-metrics + composite5 を計算する。

  - direction: OH→FP / FP→OH
  - mode: cluster_labels_matrix の列名プレフィックス
      例）nonlinear_cumeig_silhouette → mode="nonlinear_cumeig", index="silhouette"
  - index: NbClust 指標（silhouette, dunn, gap, ch, db, ptbiserial）

出力:
  <BASE>/plots_sixmetrics_samples_sub/Table_sixmetrics_samples_bidirectional_composite5_sub.csv
"""

from pathlib import Path
import argparse
import re
import itertools

import numpy as np
import pandas as pd


# =========================
# 既定パス
# =========================
DEFAULT_BASE = (
    "/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/"
    "cal_cluster_No/20251026_under_25clusters_program_and_result/"
    "for_checking_20251127/sub/03_clustering_STEP3.2_signlessCorr"
)

DEFAULT_RUN_ID = "20251130_153500"

DEFAULT_FEAT_ROOT = (
    "/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/"
    "cal_cluster_No/20251026_under_25clusters_program_and_result/"
    "for_checking_20251127/data/for_MDS_STEP2"
)

# サンプルクラスタのデータセット名（フォルダ名の一部）
DATASET_OH = "A_OH_plus_others"
DATASET_FP = "B_FP_plus_others"

# NbClust 指標の順序
IDX_ORDER = ["silhouette", "dunn", "gap", "ch", "db", "ptbiserial"]

# 目的変数（特徴量から除外）
TARGETS = {"PCEmax", "Jsc", "Voc", "FF"}


def debug(msg: str):
    print(f"[DEBUG] {msg}")


# =========================
# サンプルクラスタ割当の読み込み
# =========================
def get_cluster_matrix_path(base: Path, run_id: str, dataset: str) -> Path:
    """
    例:
      base/run_20251130_153500/samples/A_OH_plus_others/
        cluster_labels_matrix_samples_A_OH_plus_others_20251130_153500.csv
    """
    fname = f"cluster_labels_matrix_samples_{dataset}_{run_id}.csv"
    return base / f"run_{run_id}" / "samples" / dataset / fname


def read_assignments_from_cluster_matrix(base: Path,
                                         run_id: str,
                                         dataset: str) -> pd.DataFrame:
    """
    cluster_labels_matrix_samples_<dataset>_<run_id>.csv を読み、
    次の tidy 形式に変換する：

      - id: サンプルID
      - cluster: クラスタ番号（整数）
      - mode: 列名プレフィックス（例: "nonlinear_cumeig"）
      - index: NbClust 指標名（silhouette, dunn, gap, ch, db, ptbiserial）

    例）列 "nonlinear_cumeig_silhouette" →
        mode="nonlinear_cumeig", index="silhouette"
    """
    path = get_cluster_matrix_path(base, run_id, dataset)
    if not path.exists():
        debug(f"Cluster matrix file missing for {dataset}: {path}")
        return pd.DataFrame(columns=["id", "cluster", "mode", "index"])

    debug(f"Reading cluster matrix for {dataset}: {path}")
    df = pd.read_csv(path)

    # id 列を決定
    id_col = None
    for c in df.columns:
        cl = str(c).lower()
        if cl in ["id", "sample", "sample_id", "sno"]:
            id_col = c
            break
    if id_col is None:
        id_col = df.columns[0]

    ids = df[id_col].astype(str).tolist()

    rows = []
    for col in df.columns:
        if col == id_col:
            continue
        # パターン: <mode>_<index>
        m = re.match(r"(.+?)_(silhouette|dunn|gap|ch|db|ptbiserial)$", str(col))
        if not m:
            # 対象外の列（例えば別メタ情報など）はスキップ
            continue
        mode = m.group(1)      # 例: "nonlinear_cumeig"
        index = m.group(2)     # 例: "silhouette"
        clab = pd.to_numeric(df[col], errors="coerce").astype("Int64").tolist()
        rows.append(
            pd.DataFrame(
                {"id": ids, "cluster": clab, "mode": mode, "index": index}
            )
        )

    if not rows:
        debug(f"No usable cluster columns found for {dataset}")
        return pd.DataFrame(columns=["id", "cluster", "mode", "index"])

    A = pd.concat(rows, ignore_index=True)
    A = A.dropna(subset=["cluster"])
    A["cluster"] = A["cluster"].astype(int)
    A["id"] = A["id"].astype(str)
    return A


# =========================
# 特徴量（サンプル行）読み込み
# =========================
def load_features_rows_as_samples(path: Path) -> pd.DataFrame:
    """
    preprocessed_features_{OH,FP}.csv を読み、
    行=サンプル, 列=変数 として処理。
    - 目的変数列（PCEmax,Jsc,Voc,FF）は除外
    - NaN → 0, 負値 → 0
    """
    if not path.exists():
        debug(f"Feature file missing: {path}")
        return pd.DataFrame()

    df = pd.read_csv(path, header=0, index_col=0)
    keep = [c for c in df.columns if c not in TARGETS]
    df = df.loc[:, keep]

    # 数値変換 + NaN → 0, 負値 → 0
    for c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")
    df = df.fillna(0.0)
    df[df < 0.0] = 0.0

    df.index = df.index.astype(str)
    return df


def row_normalize_to_prob(X: pd.DataFrame) -> pd.DataFrame:
    """
    各行（=サンプル）を確率分布に正規化（sum=1）。全0行はそのまま0。
    """
    S = X.sum(axis=1)
    S = S.replace(0, np.nan)
    Y = X.div(S, axis=0).fillna(0.0)
    return Y


# =========================
# 距離・情報量系ヘルパ
# =========================
def cosine(u: np.ndarray, v: np.ndarray) -> float:
    nu = np.linalg.norm(u)
    nv = np.linalg.norm(v)
    if nu == 0 or nv == 0:
        return 0.0
    return float(np.dot(u, v) / (nu * nv))


def js_divergence(p: np.ndarray, q: np.ndarray) -> float:
    """
    Jensen-Shannon divergence (log2).
    p, q: 確率分布（要素 >=0, sum=1 想定）
    """
    m = 0.5 * (p + q)

    def kl(a, b):
        nz = (a > 0) & (b > 0)
        return np.sum(a[nz] * np.log2(a[nz] / b[nz]))

    return 0.5 * kl(p, m) + 0.5 * kl(q, m)


def entropy_from_counts(counts: pd.Series) -> float:
    """
    離散分布の Shannon エントロピー（log2）。
    counts: Rラベルの個数など
    """
    tot = counts.sum()
    if tot <= 0:
        return 0.0
    p = counts[counts > 0] / tot
    return float(-np.sum(p * np.log2(p)))


# =========================
# 6 指標（1方向）
# =========================
def six_metrics_direction(left_assign: pd.DataFrame,
                          right_assign: pd.DataFrame,
                          right_vec_prob: pd.DataFrame,
                          mode: str,
                          index: str,
                          thr=(0.6, 0.6, 0.5)):
    """
    left_assign のクラスタ（id=サンプル）を right 側で眺めたときの整合度（6指標）

      - mean_purity:
          各 left クラスタ内での right クラスタラベル分布の最大比率の平均
      - pair_major_same_rate:
          left クラスタ内ペアにおいて、right ラベル一致の割合
      - pair_mean_cosine_similarity:
          left クラスタ内ペアの right_vec 確率分布のコサイン平均
      - cluster_median_cohesive_ratio:
          left クラスタ内ペアのうち、
            (R同一 & purity>=thr0) or (cos>=thr1) or (JS<=thr2)
          を満たす割合の中央値
      - mean_entropy:
          right ラベル分布のエントロピー平均
      - pair_mean_JS_divergence:
          left クラスタ内ペアの JS ダイバージェンス平均
    """
    L = left_assign[(left_assign["mode"] == mode) &
                    (left_assign["index"] == index)][["id", "cluster"]].rename(columns={"cluster": "L"})
    R = right_assign[(right_assign["mode"] == mode) &
                     (right_assign["index"] == index)][["id", "cluster"]].rename(columns={"cluster": "R"})
    M = pd.merge(L, R, on="id", how="inner")
    if M.empty:
        return None

    pur_list, ent_list = [], []
    cos_vals, js_vals = [], []
    coh_ratio = []
    pair_same_n = 0
    pair_tot_n = 0

    for l, grp in M.groupby("L"):
        # Rラベル分布 → purity / entropy
        vc = grp["R"].value_counts()
        purity = vc.max() / vc.sum() if vc.sum() > 0 else np.nan
        ent = entropy_from_counts(vc)
        pur_list.append(purity)
        ent_list.append(ent)

        # L内ペアでコサイン・JS・R同一・cohesive
        ids = grp["id"].tolist()
        if len(ids) >= 2:
            sub = right_vec_prob.loc[right_vec_prob.index.intersection(ids)]
            mat = sub.to_numpy()
            same = 0
            total = 0
            coh = 0
            for i, j in itertools.combinations(range(mat.shape[0]), 2):
                total += 1
                same += int(grp["R"].iloc[i] == grp["R"].iloc[j])
                c = cosine(mat[i], mat[j])
                js = js_divergence(mat[i], mat[j])
                cos_vals.append(c)
                js_vals.append(js)

                cond1 = (grp["R"].iloc[i] == grp["R"].iloc[j]) and (purity >= thr[0])
                cond2 = (c >= thr[1])
                cond3 = (js <= thr[2])
                if cond1 or cond2 or cond3:
                    coh += 1

            pair_same_n += same
            pair_tot_n += total
            if total > 0:
                coh_ratio.append(coh / total)

    return {
        "mean_purity": np.nanmean(pur_list) if pur_list else np.nan,
        "pair_major_same_rate": (pair_same_n / pair_tot_n) if pair_tot_n > 0 else np.nan,
        "pair_mean_cosine_similarity": np.nanmean(cos_vals) if cos_vals else np.nan,
        "cluster_median_cohesive_ratio": np.nanmedian(coh_ratio) if coh_ratio else np.nan,
        "mean_entropy": np.nanmean(ent_list) if ent_list else np.nan,
        "pair_mean_JS_divergence": np.nanmean(js_vals) if js_vals else np.nan,
        "k_left": M["L"].nunique(),
        "k_right": M["R"].nunique(),
        "n": len(M),
    }


# =========================
# メイン
# =========================
def main():
    ap = argparse.ArgumentParser()
    ap.add_argument("--base", type=str, default=DEFAULT_BASE,
                    help="STEP3.2_signlessCorr の sub フォルダ（run_xxx の親）")
    ap.add_argument("--run_id", type=str, default=DEFAULT_RUN_ID,
                    help="run_<RUN_ID> の RUN_ID")
    ap.add_argument("--featroot", type=str, default=DEFAULT_FEAT_ROOT,
                    help="preprocessed_features_{OH,FP}.csv が置いてある for_MDS_STEP2 のパス")
    args, _ = ap.parse_known_args()

    BASE = Path(args.base).resolve()
    RUN_ID = args.run_id
    FEAT_ROOT = Path(args.featroot).resolve()

    print("[INFO] BASE:", BASE)
    print("[INFO] RUN_ID:", RUN_ID)
    print("[INFO] FEAT_ROOT:", FEAT_ROOT)

    OUT_DIR = BASE / "run_{}".format(RUN_ID) / "plots_sixmetrics_samples_sub"
    OUT_DIR.mkdir(parents=True, exist_ok=True)
    print("[INFO] Output dir:", OUT_DIR)

    # 1) サンプルクラスタ割り当て読み込み
    print("[INFO] Reading sample-wise assignments from cluster_labels_matrix (OH/FP)...")
    A_OH = read_assignments_from_cluster_matrix(BASE, RUN_ID, DATASET_OH)
    A_FP = read_assignments_from_cluster_matrix(BASE, RUN_ID, DATASET_FP)
    print("[INFO] OH assignments:", A_OH.shape)
    print("[INFO] FP assignments:", A_FP.shape)

    if A_OH.empty or A_FP.empty:
        print("❌ サンプル割当が読み込めませんでした。パス・RUN_ID・ファイル名を確認してください。")
        return

    # 2) 特徴量（行=サンプル）読み込み → 行確率化
    X_OH = load_features_rows_as_samples(FEAT_ROOT / "preprocessed_features_OH.csv")
    X_FP = load_features_rows_as_samples(FEAT_ROOT / "preprocessed_features_FP.csv")
    print("[INFO] Features (samples x variables): OH", X_OH.shape, "FP", X_FP.shape)

    if X_OH.empty or X_FP.empty:
        print("❌ 特徴量CSVが読み込めませんでした。--featroot を確認してください。")
        return

    # 共通サンプルIDのみ使用
    common_ids = sorted(set(A_OH["id"]) & set(A_FP["id"]) &
                        set(X_OH.index.astype(str)) & set(X_FP.index.astype(str)))
    print(f"[INFO] common ids (samples) = {len(common_ids)}")
    if len(common_ids) == 0:
        print("❌ 共通サンプルIDが 0 件です。割当CSVの id と特徴量CSVの行名が一致しているか確認してください。")
        return

    A_OH = A_OH[A_OH["id"].isin(common_ids)].copy()
    A_FP = A_FP[A_FP["id"].isin(common_ids)].copy()
    X_OH = X_OH.loc[common_ids]
    X_FP = X_FP.loc[common_ids]

    P_OH = row_normalize_to_prob(X_OH)
    P_FP = row_normalize_to_prob(X_FP)

    # 利用可能な mode 一覧（cluster_labels_matrix から自動取得）
    modes = sorted(set(A_OH["mode"]).intersection(set(A_FP["mode"])))
    print("[INFO] modes (common between OH/FP):", modes)

    rows = []
    for md in modes:
        for ix in IDX_ORDER:
            a = six_metrics_direction(A_OH, A_FP, P_FP, md, ix)  # OH→FP
            b = six_metrics_direction(A_FP, A_OH, P_OH, md, ix)  # FP→OH
            if a:
                rows.append({"direction": "OH→FP", "mode": md, "index": ix, **a})
            if b:
                rows.append({"direction": "FP→OH", "mode": md, "index": ix, **b})

    if not rows:
        print("❌ 6指標が計算できませんでした（mode/index の組み合わせが存在しない可能性）。")
        return

    MET = pd.DataFrame(rows)

    # 3) entropy_coh, js_coh, composite5
    with np.errstate(divide="ignore", invalid="ignore"):
        kr = MET["k_right"].clip(lower=2)
        maxH = np.log2(kr)
        MET["entropy_coh"] = 1.0 - (MET["mean_entropy"] / maxH)
    MET["entropy_coh"] = MET["entropy_coh"].clip(0, 1)

    MET["js_coh"] = 1.0 - MET["pair_mean_JS_divergence"]
    MET["js_coh"] = MET["js_coh"].clip(0, 1)

    MET["composite5"] = MET[
        [
            "mean_purity",
            "pair_major_same_rate",
            "pair_mean_cosine_similarity",
            "cluster_median_cohesive_ratio",
            "entropy_coh",
        ]
    ].mean(axis=1)

    # 4) 保存
    out_csv = OUT_DIR / "Table_sixmetrics_samples_bidirectional_composite5_sub.csv"
    MET.to_csv(out_csv, index=False, encoding="utf-8-sig")
    print("✅ Saved table:", out_csv)


if __name__ == "__main__":
    main()
```
