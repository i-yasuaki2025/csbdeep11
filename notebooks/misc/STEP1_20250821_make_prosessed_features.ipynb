```r
# --- 必要パッケージ ---
library(dplyr)
library(tidyr)
library(ggplot2)

# =========================
# 入力ファイル
# =========================
data_file    <- "DataMerge20211220oh_20250717_OH_delSMILES.csv"  # 入力CSV
cluster_file <- "cluster_assignment_OH.csv"                       # クラスタ割当（OH版）

# =========================
# データ読み込み & 前処理
# =========================
feature_df <- read.csv(data_file, row.names = 1)

# 目的変数を除去
target_vars <- c("PCEmax", "Jsc", "Voc", "FF")
expl_vars <- feature_df[, !(colnames(feature_df) %in% target_vars), drop = FALSE]

# 文字列列の除外（英字を含む値がある列を除去。文字列の "NA" は無視）
is_char <- sapply(expl_vars, function(col) any(grepl("[A-Za-z]", col) & col != "NA", na.rm = TRUE))
expl_vars <- expl_vars[, !is_char, drop = FALSE]

# --- SD=0の確認関数 ---
check_sd0_vars <- function(df, label) {
  sd0_vars <- colnames(df)[apply(df, 2, sd, na.rm = TRUE) == 0]
  if (length(sd0_vars) > 0) {
    cat("\n⚠️", label, "でSD=0の変数:\n", paste(sd0_vars, collapse = ", "), "\n")
  } else {
    cat("\n✅", label, "にはSD=0の変数はありません\n")
  }
}

# SD=0の確認 & 除去（ログをCSV出力）
check_sd0_vars(expl_vars, "expl_vars (raw)")
sd_nonzero <- apply(expl_vars, 2, sd, na.rm = TRUE) != 0
numData <- expl_vars[, sd_nonzero, drop = FALSE]
removed_sd0_vars <- setdiff(colnames(expl_vars), colnames(numData))
write.csv(data.frame(Removed_SD0 = removed_sd0_vars),
          "removed_sd0_vars_OH.csv", row.names = FALSE)

# スケーリング
numData_scaled <- scale(numData)

# 完全相関（|r|=1）の片側除去（pairwise 相関）
cor_mat <- cor(numData_scaled, use = "pairwise.complete.obs")
perfect_corr <- which(abs(cor_mat) == 1 & lower.tri(cor_mat), arr.ind = TRUE)

if (nrow(perfect_corr) > 0) {
  removed_pairs <- unique(data.frame(
    Representative = colnames(cor_mat)[perfect_corr[, 1]],
    Removed       = colnames(cor_mat)[perfect_corr[, 2]]
  ))
  write.csv(removed_pairs, "removed_perfect_corr_pairs_OH.csv", row.names = FALSE)
  vars_to_remove <- unique(removed_pairs$Removed)
  numData_filtered <- numData_scaled[, !colnames(numData_scaled) %in% vars_to_remove, drop = FALSE]
} else {
  cat("✅ 完全相関で除外すべきペアはありません\n")
  numData_filtered <- numData_scaled
}

# 前処理後データフレーム
feature_df_proc <- as.data.frame(numData_filtered)

# === 前処理済みデータのCSV出力 ===
out_csv <- "preprocessed_features_OH.csv"
write.csv(feature_df_proc, out_csv, row.names = TRUE)
cat("✅ 前処理済みデータを保存しました:", out_csv, "\n")

# =========================
# クラスタ割当の読み込み（無ければ自動生成）
# =========================
set.seed(42)

if (file.exists(cluster_file)) {
  cluster_df <- read.csv(cluster_file, row.names = 1)
  colnames(cluster_df) <- "Cluster"
  cat("✅ 既存のクラスタ割当を読み込みました：", cluster_file, "\n")
} else {
  cat("⚠️ クラスタ割当が見つかりません。Ward.D2, k=3 で簡易クラスタを作成します。\n")
  # 単純補完（列中央値）→ 標準化 → 距離 → 階層クラスタリング
  feat_imp <- feature_df_proc
  for (j in seq_len(ncol(feat_imp))) {
    med_j <- median(feat_imp[[j]], na.rm = TRUE)
    feat_imp[[j]][is.na(feat_imp[[j]])] <- med_j
  }
  x_scaled <- scale(feat_imp)
  d  <- dist(x_scaled, method = "euclidean")
  hc <- hclust(d, method = "ward.D2")
  k <- 3
  cl <- cutree(hc, k = k)
  cluster_df <- data.frame(Cluster = cl)
  write.csv(cluster_df, cluster_file)
  cat("✅", cluster_file, "を出力しました（k=3）\n")
}

# =========================
# クラスタ平均の可視化
# =========================
feature_df_proc$SampleID <- rownames(feature_df_proc)
cluster_df$SampleID      <- rownames(cluster_df)

merged_df <- merge(feature_df_proc, cluster_df, by = "SampleID")
merged_df$Cluster <- as.factor(merged_df$Cluster)

# クラスタごとの平均
cluster_mean <- merged_df %>%
  select(-SampleID) %>%
  group_by(Cluster) %>%
  summarise(across(everything(), ~ mean(.x, na.rm = TRUE)))

# 長い形式へ
long_cluster_mean <- cluster_mean %>%
  pivot_longer(cols = -Cluster, names_to = "Feature", values_to = "Mean")

# 変動が大きい上位特徴量（クラスタ間分散が大きい順）
top_features <- long_cluster_mean %>%
  group_by(Feature) %>%
  summarise(VarAcrossClusters = var(Mean)) %>%
  arrange(desc(VarAcrossClusters)) %>%
  slice_head(n = 20) %>%
  pull(Feature)

plot_data <- long_cluster_mean %>%
  filter(Feature %in% top_features)

# 可視化
ggplot(plot_data, aes(x = Cluster, y = Mean, group = Feature, color = Feature)) +
  geom_line() +
  facet_wrap(~ Feature, scales = "free_y") +
  theme_minimal() +
  labs(title = "Cluster-wise Feature Averages (after SD=0 & perfect-corr filtering)",
       y = "Mean", x = "Cluster")
```

```r
# --- 必要パッケージ ---
library(dplyr)
library(tidyr)
library(ggplot2)

# =========================
# 入力ファイル
# =========================
data_file    <- "DataMerge20211220FP_20250715.csv"   # 入力CSV
cluster_file <- "cluster_assignment_FP.csv"          # クラスタ割当（FP版）

# =========================
# データ読み込み & 前処理
# =========================
feature_df <- read.csv(data_file, row.names = 1)

# 目的変数を除去
target_vars <- c("PCEmax", "Jsc", "Voc", "FF")
expl_vars <- feature_df[, !(colnames(feature_df) %in% target_vars), drop = FALSE]

# 文字列列の除外（英字を含む値がある列を除去。文字列の "NA" は無視）
is_char <- sapply(expl_vars, function(col) any(grepl("[A-Za-z]", col) & col != "NA", na.rm = TRUE))
expl_vars <- expl_vars[, !is_char, drop = FALSE]

# --- SD=0の確認関数 ---
check_sd0_vars <- function(df, label) {
  sd0_vars <- colnames(df)[apply(df, 2, sd, na.rm = TRUE) == 0]
  if (length(sd0_vars) > 0) {
    cat("\n⚠️", label, "でSD=0の変数:\n", paste(sd0_vars, collapse = ", "), "\n")
  } else {
    cat("\n✅", label, "にはSD=0の変数はありません\n")
  }
}

# SD=0の確認 & 除去（ログをCSV出力）
check_sd0_vars(expl_vars, "expl_vars (raw)")
sd_nonzero <- apply(expl_vars, 2, sd, na.rm = TRUE) != 0
numData <- expl_vars[, sd_nonzero, drop = FALSE]
removed_sd0_vars <- setdiff(colnames(expl_vars), colnames(numData))
write.csv(data.frame(Removed_SD0 = removed_sd0_vars),
          "removed_sd0_vars_FP.csv", row.names = FALSE)

# スケーリング
numData_scaled <- scale(numData)

# 完全相関（|r|=1）の片側除去（pairwise 相関）
cor_mat <- cor(numData_scaled, use = "pairwise.complete.obs")
perfect_corr <- which(abs(cor_mat) == 1 & lower.tri(cor_mat), arr.ind = TRUE)

if (nrow(perfect_corr) > 0) {
  removed_pairs <- unique(data.frame(
    Representative = colnames(cor_mat)[perfect_corr[, 1]],
    Removed       = colnames(cor_mat)[perfect_corr[, 2]]
  ))
  write.csv(removed_pairs, "removed_perfect_corr_pairs_FP.csv", row.names = FALSE)
  vars_to_remove <- unique(removed_pairs$Removed)
  numData_filtered <- numData_scaled[, !colnames(numData_scaled) %in% vars_to_remove, drop = FALSE]
} else {
  cat("✅ 完全相関で除外すべきペアはありません\n")
  numData_filtered <- numData_scaled
}

# 前処理後データフレーム
feature_df_proc <- as.data.frame(numData_filtered)

# === 前処理済みデータのCSV出力 ===
out_csv <- "preprocessed_features_FP.csv"
write.csv(feature_df_proc, out_csv, row.names = TRUE)
cat("✅ 前処理済みデータを保存しました:", out_csv, "\n")

# =========================
# クラスタ割当の読み込み（無ければ自動生成）
# =========================
set.seed(42)

if (file.exists(cluster_file)) {
  cluster_df <- read.csv(cluster_file, row.names = 1)
  colnames(cluster_df) <- "Cluster"
  cat("✅ 既存のクラスタ割当を読み込みました：", cluster_file, "\n")
} else {
  cat("⚠️ クラスタ割当が見つかりません。Ward.D2, k=3 で簡易クラスタを作成します。\n")
  # 単純補完（列中央値）→ 標準化 → 距離 → 階層クラスタリング
  feat_imp <- feature_df_proc
  for (j in seq_len(ncol(feat_imp))) {
    med_j <- median(feat_imp[[j]], na.rm = TRUE)
    feat_imp[[j]][is.na(feat_imp[[j]])] <- med_j
  }
  x_scaled <- scale(feat_imp)
  d  <- dist(x_scaled, method = "euclidean")
  hc <- hclust(d, method = "ward.D2")
  k <- 
  cl <- cutree(hc, k = k)
  cluster_df <- data.frame(Cluster = cl)
  write.csv(cluster_df, cluster_file)
  cat("✅", cluster_file, "を出力しました（k=3）\n")
}

# =========================
# クラスタ平均の可視化
# =========================
feature_df_proc$SampleID <- rownames(feature_df_proc)
cluster_df$SampleID      <- rownames(cluster_df)

merged_df <- merge(feature_df_proc, cluster_df, by = "SampleID")
merged_df$Cluster <- as.factor(merged_df$Cluster)

# クラスタごとの平均
cluster_mean <- merged_df %>%
  select(-SampleID) %>%
  group_by(Cluster) %>%
  summarise(across(everything(), ~ mean(.x, na.rm = TRUE)))

# 長い形式へ
long_cluster_mean <- cluster_mean %>%
  pivot_longer(cols = -Cluster, names_to = "Feature", values_to = "Mean")

# 変動が大きい上位特徴量（クラスタ間分散が大きい順）
top_features <- long_cluster_mean %>%
  group_by(Feature) %>%
  summarise(VarAcrossClusters = var(Mean)) %>%
  arrange(desc(VarAcrossClusters)) %>%
  slice_head(n = 20) %>%
  pull(Feature)

plot_data <- long_cluster_mean %>%
  filter(Feature %in% top_features)

# 可視化
ggplot(plot_data, aes(x = Cluster, y = Mean, group = Feature, color = Feature)) +
  geom_line() +
  facet_wrap(~ Feature, scales = "free_y") +
  theme_minimal() +
  labs(title = "Cluster-wise Feature Averages (after SD=0 & perfect-corr filtering)",
       y = "Mean", x = "Cluster")
```

```r
# --- 必要パッケージ ---
library(dplyr)
library(tidyr)
library(ggplot2)
if (!require(igraph)) { install.packages("igraph"); library(igraph) }

# =========================
# 入力ファイル
# =========================
data_file    <- "DataMerge20211220oh_20250717_OH_delSMILES.csv"  # 入力CSV
cluster_file <- "cluster_assignment_OH.csv"                       # クラスタ割当（OH版）

# =========================
# データ読み込み & 前処理
# =========================
feature_df <- read.csv(data_file, row.names = 1)

# 目的変数を除去
target_vars <- c("PCEmax", "Jsc", "Voc", "FF")
expl_vars <- feature_df[, !(colnames(feature_df) %in% target_vars), drop = FALSE]

# 文字列列の除外（英字を含む値がある列を除去。文字列の "NA" は無視）
is_char <- sapply(expl_vars, function(col) any(grepl("[A-Za-z]", col) & col != "NA", na.rm = TRUE))
expl_vars <- expl_vars[, !is_char, drop = FALSE]

# --- SD=0の確認関数 ---
check_sd0_vars <- function(df, label) {
  sd0_vars <- colnames(df)[apply(df, 2, sd, na.rm = TRUE) == 0]
  if (length(sd0_vars) > 0) {
    cat("\n⚠️", label, "でSD=0の変数:\n", paste(sd0_vars, collapse = ", "), "\n")
  } else {
    cat("\n✅", label, "にはSD=0の変数はありません\n")
  }
}

# SD=0の確認 & 除去（ログをCSV出力）
check_sd0_vars(expl_vars, "expl_vars (raw)")
sd_nonzero <- apply(expl_vars, 2, sd, na.rm = TRUE) != 0
numData <- expl_vars[, sd_nonzero, drop = FALSE]
removed_sd0_vars <- setdiff(colnames(expl_vars), colnames(numData))
write.csv(data.frame(Removed_SD0 = removed_sd0_vars),
          "removed_sd0_vars_OH.csv", row.names = FALSE)

# スケーリング
numData_scaled <- scale(numData)

# =========================
# 完全相関（±1）の「グループ」から代表1つ以外を除去  <<<< ここが新規ブロック
# =========================
# 相関（pairwise）を計算（スケーリング後データを使用）
cor_mat <- cor(numData_scaled, use = "pairwise.complete.obs")

# （任意）±1扱いとするための最小共通サンプル数（小さすぎると偶然±1になりやすい）
min_pair_n <- 3
pair_n <- outer(colnames(numData), colnames(numData), Vectorize(function(a, b) {
  sum(complete.cases(numData[, c(a, b)]))
}))

# ±1 & 下三角のみを抽出、かつ共通サンプル数がしきい値以上
mask_edges <- (abs(cor_mat) == 1) & lower.tri(cor_mat) & (pair_n >= min_pair_n)
edge_idx <- which(mask_edges, arr.ind = TRUE)

if (nrow(edge_idx) > 0) {
  edge_list <- data.frame(
    from = colnames(cor_mat)[edge_idx[, 1]],
    to   = colnames(cor_mat)[edge_idx[, 2]],
    stringsAsFactors = FALSE
  )
  g <- igraph::graph_from_data_frame(
    edge_list, directed = FALSE,
    vertices = data.frame(name = colnames(cor_mat))
  )

  comp <- igraph::components(g)$membership
  comp_df <- data.frame(Variable = names(comp), Component = as.integer(comp), row.names = NULL)

  # 代表の選び方：欠損が最も少ない列（同数なら名前で最小）
  na_counts <- colSums(is.na(numData[, comp_df$Variable, drop = FALSE]))
  comp_df$NAcount <- na_counts[comp_df$Variable]

  pick_representative <- function(df) {
    df2 <- df[order(df$NAcount, df$Variable), , drop = FALSE]
    df2$Variable[1]
  }
  reps <- tapply(
    X        = seq_len(nrow(comp_df)),
    INDEX    = comp_df$Component,
    FUN      = function(idx) pick_representative(comp_df[idx, , drop = FALSE])
  )
  reps <- unname(unlist(reps))

  all_vars <- colnames(cor_mat)
  vars_to_keep   <- unique(c(reps, all_vars[degree(g) == 0]))  # 孤立点（±1相関なし）は全て保持
  vars_to_remove <- setdiff(all_vars, vars_to_keep)

  # ログ出力：各コンポーネントの代表と除外された変数一覧
  comp_map <- split(comp_df$Variable, comp_df$Component)
  log_rows <- do.call(rbind, lapply(names(comp_map), function(cid) {
    vars <- sort(comp_map[[cid]])
    repv <- intersect(vars, reps)
    removed <- setdiff(vars, repv)
    if (length(repv) == 0) repv <- NA_character_
    data.frame(Component = as.integer(cid),
               Representative = paste(repv, collapse = ";"),
               Removed = ifelse(length(removed) == 0, "", paste(removed, collapse = ";")),
               stringsAsFactors = FALSE)
  }))
  write.csv(log_rows, "removed_perfect_corr_groups_OH.csv", row.names = FALSE)
  write.csv(data.frame(Kept_Variables = vars_to_keep), "kept_vars_OH.csv", row.names = FALSE)

  numData_filtered <- numData_scaled[, vars_to_keep, drop = FALSE]
  cat("✅ 完全相関グループから代表のみを残しました。保持数:", ncol(numData_filtered), "\n")
} else {
  cat("✅ 完全相関（±1）のエッジは見つかりませんでした。\n")
  numData_filtered <- numData_scaled
}

# 前処理後データフレーム
feature_df_proc <- as.data.frame(numData_filtered)

# === 前処理済みデータのCSV出力 ===
out_csv <- "preprocessed_features_OH.csv"
write.csv(feature_df_proc, out_csv, row.names = TRUE)
cat("✅ 前処理済みデータを保存しました:", out_csv, "\n")

# # =========================
# # クラスタ割当の読み込み（無ければ自動生成）
# # =========================
# set.seed(42)

# if (file.exists(cluster_file)) {
#   cluster_df <- read.csv(cluster_file, row.names = 1)
#   colnames(cluster_df) <- "Cluster"
#   cat("✅ 既存のクラスタ割当を読み込みました：", cluster_file, "\n")
# } else {
#   cat("⚠️ クラスタ割当が見つかりません。Ward.D2, k=3 で簡易クラスタを作成します。\n")
#   # 単純補完（列中央値）→ 標準化 → 距離 → 階層クラスタリング
#   feat_imp <- feature_df_proc
#   for (j in seq_len(ncol(feat_imp))) {
#     med_j <- median(feat_imp[[j]], na.rm = TRUE)
#     feat_imp[[j]][is.na(feat_imp[[j]])] <- med_j
#   }
#   x_scaled <- scale(feat_imp)
#   d  <- dist(x_scaled, method = "euclidean")
#   hc <- hclust(d, method = "ward.D2")
#   k <- 3
#   cl <- cutree(hc, k = k)
#   cluster_df <- data.frame(Cluster = cl)
#   write.csv(cluster_df, cluster_file)
#   cat("✅", cluster_file, "を出力しました（k=3）\n")
# }

# # =========================
# # クラスタ平均の可視化
# # =========================
# feature_df_proc$SampleID <- rownames(feature_df_proc)
# cluster_df$SampleID      <- rownames(cluster_df)

# merged_df <- merge(feature_df_proc, cluster_df, by = "SampleID")
# merged_df$Cluster <- as.factor(merged_df$Cluster)

# # クラスタごとの平均
# cluster_mean <- merged_df %>%
#   select(-SampleID) %>%
#   group_by(Cluster) %>%
#   summarise(across(everything(), ~ mean(.x, na.rm = TRUE)))

# # 長い形式へ
# long_cluster_mean <- cluster_mean %>%
#   pivot_longer(cols = -Cluster, names_to = "Feature", values_to = "Mean")

# # 変動が大きい上位特徴量（クラスタ間分散が大きい順）
# top_features <- long_cluster_mean %>%
#   group_by(Feature) %>%
#   summarise(VarAcrossClusters = var(Mean)) %>%
#   arrange(desc(VarAcrossClusters)) %>%
#   slice_head(n = 20) %>%
#   pull(Feature)

# plot_data <- long_cluster_mean %>%
#   filter(Feature %in% top_features)

# # 可視化
# ggplot(plot_data, aes(x = Cluster, y = Mean, group = Feature, color = Feature)) +
#   geom_line() +
#   facet_wrap(~ Feature, scales = "free_y") +
#   theme_minimal() +
#   labs(title = "Cluster-wise Feature Averages (after SD=0 & perfect-corr filtering)",
#        y = "Mean", x = "Cluster")
```
