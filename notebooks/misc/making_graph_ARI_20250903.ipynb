```r
# =============== 必要パッケージ ===============
if (!require(mclust)) install.packages("mclust"); library(mclust)

# =============== 入力ディレクトリ（OH/FP） ===============
base_dir_OH <- "/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/cal_cluster_No/figs_OH"
base_dir_FP <- "/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/cal_cluster_No/figs_FP"

# =============== 出力（同フォルダ・別CSV） ===============
# OH → base_dir_OH/ARI_matrices/
# FP → base_dir_FP/ARI_matrices/
out_subdir <- "ARI_matrices"

# =============== 指標・次元設定 ===============
indices  <- c("silhouette","dunn","gap","ch","db","ptbiserial")
dim_tags <- c("top3","cumeig")   # 両方出力

# =============== ユーティリティ：最新ファイルを探す ===============
find_latest <- function(base_dir, dataset_tag, dim_tag, index) {
  # 例: ClusterAssign_top3_silhouette_OH_*.csv
  pat <- file.path(base_dir, sprintf("ClusterAssign_%s_%s_%s_*.csv", dim_tag, index, dataset_tag))
  candidates <- Sys.glob(pat)
  if (length(candidates) == 0) {
    stop("ファイルが見つかりません: ", pat)
  }
  info <- file.info(candidates)
  latest <- rownames(info)[which.max(info$mtime)]
  return(latest)
}

# =============== ユーティリティ：CSV → (Variable, Cluster) ===============
read_assign <- function(path) {
  df <- read.csv(path, header = TRUE, stringsAsFactors = FALSE, check.names = FALSE)
  if (all(c("Variable","Cluster") %in% names(df))) {
    out <- df[, c("Variable","Cluster")]
  } else {
    rn <- rownames(df)
    num_cols <- names(df)[vapply(df, function(x) is.numeric(x) || all(grepl("^\\d+$", x)), logical(1))]
    chr_cols <- names(df)[vapply(df, function(x) is.character(x) || is.factor(x), logical(1))]
    if (!is.null(rn) && length(rn) == nrow(df) && any(nchar(rn) > 0) && length(num_cols) >= 1) {
      out <- data.frame(Variable = rn, Cluster = df[[num_cols[1]]], stringsAsFactors = FALSE)
    } else if (length(chr_cols) >= 1 && length(num_cols) >= 1) {
      out <- data.frame(Variable = df[[chr_cols[1]]], Cluster = df[[num_cols[1]]], stringsAsFactors = FALSE)
    } else if (ncol(df) >= 2) {
      out <- data.frame(Variable = as.character(df[[1]]), Cluster = df[[2]], stringsAsFactors = FALSE)
    } else {
      stop("クラスタ割当の列を自動認識できませんでした: ", path)
    }
  }
  # クラスタは可能なら整数化
  suppressWarnings({
    cnum <- as.integer(out$Cluster)
    if (!any(is.na(cnum)) || sum(is.na(cnum)) < length(cnum)/2) out$Cluster <- cnum
  })
  out
}

# =============== メイン：データセット × 次元設定ごとに出力 ===============
run_one <- function(base_dir, dataset_tag, dim_tag) {
  # 出力先（同フォルダ内のサブフォルダ）
  outdir <- file.path(base_dir, out_subdir)
  if (!dir.exists(outdir)) dir.create(outdir, recursive = TRUE, showWarnings = FALSE)

  # 全指標の最新ファイルを読み込み
  all_paths <- setNames(nm = indices,
                        object = vapply(indices, function(ix) {
                          find_latest(base_dir, dataset_tag, dim_tag, ix)
                        }, character(1)))
  all_dat <- lapply(all_paths, read_assign)

  # マトリクス計算（行=列=同じ指標集合）
  n <- length(indices)
  rowcol_names <- paste0(indices, "_", dim_tag)
  fmt_mat <- matrix(NA_character_, n, n, dimnames = list(rowcol_names, rowcol_names))
  ari_mat <- matrix(NA_real_,      n, n, dimnames = list(rowcol_names, rowcol_names))

  for (i in seq_along(indices)) {
    for (j in seq_along(indices)) {
      dfi <- all_dat[[i]]
      dfj <- all_dat[[j]]
      merged <- merge(dfi, dfj, by = "Variable", suffixes = c("_i","_j"))
      if (nrow(merged) == 0) next
      ari <- mclust::adjustedRandIndex(merged$Cluster_i, merged$Cluster_j)
      k_i <- length(unique(merged$Cluster_i))
      k_j <- length(unique(merged$Cluster_j))
      fmt_mat[i, j] <- sprintf("%.2f (%d/%d)", ari, k_i, k_j)
      ari_mat[i, j] <- ari
    }
  }

  # 出力ファイル名（同フォルダ・別CSV）
  outfile_fmt <- file.path(outdir, sprintf("ARI_matrix_%s_%s_formatted.csv", dataset_tag, dim_tag))
  outfile_ari <- file.path(outdir, sprintf("ARI_matrix_%s_%s_numeric.csv",   dataset_tag, dim_tag))
  outfile_used <- file.path(outdir, sprintf("files_used_%s_%s.txt",          dataset_tag, dim_tag))

  # CSV保存
  write.csv(as.data.frame(fmt_mat, stringsAsFactors = FALSE), outfile_fmt, row.names = TRUE, fileEncoding = "UTF-8")
  write.csv(as.data.frame(ari_mat), outfile_ari, row.names = TRUE)

  # 使ったファイルの一覧も保存（トレーサビリティ）
  cat(paste0("[", dataset_tag, " / ", dim_tag, "] ファイル対応一覧:\n"), file = outfile_used)
  for (ix in indices) {
    cat(sprintf("%-14s : %s\n", ix, all_paths[[ix]]), file = outfile_used, append = TRUE)
  }

  # コンソール表示
  cat("\n===== ARI マトリクス（", dataset_tag, " / ", dim_tag, "）フォーマット済み =====\n", sep = "")
  print(as.data.frame(fmt_mat, stringsAsFactors = FALSE))
  cat("\n===== ARI マトリクス（", dataset_tag, " / ", dim_tag, "）数値のみ =====\n", sep = "")
  print(round(as.data.frame(ari_mat), 3))

  cat("\n✅ 出力完了（", dataset_tag, " / ", dim_tag, "）:\n  - 表示用CSV: ", outfile_fmt,
      "\n  - 数値CSV:   ", outfile_ari,
      "\n  - 使用ファイル: ", outfile_used, "\n", sep = "")
}

# =============== 実行：OH / FP ×（top3 / cumeig） ===============
for (dim_tag in dim_tags) {
  run_one(base_dir_OH, "OH", dim_tag)
  run_one(base_dir_FP, "FP", dim_tag)
}
```
