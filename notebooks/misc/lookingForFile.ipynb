```python
import os
import json

# 検索対象ディレクトリ
search_dir = '/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/cal_cluster_No/20251026_under_25clusters_program_and_result/for_checking_20251127/'

# 探したい文字列（ファイル名の一部）
target_str = "F_4_4_2_Main_Performance_Grid"

print(f"ディレクトリ: {search_dir}")
print(f"出力ログ(Console Output)から '{target_str}' を検索中...\n")

found_any = False

for file_name in os.listdir(search_dir):
    if file_name.endswith(".ipynb"):
        path = os.path.join(search_dir, file_name)
        try:
            with open(path, 'r', encoding='utf-8') as f:
                nb = json.load(f)
                
                for cell_index, cell in enumerate(nb.get('cells', [])):
                    # コードセルの場合、出力(outputs)を確認
                    if cell.get('cell_type') == 'code':
                        for output in cell.get('outputs', []):
                            
                            # テキスト出力 (print文など) を取得
                            text_content = ""
                            
                            # stream形式 (標準出力)
                            if output.get('output_type') == 'stream':
                                text_content = "".join(output.get('text', []))
                            
                            # execute_result形式 (戻り値表示)
                            elif output.get('output_type') == 'execute_result':
                                text_content = "".join(output.get('data', {}).get('text/plain', []))
                            
                            # 検索ヒット判定
                            if target_str in text_content:
                                print(f"🎯 発見しました！: {file_name}")
                                print(f"   Cell {cell_index + 1} の出力に含まれています。")
                                # 見つかった行周辺を表示
                                lines = text_content.split('\n')
                                for line in lines:
                                    if target_str in line:
                                        print(f"   ログ内容: {line.strip()}")
                                print("-" * 20)
                                found_any = True
                                
        except Exception as e:
            print(f"エラー読み込み不可: {file_name}")

if not found_any:
    print("出力ログの中にも見つかりませんでした。")
```

```python
import os
import json

# 親ディレクトリ（ここから下の sub なども全部探します）
base_dir = '/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/cal_cluster_No/20251026_under_25clusters_program_and_result/for_checking_20251127/'

# 検索キーワード：フォルダ名や生成ロジックに関連するもの
keywords = ["Thesis_Figures", "run_", "strftime", "F_4"]

print(f"ディレクトリ全体（サブフォルダ含む）を検索中: {base_dir}\n")

found_files = []

for root, dirs, files in os.walk(base_dir):
    # .ipynb_checkpoints などの隠しフォルダは除外
    dirs[:] = [d for d in dirs if not d.startswith('.')]
    
    for file in files:
        if file.endswith(".ipynb"):
            path = os.path.join(root, file)
            try:
                with open(path, 'r', encoding='utf-8') as f:
                    nb = json.load(f)
                    
                    # ソースコード全体を1つの文字列にする
                    source_code = ""
                    for cell in nb.get('cells', []):
                        if cell.get('cell_type') == 'code':
                            source_code += "".join(cell.get('source', [])) + "\n"
                    
                    # キーワードチェック
                    hits = [kw for kw in keywords if kw in source_code]
                    if hits:
                        # パスを短く表示するために base_dir を削る
                        rel_path = os.path.relpath(path, base_dir)
                        found_files.append((rel_path, hits))
                        print(f"📄 {rel_path}")
                        print(f"   含まれるキーワード: {hits}")
                        
                        # 文脈を表示
                        lines = source_code.split('\n')
                        for i, line in enumerate(lines):
                            for kw in hits:
                                if kw in line and len(line) < 200:
                                    print(f"   Line {i+1}: {line.strip()}")
                                    break
                        print("-" * 20)
                        
            except Exception as e:
                pass # 読み込みエラーは無視

if not found_files:
    print("キーワードは見つかりませんでした。")
else:
    print(f"\n合計 {len(found_files)} 個のファイルが見つかりました。")
```
