```r
# ==============================================================================
# ステップ 2：変数重要度解析（次元エラー回避 ＋ CSV/PDF出力版）
# ==============================================================================
library(earth)
library(ggplot2)

# 1. パスの設定
dir_path <- "/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/OOD_check/gcvEarth/"
file_name <- "fixed20250702_model_data_gcvEarth_n_all_FP_PCEmax_20251212.rds"
target_path <- paste0(dir_path, file_name)
today <- format(Sys.Date(), "%Y%m%d")

# ロード
bundle <- readRDS(target_path)
actual_model <- bundle$model$finalModel

# 2. 変数重要度（evimp）の取得と変換
ev_matrix <- evimp(actual_model)

# 【重要修正】行列からデータフレームへ安全に変換
# 行列の中身だけを取り出し、行名（変数名）を新しい列として明示的に作成します
df_ev <- data.frame(
  variable = rownames(ev_matrix),
  matrix(as.numeric(ev_matrix), nrow = nrow(ev_matrix), dimnames = dimnames(ev_matrix)),
  stringsAsFactors = FALSE
)

# 3. 列名の正規化（gcvカラムを確実に特定）
# evimpの結果には "gcv" というカラムが含まれるため、これを使用します
main_metric <- "gcv"

# CSV保存
csv_ev_file <- paste0("step2_variable_importance_fixed_", today, ".csv")
write.csv(df_ev, csv_ev_file, row.names = FALSE)

# 4. グラフ作成（上位15変数）
# 重要度（gcv）の降順にソート
df_plot <- df_ev[order(df_ev[[main_metric]], decreasing = TRUE), ]
df_plot <- head(df_plot, 15)

# グラフ描画
p2 <- ggplot(df_plot, aes(x = reorder(variable, get(main_metric)), y = get(main_metric))) +
  geom_bar(stat = "identity", fill = "steelblue") +
  coord_flip() +
  theme_minimal(base_size = 14) +
  labs(title = "Feature Importance: fixed_n_all_FP_PCEmax",
       subtitle = paste("Based on", main_metric, "score"),
       x = "Physicochemical / Structural Descriptors",
       y = "Importance Score (GCV)")

# PDF保存
pdf_ev_file <- paste0("step2_variable_importance_plot_", today, ".pdf")
ggsave(pdf_ev_file, p2, width = 8, height = 6)

# --- コンソールへの最終レポート ---
cat("====================================================\n")
cat("  ステップ 2：解析完了レポート\n")
cat("====================================================\n")
cat("最重要変数 (Top 5):\n")
print(head(df_plot[, c("variable", "gcv")], 5))

cat("\n--- 出力ファイル一覧 ---\n")
cat("1. 重要度CSV: ", csv_ev_file, "\n")
cat("2. グラフPDF: ", pdf_ev_file, "\n")
cat("====================================================\n")
```
