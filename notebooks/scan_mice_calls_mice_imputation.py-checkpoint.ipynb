```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

"""
scan_mice_calls_mice_imputation.py
----------------------------------
R の履歴ファイル (例: RStudio_rireki.txt) から:

- mice() 呼び出しをすべて検出
- 各呼び出しについて:
    - 補完前データ名 (mice()の第1引数とその変数の定義)
    - 補完方法 method= の中身（変数なら定義を追跡）
    - 補完後オブジェクト (mdata <- mice())
    - complete(mdata, ...) による最終データ名

を抽出し、時系列順に CSV に出力する。
"""

import sys
import re
import csv
from pathlib import Path
from typing import List, Dict, Optional, Tuple

IDENT_CHARS = set("abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789_.")


def load_text(path: Path) -> str:
    return path.read_text(encoding="utf-8", errors="ignore")


def find_mice_calls(text: str) -> List[Dict]:
    """
    Find all 'mice(...)' calls in the text and return a list of dicts
    with basic info (positions, raw argument string, assigned variable name).
    """
    results = []
    i = 0
    n = len(text)

    while True:
        idx = text.find("mice(", i)
        if idx == -1:
            break

        # ensure 'mice' is a standalone identifier
        if idx > 0 and text[idx - 1] in IDENT_CHARS:
            i = idx + 4
            continue

        # find matching ')'
        start_args = idx + len("mice(")
        depth = 1
        j = start_args
        in_str: Optional[str] = None

        while j < n and depth > 0:
            ch = text[j]
            if in_str is not None:
                if ch == "\\":
                    j += 2
                    continue
                if ch == in_str:
                    in_str = None
            else:
                if ch in ('"', "'"):
                    in_str = ch
                elif ch == "(":
                    depth += 1
                elif ch == ")":
                    depth -= 1
                    if depth == 0:
                        break
            j += 1

        if depth != 0:
            break  # unmatched

        args_str = text[start_args:j]
        lhs = find_lhs_variable(text, idx)

        results.append(
            {
                "start_index": idx,
                "end_index": j + 1,
                "args_str": args_str,
                "assigned_var": lhs,
            }
        )
        i = j + 1

    return results


def find_lhs_variable(text: str, func_pos: int) -> Optional[str]:
    """
    Look backwards from 'func_pos' to see if there is an assignment "<-" or "="
    like: mdata <- mice(...)
    """
    start = max(0, func_pos - 200)
    snippet = text[start:func_pos]

    patterns = [
        r"([A-Za-z_][A-Za-z0-9_.]*)\s*<-\s*$",
        r"([A-Za-z_][A-Za-z0-9_.]*)\s*=\s*$",
    ]

    for pat in patterns:
        matches = list(re.finditer(pat, snippet, re.MULTILINE))
        if matches:
            return matches[-1].group(1)

    return None


def split_args_top_level(args_str: str) -> List[str]:
    """
    Split "inside of mice(...)" into arguments, ignoring commas in nested
    parentheses/brackets/braces and strings.
    """
    args = []
    current = []
    depth_paren = 0
    depth_brack = 0
    depth_brace = 0
    in_str: Optional[str] = None

    i = 0
    n = len(args_str)

    while i < n:
        ch = args_str[i]

        if in_str is not None:
            current.append(ch)
            if ch == "\\" and i + 1 < n:
                current.append(args_str[i + 1])
                i += 2
                continue
            elif ch == in_str:
                in_str = None
        else:
            if ch in ('"', "'"):
                in_str = ch
                current.append(ch)
            elif ch == "(":
                depth_paren += 1
                current.append(ch)
            elif ch == ")":
                depth_paren -= 1
                current.append(ch)
            elif ch == "[":
                depth_brack += 1
                current.append(ch)
            elif ch == "]":
                depth_brack -= 1
                current.append(ch)
            elif ch == "{":
                depth_brace += 1
                current.append(ch)
            elif ch == "}":
                depth_brace -= 1
                current.append(ch)
            elif (
                ch == ","
                and depth_paren == 0
                and depth_brack == 0
                and depth_brace == 0
            ):
                arg = "".join(current).strip()
                if arg:
                    args.append(arg)
                current = []
            else:
                current.append(ch)
        i += 1

    last = "".join(current).strip()
    if last:
        args.append(last)

    return args


def parse_mice_args(args_str: str) -> Dict:
    arg_list = split_args_top_level(args_str)
    res = {
        "first_arg_expr": arg_list[0].strip() if arg_list else None,
        "method_expr": None,
    }
    for arg in arg_list:
        parts = arg.split("=", 1)
        if len(parts) != 2:
            continue
        name = parts[0].strip()
        value = parts[1].strip()
        if name == "method":
            res["method_expr"] = value
    return res


def extract_var_name_from_expr(expr: Optional[str]) -> Optional[str]:
    if expr is None:
        return None
    tokens = re.split(r"[^A-Za-z0-9_.]+", expr)
    tokens = [t for t in tokens if t]
    return tokens[-1] if tokens else None


def find_last_assignment_before(text: str, varname: str, pos: int) -> Optional[Tuple[str, int]]:
    """
    Search in text[:pos] for the last "varname <- ..." line.
    """
    sub = text[:pos]
    lines = sub.splitlines()
    pattern = re.compile(rf"^\s*{re.escape(varname)}\s*<-\s*(.+)$")
    last_line = None
    last_lineno = None

    for lineno, line in enumerate(lines, start=1):
        m = pattern.match(line)
        if m:
            last_line = line
            last_lineno = lineno

    if last_line is None:
        return None
    return last_line, last_lineno


def extract_file_from_assignment(line: str) -> Optional[str]:
    """
    From e.g.:
        data <- read.csv("path/to/file.csv")
    extract "path/to/file.csv"
    """
    m = re.search(r'["\']([^"\']+)["\']', line)
    if m:
        return m.group(1)
    return None


def is_plain_identifier(expr: Optional[str]) -> bool:
    if expr is None:
        return False
    expr_stripped = expr.strip()
    return re.fullmatch(r"[A-Za-z_][A-Za-z0-9_.]*", expr_stripped) is not None


def find_complete_outputs(text: str, mdata_var: str, start_pos: int) -> List[Tuple[str, int]]:
    """
    After position start_pos, find lines like:
        comp <- complete(mdata_var, ...)
    """
    sub = text[start_pos:]
    lines = sub.splitlines()
    results = []
    before_lines = text[:start_pos].count("\n")

    pattern = re.compile(
        rf"^\s*([A-Za-z_][A-Za-z0-9_.]*)\s*<-\s*complete\(\s*{re.escape(mdata_var)}\b"
    )

    for i, line in enumerate(lines, start=1):
        m = pattern.match(line)
        if m:
            lhs = m.group(1)
            results.append((lhs, before_lines + i))

    return results


def main():
    if len(sys.argv) != 2:
        print("Usage: python scan_mice_calls_mice_imputation.py <RStudio_rireki.txt>", file=sys.stderr)
        sys.exit(1)

    txt_path = Path(sys.argv[1]).expanduser().resolve()
    if not txt_path.exists():
        print(f"[ERROR] Input text file not found: {txt_path}", file=sys.stderr)
        sys.exit(1)

    text = load_text(txt_path)
    calls = find_mice_calls(text)

    if not calls:
        print("[INFO] No mice() calls found.")
        return

    print(f"[INFO] Found {len(calls)} mice() call(s) in {txt_path.name}\n")

    rows = []

    for idx, call in enumerate(calls, start=1):
        args_info = parse_mice_args(call["args_str"])
        first_arg_expr = args_info["first_arg_expr"]
        method_expr = args_info["method_expr"]

        imput_input_var = extract_var_name_from_expr(first_arg_expr)
        imput_result_var = call["assigned_var"]

        # input var assignment
        input_assign_line = ""
        input_file_path = ""
        if imput_input_var:
            assign_info = find_last_assignment_before(text, imput_input_var, call["start_index"])
            if assign_info:
                line, _lineno = assign_info
                input_assign_line = line
                fp = extract_file_from_assignment(line)
                if fp:
                    input_file_path = fp

        # method definition
        method_is_var = False
        method_assign_line = ""
        if method_expr:
            if is_plain_identifier(method_expr):
                method_is_var = True
                m_assign = find_last_assignment_before(text, method_expr.strip(), call["start_index"])
                if m_assign:
                    m_line, _m_lineno = m_assign
                    method_assign_line = m_line

        # complete(mdata, ...) outputs
        completed_names: List[str] = []
        if imput_result_var:
            completes = find_complete_outputs(text, imput_result_var, call["end_index"])
            for lhs, _lineno in completes:
                completed_names.append(lhs)
        completed_str = ";".join(completed_names) if completed_names else ""

        rows.append(
            {
                "call_index": idx,
                "raw_args": call["args_str"].strip(),
                "imputed_object_var": imput_result_var or "",
                "input_first_arg_expr": first_arg_expr or "",
                "input_var_name": imput_input_var or "",
                "input_last_assignment_line": input_assign_line,
                "input_file_path": input_file_path,
                "method_expr": method_expr or "",
                "method_is_variable": str(method_is_var),
                "method_last_assignment_line": method_assign_line,
                "completed_data_vars": completed_str,
            }
        )

    # CSV 出力
    out_csv = txt_path.with_name("mice_imputation_summary.csv")
    fieldnames = [
        "call_index",
        "raw_args",
        "imputed_object_var",
        "input_first_arg_expr",
        "input_var_name",
        "input_last_assignment_line",
        "input_file_path",
        "method_expr",
        "method_is_variable",
        "method_last_assignment_line",
        "completed_data_vars",
    ]
    with out_csv.open("w", encoding="utf-8", newline="") as f:
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(rows)

    print(f"[INFO] Summary CSV written to: {out_csv}")


if __name__ == "__main__":
    main()
```
