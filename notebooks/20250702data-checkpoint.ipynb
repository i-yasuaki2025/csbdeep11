```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import re

# --- 入力ファイルパス ---
vip_file = "/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/graph/PLS/PLS_csvdata_20250702/new202506_vip_m_all_FF_PLS_20250626.csv"
definition_file = "/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/graph/MACCS166_definitions.csv"

# --- 出力ファイルパス ---
output_dir = os.path.dirname(vip_file)
basename = os.path.splitext(os.path.basename(vip_file))[0]
output_csv = os.path.join(output_dir, f"{basename}_renamed.csv")
output_png = os.path.join(output_dir, f"{basename}_top20.png")

# --- データの読み込み ---
vip_df = pd.read_csv(vip_file)
definition_df = pd.read_csv(definition_file)

# --- 定義の整形 ---
def clean_definition(text):
    text = re.sub(r'[\s\(\)\-]', '', str(text))  # スペース・( )・- を削除
    return text

definition_df["Key"] = definition_df["Key"].astype(str).apply(lambda x: f"X{x.strip()}")
definition_df["Definition"] = definition_df["Definition"].apply(clean_definition)

# --- 辞書の作成 ---
conversion_dict = dict(zip(definition_df["Key"], definition_df["Definition"]))

# --- VIP変数名変換（X130.4 → QQ>1&..._4）---
def convert_variable_name(var):
    match = re.match(r"(X\d+)\.(\d+)", var)
    if match:
        key, suffix = match.groups()
        base = conversion_dict.get(key, key)
        return f"{base}_{suffix}"
    return var  # フォーマット外はそのまま

vip_df["Variable"] = vip_df["Variable"].apply(convert_variable_name)

# --- CSV保存 ---
vip_df.to_csv(output_csv, index=False)

# --- 図の作成 ---
df_top20 = vip_df.sort_values(by="VIP", ascending=False).head(20)
plt.figure(figsize=(10, 6))
plt.barh(df_top20["Variable"], df_top20["VIP"], edgecolor='black')
plt.xlabel("VIP Score")
plt.title(f"Top 20 VIP Scores - {basename}")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig(output_png, dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# --- 設定 ---
data_parts = [
    "n_base", "n_base_OH", "n_base_FP",
    "n_all", "n_all_OH", "n_all_FP",
    "m_base", "m_base_OH", "m_base_FP",
    "m_all", "m_all_OH", "m_all_FP"
]

target_parts = ["PCEmax", "Jsc", "Voc", "FF"]

base_dirs = {
    "PLS": "/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/graph/PLS/PLS_csvdata_20250702",
    "gcvEarth": "/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/graph/gcvEarth/gcvEarth_csvdata_20250702",
    "Rborist": "/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/graph/Rborist/Rborist_csvdata_20250702"
}

# --- MACCS変換辞書 ---
def make_maccs_dict(csv_path):
    df = pd.read_csv(csv_path)
    maccs_dict = {}
    for _, row in df.iterrows():
        key = f"X{row['Key']}"
        cleaned = ''.join(c for c in str(row['Definition']) if c.isalnum())
        maccs_dict[key] = cleaned
    return maccs_dict

maccs_def_path = "/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/graph/MACCS166_definitions.csv"
maccs_dict = make_maccs_dict(maccs_def_path)

# --- ラベル重複解消関数 ---
def deduplicate_labels(labels):
    seen = {}
    new_labels = []
    for label in labels:
        if label not in seen:
            seen[label] = 0
            new_labels.append(label)
        else:
            seen[label] += 1
            new_labels.append(f"{label}.{seen[label]}")
    return new_labels

# --- ループ開始 ---
for data_part in data_parts:
    for target in target_parts:
        file_suffix = f"{data_part}_{target}"
        filenames = {
            "PLS": f"new20250702_vip_{file_suffix}_PLS_20250702.csv",
            "gcvEarth": f"new20250702_varimp_gcvEarth_{file_suffix}_20250702.csv",
        }

        data = {}
        for model, base_dir in base_dirs.items():
            path = os.path.join(base_dir, filenames[model])
            print(f"🔍 チェック中: {model} → {path}")

            if not os.path.exists(path):
                print(f"⚠️ {model} のファイルが見つかりません: {path}")
                continue  # 次のモデルへ

            try:
                df = pd.read_csv(path)
                df.columns = [col.strip() for col in df.columns]

                if model == "Rborist":
                    df.columns = ["Variable", "Score"]
                elif "VIP" in df.columns:
                    df = df.rename(columns={"VIP": "Score"})
                elif "Overall" in df.columns:
                    df = df.rename(columns={"Overall": "Score"})

                var_col = next((col for col in df.columns if col in ["Variable", "Feature", "X", "Var", "Name", "Unnamed: 0"]), None)
                if var_col is None:
                    raise ValueError(f"{model} のファイルで変数名列が見つかりません: {df.columns.tolist()}")
                df = df.rename(columns={var_col: "Variable"})

                df_top = df.sort_values("Score", ascending=False).head(10)
                df_top["Score"] = df_top["Score"] / df_top["Score"].max()
                df_top["Label"] = df_top["Variable"].apply(lambda x: maccs_dict.get(str(x).split(".")[0], str(x)))
                df_top["Label"] = deduplicate_labels(df_top["Label"].tolist())

                data[model] = df_top.set_index("Label")["Score"]

            except Exception as e:
                print(f"❌ {model} の処理エラー: {e}")
                continue

        # モデルが1つ以上読み込めていれば描画
        if data:
            try:
                merged_df = pd.concat(data.values(), axis=1, keys=data.keys()).fillna(0)
                ax = merged_df.plot(kind='barh', figsize=(12, 8))
                plt.xlabel("Normalized Importance Score")
                plt.ylabel("Variable (MACCS label)")
                plt.title(f"Top 10 Variables Comparison ({target} - {data_part})")
                plt.tight_layout()
                plt.gca().invert_yaxis()
                plt.legend(title="Model")

                output_file = f"importance_comparison_{data_part}_{target}.png"
                plt.savefig(output_file)
                plt.close()
                print(f"✅ グラフ保存完了: {output_file}")
            except Exception as e:
                print(f"❌ グラフ生成エラー: {e}")
        else:
            print(f"⚠️ 有効なモデルがないためスキップ: {file_suffix}")

```
