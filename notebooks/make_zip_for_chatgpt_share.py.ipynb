```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

"""
make_zip_for_chatgpt_share.py
-----------------------------
FP→OH 解析結果（reverse/analysis_csv）を ChatGPT 共有用に ZIP 圧縮するスクリプト。

出力例：
  /Volumes/csbdeep11/_yasu-i/20250303_rebuiled/cal_cluster_No/
      paper_4.2.4.4_OHFP_reverse_analysis_csv_20251022.zip
"""

from pathlib import Path
import zipfile
import datetime

# ========= 設定 =========
ROOT = Path("/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/cal_cluster_No")
TARGET_DIR = ROOT / "paper_4.2.4.4_OHFP_20251022" / "reverse" / "analysis_csv"

# 出力ZIPファイル名（日時入り）
date_str = datetime.datetime.now().strftime("%Y%m%d")
OUT_ZIP = ROOT / f"paper_4.2.4.4_OHFP_reverse_analysis_csv_{date_str}.zip"

# ========= 実行 =========
if not TARGET_DIR.exists():
    raise FileNotFoundError(f"解析結果フォルダが見つかりません: {TARGET_DIR}")

with zipfile.ZipFile(OUT_ZIP, "w", compression=zipfile.ZIP_DEFLATED) as zf:
    for path in TARGET_DIR.rglob("*"):
        if path.is_file():
            # 相対パスで保存（サブフォルダ構造を保つ）
            arcname = path.relative_to(ROOT)
            zf.write(path, arcname)
            print(f"[ADD] {arcname}")

print(f"\n✅ ZIP作成完了: {OUT_ZIP}")
```
