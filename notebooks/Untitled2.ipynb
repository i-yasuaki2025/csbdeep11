```python
import os
import pandas as pd
import numpy as np

# =============================================================================
# 1. 設定：パスと対象の定義
# =============================================================================
RAW_DATA_PATH = "/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/20220127_PLS 202201 ALL/20251216_for_making_collect_data_nm_baseall_OHFP_csv/rebuilt_merged_data/"
FI_MASTER_FILE = "Feature_Importance_All_Models_Master_20260112.csv"

target_configs = [
    {"Target": "PCEmax", "Type": "補完",   "File": "m_all_FP_rebuilt.csv"},
    {"Target": "PCEmax", "Type": "未補完", "File": "n_base_OH_rebuilt.csv"},
    {"Target": "Jsc",    "Type": "補完",   "File": "m_all_FP_rebuilt.csv"},
    {"Target": "Jsc",    "Type": "未補完", "File": "n_base_OH_rebuilt.csv"},
    {"Target": "Voc",    "Type": "補完",   "File": "m_base_FP_rebuilt.csv"},
    {"Target": "Voc",    "Type": "未補完", "File": "n_base_FP_rebuilt.csv"},
    {"Target": "FF",     "Type": "補完",   "File": "m_all_OH_rebuilt.csv"},
    {"Target": "FF",     "Type": "未補完", "File": "n_base_OH_rebuilt.csv"}
]

# =============================================================================
# 2. 解析処理（規格化・ソフトバイナリ対応版）
# =============================================================================
fi_master = pd.read_csv(FI_MASTER_FILE)
impact_results = []

print(f"{'='*100}")
print(f" Feature Impact Analysis: Robust Trend Investigation (Handles Soft-Binary & Normalized Data)")
print(f"{'='*100}")

for cfg in target_configs:
    tgt, imp_type, raw_file = cfg['Target'], cfg['Type'], cfg['File']
    
    top_15 = fi_master[(fi_master['Target'] == tgt) & 
                       (fi_master['Imputation_Type'] == imp_type)].sort_values('Normalized_Importance', ascending=False).head(15)
    
    data_path = os.path.join(RAW_DATA_PATH, raw_file)
    if not os.path.exists(data_path) or top_15.empty:
        continue
    
    df_raw = pd.read_csv(data_path)
    print(f"\nProcessing: {tgt} ({imp_type}) | Source: {raw_file}")

    for _, row in top_15.iterrows():
        feature = row['Feature']
        if feature not in df_raw.columns: continue
            
        # 欠損値を除去してペアを作る
        valid_idx = df_raw[[feature, tgt]].dropna().index
        feat_vals = df_raw.loc[valid_idx, feature]
        target_vals = df_raw.loc[valid_idx, tgt]

        # ---------------------------------------------------------
        # 影響度の計算ロジック（改修ポイント）
        # ---------------------------------------------------------
        # 1. 変数が「厳密なバイナリ(0 or 1)」かどうかを判定
        is_strictly_binary = set(feat_vals.unique()).issubset({0, 1})

        if is_strictly_binary:
            # 厳密な0/1変数の場合は平均値の差
            m1 = target_vals[feat_vals == 1].mean()
            m0 = target_vals[feat_vals == 0].mean()
            impact_val = m1 - m0
            method = "Mean Diff (1 vs 0)"
        else:
            # 補完されたバイナリ(0.99等)や連続値、規格化済み変数はすべて「相関係数」で評価
            # これによりスケール(標準化等)の影響を排除し、純粋な「向き」を取り出せる
            impact_val = feat_vals.corr(target_vals)
            method = "Pearson Correlation"

        # 方向性の判定（数値が極めて小さい場合は誤差として扱う）
        if pd.isna(impact_val):
            direction = "No Variation / Constant"
        elif impact_val > 0.0001:
            direction = "Positive (+)"
        elif impact_val < -0.0001:
            direction = "Negative (-)"
        else:
            direction = "Neutral / Stable"
            
        impact_results.append({
            "Target": tgt, "Imputation": imp_type, "Feature": feature,
            "FI_Rank": len([x for x in impact_results if x['Target']==tgt and x['Imputation']==imp_type]) + 1,
            "Importance": row['Normalized_Importance'],
            "Impact_Value": round(impact_val, 4) if pd.notna(impact_val) else np.nan,
            "Method": method, "Trend": direction
        })
        print(f"  [{direction:12}] {feature:25} (Val: {impact_val:8.4f} | {method})")

# 結果保存
pd.DataFrame(impact_results).to_csv("Feature_Impact_Trends_Robust_Top15.csv", index=False, encoding='utf-8-sig')
print(f"\n{'='*100}\n SUCCESS: Results saved to 'Feature_Impact_Trends_Robust_Top15.csv'\n{'='*100}")
```

```python
import os
import pandas as pd

# ==========================================
# 1. 設定
# ==========================================
RAW_DATA_DIR = "/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/20220127_PLS 202201 ALL/20251216_for_making_collect_data_nm_baseall_OHFP_csv/rebuilt_merged_data/"
MASTER_CSV = "Feature_Importance_All_Models_Master_20260112.csv"

# ==========================================
# 2. 解析対象の抽出（重要度マスタから）
# ==========================================
master_df = pd.read_csv(MASTER_CSV)

# PCEmaxの未補完モデルでの上位材料（OH）
top_materials = master_df[
    (master_df['Target'] == 'PCEmax') & 
    (master_df['Imputation_Type'] == '未補完') & 
    (master_df['Feature'].str.startswith('SMILESsnamep1M'))
].sort_values('Normalized_Importance', ascending=False).head(10)['Feature'].tolist()

# PCEmaxの補完モデルでの上位構造ビット（FP）
top_fp_bits = master_df[
    (master_df['Target'] == 'PCEmax') & 
    (master_df['Imputation_Type'] == '補完') & 
    (master_df['Feature'].str.startswith('X'))
].sort_values('Normalized_Importance', ascending=False).head(15)['Feature'].tolist()

# ==========================================
# 3. データの読み込みとサンプル番号の特定
# ==========================================
df_m_oh = pd.read_csv(os.path.join(RAW_DATA_DIR, "m_all_OH_rebuilt.csv"))
df_m_fp = pd.read_csv(os.path.join(RAW_DATA_DIR, "m_all_FP_rebuilt.csv"))

# １列目のカラム名を特定（サンプル番号/IDとして使用）
id_col_oh = df_m_oh.columns[0]
id_col_fp = df_m_fp.columns[0]
print(f"Matching using ID columns: [OH: {id_col_oh}] and [FP: {id_col_fp}]")

cross_analysis = []

for mat in top_materials:
    # 1. OH版ファイルから、対象材料が含まれるサンプルの「ID番号」をリストアップ
    if mat not in df_m_oh.columns:
        continue
    
    # 値が0.5以上（補完後のソフトバイナリ対応）の行のIDを取得
    target_sample_ids = df_m_oh[df_m_oh[mat] > 0.5][id_col_oh].tolist()
    
    if not target_sample_ids:
        continue
        
    # 2. FP版ファイルから、上記IDに一致する行を抽出
    samples_fp = df_m_fp[df_m_fp[id_col_fp].isin(target_sample_ids)]
    
    if samples_fp.empty:
        continue
        
    # 各構造ビット（FP）の平均含有率を計算
    existing_bits = [b for b in top_fp_bits if b in samples_fp.columns]
    presence_stats = samples_fp[existing_bits].mean().to_dict()
    
    # メタ情報付与
    presence_stats['Material_Name'] = mat.replace('SMILESsnamep1M_', '')
    presence_stats['Sample_Count'] = len(samples_fp)
    
    cross_analysis.append(presence_stats)

# ==========================================
# 4. 結果の出力
# ==========================================
if cross_analysis:
    result_df = pd.DataFrame(cross_analysis).set_index('Material_Name')
    
    # カラムの順序を整理（ビットの重要度順にする）
    cols = [b for b in top_fp_bits if b in result_df.columns] + ['Sample_Count']
    result_df = result_df[cols]
    
    result_df.to_csv("PCEmax_CrossCheck_By_SampleID.csv", encoding='utf-8-sig')
    print("\n--- PCEmax Cross-Check Result (ID Matched) ---")
    print(result_df)
else:
    print("\n[ERROR] サンプル番号による照合に失敗しました。１列目の内容が一致しているか確認してください。")
```

```python
import os
import pandas as pd

# ==========================================
# 1. 設定
# ==========================================
base_path = "/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/20220127_PLS 202201 ALL/20251216_for_making_collect_data_nm_baseall_OHFP_csv/rebuilt_merged_data/"

file_names = [
    "n_base.csv", "n_base_OH_rebuilt.csv", "n_base_FP_rebuilt.csv",
    "n_all.csv",  "n_all_OH_rebuilt.csv",  "n_all_FP_rebuilt.csv",
    "m_base.csv", "m_base_OH_rebuilt.csv", "m_base_FP_rebuilt.csv",
    "m_all.csv",  "m_all_OH_rebuilt.csv",  "m_all_FP_rebuilt.csv"
]

prefix = "SMILESsnamep1M_"

# 結果格納用
list_results = []
summary_results = []

# ==========================================
# 2. 処理実行
# ==========================================
for f in file_names:
    full_path = os.path.join(base_path, f)
    
    if os.path.exists(full_path):
        try:
            # データの読み込み
            df = pd.read_csv(full_path)
            
            # 対象となる変数を特定
            target_cols = [c for c in df.columns if c.startswith(prefix)]
            
            # 処理前 (Before)
            before_materials = [c.replace(prefix, "") for c in target_cols]
            
            # 処理後 (After: 全ての値が同じものを除外)
            # nunique() > 1 のものだけを残す
            active_cols = [c for c in target_cols if df[c].nunique() > 1]
            after_materials = [c.replace(prefix, "") for c in active_cols]
            
            # --- 1つ目のCSV用データの作成 (Material List) ---
            # 全材料をマージして Before/After のフラグを立てる
            all_mats_in_file = sorted(list(set(before_materials)))
            for mat in all_mats_in_file:
                list_results.append({
                    "File": f,
                    "Material": mat,
                    "Status_Before": "Yes",
                    "Status_After": "Yes" if mat in after_materials else "No (Excluded)"
                })
            
            # --- 2つ目のCSV用データの作成 (Counts Summary) ---
            summary_results.append({
                "File": f,
                "Count_Before": len(before_materials),
                "Count_After": len(after_materials),
                "Excluded_Count": len(before_materials) - len(after_materials)
            })
            
            print(f"Success: {f} (Before: {len(before_materials)}, After: {len(after_materials)})")
            
        except Exception as e:
            print(f"Error processing {f}: {e}")
    else:
        print(f"Skipped: {f} (File not found)")

# ==========================================
# 3. CSV出力
# ==========================================
# 1. 材料名詳細リスト
if list_results:
    df_list = pd.DataFrame(list_results)
    df_list.to_csv("Detailed_Material_List_Processing.csv", index=False, encoding='utf-8-sig')
    print("\nGenerated: Detailed_Material_List_Processing.csv")

# 2. 種類数サマリー
if summary_results:
    df_summary = pd.DataFrame(summary_results)
    df_summary.to_csv("Material_Counts_Summary.csv", index=False, encoding='utf-8-sig')
    print("Generated: Material_Counts_Summary.csv")
```

```python
import os
import pandas as pd
import shutil
import re

# ==========================================
# 1. パスと設定
# ==========================================
base_path = "/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/20220127_PLS 202201 ALL/20251216_for_making_collect_data_nm_baseall_OHFP_csv/rebuilt_merged_data/"

file_names = [
    "n_base.csv", "n_base_OH_rebuilt.csv", "n_base_FP_rebuilt.csv",
    "n_all.csv",  "n_all_OH_rebuilt.csv",  "n_all_FP_rebuilt.csv",
    "m_base.csv", "m_base_OH_rebuilt.csv", "m_base_FP_rebuilt.csv",
    "m_all.csv",  "m_all_OH_rebuilt.csv",  "m_all_FP_rebuilt.csv"
]

prefix_name = "SMILESsnamep1M_"
# ★ X(数字).4 のパターンを定義
pattern_struct = r'^X\d+\.4$' 

output_dir = "Processed_Full_Summary_Final"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

summary_rows = []

# ヘッダー表示（コンソール）: 構造記述子の列を追加
print(f"{'File Name':<22} | {'Samples':<11} | {'p1M (B->A)':<11} | {'Struct.4 (B->A)':<15} | {'Cols excl.ID'}")
print("-" * 105)

# ==========================================
# 2. 処理実行
# ==========================================
for f in file_names:
    full_path = os.path.join(base_path, f)
    if not os.path.exists(full_path):
        continue

    try:
        df = pd.read_csv(full_path)
        
        # --- 処理前カウント ---
        rows_before = len(df)
        all_cols_excl_id = df.columns[1:]
        
        # 材料名(p1M)の特定
        p1m_names_before = [c for c in all_cols_excl_id if c.startswith(prefix_name)]
        # ★ 構造記述子(X.4)の特定
        struct_cols_before = [c for c in all_cols_excl_id if re.match(pattern_struct, c)]
        
        count_p1m_before = len(p1m_names_before)
        count_struct_before = len(struct_cols_before)
        count_total_before = len(all_cols_excl_id)

        # --- 処理：定数カラムの削除 ＆ 欠損行の削除 ---
        id_col = df.columns[0]
        
        # 1. 定数カラムの特定（nunique > 1 のものだけ残す）
        active_cols = [c for c in all_cols_excl_id if df[c].nunique() > 1]
        df_step1 = df[[id_col] + active_cols]
        
        # 2. 欠損値を含む行を削除
        df_cleaned = df_step1.dropna()
        
        # --- 処理後カウント ---
        rows_after = len(df_cleaned)
        active_p1m_names = [c for c in active_cols if c.startswith(prefix_name)]
        # ★ 処理後の構造記述子カウント
        active_struct_cols = [c for c in active_cols if re.match(pattern_struct, c)]
        
        count_p1m_after = len(active_p1m_names)
        count_struct_after = len(active_struct_cols)
        count_total_after = len(active_cols)

        # --- コンソール表示 ---
        sample_disp = f"{rows_before:>4}->{rows_after:<4}"
        p1m_disp = f"{count_p1m_before:>3}->{count_p1m_after:<3}"
        struct_disp = f"{count_struct_before:>3}->{count_struct_after:<3}"
        total_disp = f"{count_total_before:>4}->{count_total_after:<4}"
        
        print(f"{f:<22} | {sample_disp:<11} | {p1m_disp:<11} | {struct_disp:<15} | {total_disp}")

        # サマリーデータ蓄積
        summary_rows.append({
            "File": f,
            "Samples_Before": rows_before,
            "Samples_After": rows_after,
            "p1M_Names_Before": count_p1m_before,
            "p1M_Names_After": count_p1m_after,
            "Struct.4_Before": count_struct_before,
            "Struct.4_After": count_struct_after,
            "Total_Cols_Before(excl_ID)": count_total_before,
            "Total_Cols_After(excl_ID)": count_total_after
        })

        # クリーニング済みデータの保存
        df_cleaned.to_csv(os.path.join(output_dir, f"cleaned_{f}"), index=False, encoding='utf-8-sig')

    except Exception as e:
        print(f"Error processing {f}: {e}")

# ==========================================
# 3. 保存と圧縮
# ==========================================
if summary_rows:
    summary_df = pd.DataFrame(summary_rows)
    summary_df.to_csv(os.path.join(output_dir, "Full_Processing_Summary.csv"), index=False, encoding='utf-8-sig')

zip_name = "Final_Data_and_Summary_Package"
shutil.make_archive(zip_name, 'zip', output_dir)

print("-" * 105)
print(f"\n完了: 結果を '{output_dir}' に保存しました。")
print(f"全データ同梱ZIP: {zip_name}.zip")
```

```python
import os
import pandas as pd
import shutil
import re

# ==========================================
# 1. 設定
# ==========================================
base_path = "/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/20220127_PLS 202201 ALL/20251216_for_making_collect_data_nm_baseall_OHFP_csv/rebuilt_merged_data/"
file_names = [
    "n_base.csv", "n_base_OH_rebuilt.csv", "n_base_FP_rebuilt.csv",
    "n_all.csv",  "n_all_OH_rebuilt.csv",  "n_all_FP_rebuilt.csv",
    "m_base.csv", "m_base_OH_rebuilt.csv", "m_base_FP_rebuilt.csv",
    "m_all.csv",  "m_all_OH_rebuilt.csv",  "m_all_FP_rebuilt.csv"
]

output_dir = "Dataset_Summary_Report_Final"
os.makedirs(output_dir, exist_ok=True)

report_rows = []

# ==========================================
# 2. 処理と統計収集
# ==========================================
for f in file_names:
    full_path = os.path.join(base_path, f)
    if not os.path.exists(full_path): continue

    try:
        df = pd.read_csv(full_path)
        id_col = df.columns[0]
        all_features = df.columns[1:].tolist()

        # 処理前の統計
        n_samples_pre = len(df)
        n_features_pre = len(all_features)

        # クリーニング (sd=0の削除)
        # nunique > 1 のカラムを抽出
        active_features = [c for c in all_features if df[c].nunique() > 1]
        dropped_features = list(set(all_features) - set(active_features))
        
        # 欠損値削除
        df_cleaned = df[[id_col] + active_features].dropna()
        n_samples_post = len(df_cleaned)
        n_features_post = len(active_features)

        # サマリー情報の構築
        report_rows.append({
            "Dataset_Name": f,
            "Samples_Raw": n_samples_pre,
            "Samples_Cleaned": n_samples_post,
            "Sample_Loss_Rate(%)": round((1 - n_samples_post/n_samples_pre)*100, 2),
            "Features_Raw": n_features_pre,
            "Features_Active(sd>0)": n_features_post,
            "Features_Dropped(sd=0)": len(dropped_features),
            "Feature_Retention_Rate(%)": round((n_features_post/n_features_pre)*100, 2),
            "Dropped_Feature_Names": ", ".join(dropped_features[:5]) + ("..." if len(dropped_features)>5 else "")
        })

        # クリーニング済みデータの保存
        df_cleaned.to_csv(os.path.join(output_dir, f"cleaned_{f}"), index=False)

    except Exception as e:
        print(f"Error in {f}: {e}")

# ==========================================
# 3. まとめレポートの出力
# ==========================================
df_report = pd.DataFrame(report_rows)

# 全データセットを合算した「合計/平均」行を追加（オプション）
total_row = pd.Series({
    "Dataset_Name": "TOTAL / MEAN",
    "Samples_Raw": df_report["Samples_Raw"].sum(),
    "Samples_Cleaned": df_report["Samples_Cleaned"].sum(),
    "Sample_Loss_Rate(%)": df_report["Sample_Loss_Rate(%)"].mean(),
    "Features_Raw": df_report["Features_Raw"].sum(),
    "Features_Active(sd>0)": df_report["Features_Active(sd>0)"].sum(),
    "Features_Dropped(sd=0)": df_report["Features_Dropped(sd=0)"].sum(),
    "Feature_Retention_Rate(%)": df_report["Feature_Retention_Rate(%)"].mean()
})
df_report = pd.concat([df_report, total_row.to_frame().T], ignore_index=True)

# 保存
report_path = os.path.join(output_dir, "Dataset_Final_Report.csv")
df_report.to_csv(report_path, index=False, encoding='utf-8-sig')

print(f"解析まとめレポートを作成しました: {report_path}")
print(df_report[["Dataset_Name", "Samples_Cleaned", "Features_Active(sd>0)", "Feature_Retention_Rate(%)"]])
```
