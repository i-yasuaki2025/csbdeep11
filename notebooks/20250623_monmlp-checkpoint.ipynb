```r
library(caret)
library(monmlp)
library(Metrics)
library(ggplot2)

# 処理対象ファイルリスト
datanamelist <- c("/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/20220127_PLS 202201 ALL/n_base.csv")

# 各ファイルごとに処理
for (dataname in datanamelist) {
  cat("Processing:", dataname, "\n")
  
  if (!file.exists(dataname)) {
    warning("File not found:", dataname)
    next
  }
  
  dataUsed01 <- read.delim(dataname, row.names=1, header=T, sep=",", as.is=TRUE)
  
  # 標準偏差ゼロの列除去
  zero_sd_cols <- which(apply(dataUsed01, 2, sd, na.rm=TRUE) == 0)
  if (length(zero_sd_cols) > 0) {
    dataUsed02 <- dataUsed01[, -zero_sd_cols]
  } else {
    dataUsed02 <- dataUsed01
  }
  
  listVar <- c("Jsc", "Voc", "FF", "PCEmax")
  
  # 説明・目的変数に分割
  dataUsed02noVar <- dataUsed02[, !colnames(dataUsed02) %in% listVar]
  dataUsed02Var <- dataUsed02[, colnames(dataUsed02) %in% listVar]
  
  # 説明変数を標準化
  dataUsed02noVarS <- scale(dataUsed02noVar)
  
  # 結合
  dataUsed03 <- data.frame(dataUsed02Var, dataUsed02noVarS)
  
  # PCEmax 欠損行除去
  dataUsed03NA <- dataUsed03[!is.na(dataUsed03$PCEmax), ]
  
  # NA列削除
  dataUsed04 <- dataUsed03NA[, colSums(is.na(dataUsed03NA)) != nrow(dataUsed03NA)]
  
  # NA行削除
  if (all(is.na(dataUsed04))) {
    warning("All rows contain NA values, skipping:", dataname)
    next
  }
  dataUsed05 <- na.omit(dataUsed04)
  
  cat("Final dataset size:", dim(dataUsed05), "\n")
  
  set.data <- data.frame(dataUsed05)
  modelname <- "monmlp"
  
  for (i in seq_along(listVar)) {
    set.mesureVar <- i
    mindex01 <- set.data[, set.mesureVar]
    set.data01 <- data.frame(mindex01, set.data[, !colnames(set.data) %in% listVar])
    
    if (nrow(set.data01) == 0) next
    
    # 定数列除去
    constant_cols <- sapply(set.data01, function(x) length(unique(x)) == 1)
    set.data01 <- set.data01[, !constant_cols]
    
    if (ncol(set.data01) <= 1) next
    
    ydata01 <- set.data01[, 1]
    xdata01 <- set.data01[, -1]
    
    cat("Training model for:", listVar[i], "on", dataname, "\n")
    
    try({
      trCtrl <- trainControl(method = "cv", number = 10)
      
      model <- train(
        mindex01 ~ ., data = set.data01,
        method = "monmlp",
        trControl = trCtrl
      )
      
      est.Y <- predict(model, set.data01)
      R <- cor(ydata01, est.Y)
      R2 <- R^2
      RMSE01 <- rmse(ydata01, est.Y)
      
      cat(sprintf("→ %s %s %s : R² = %.3f  RMSE = %.3f\n", dataname, listVar[i], modelname, R2, RMSE01))
      
      # 描画
      df_plot <- data.frame(Observed = ydata01, Predicted = est.Y)
      p <- ggplot(df_plot, aes(x = Observed, y = Predicted)) +
        geom_point(size = 2, alpha = 0.6) +
        geom_abline(slope = 1, intercept = 0, linetype = "dashed", color = "gray40") +
        coord_fixed() +
        labs(title = paste(modelname, "on", listVar[i]),
             subtitle = paste("R² =", round(R2, 3), "RMSE =", round(RMSE01, 3))) +
        theme_minimal()
      print(p)
    })
  }
}
```

```r
library(caret)
library(Metrics)
library(ggplot2)

# データの読み込み
data_path <- "/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/20220127_PLS 202201 ALL/n_base.csv"
data_raw <- read.delim(data_path, header = TRUE, sep = ",", row.names = 1)

# 標準偏差0の列を除去
zero_sd_cols <- sapply(data_raw, function(x) sd(x, na.rm=TRUE)) == 0
data_filtered <- data_raw[, !zero_sd_cols]

# 目的変数と説明変数に分離
target_vars <- c("Jsc", "Voc", "FF", "PCEmax")
x_data <- data_filtered[, !colnames(data_filtered) %in% target_vars]
x_scaled <- scale(x_data)
data_all <- data.frame(data_filtered[, target_vars], x_scaled)

# 欠損値除去
data_clean <- na.omit(data_all)

# 学習と可視化ループ
for (target in target_vars) {
  cat("Training model for:", target, "\n")
  
  data_model <- data_clean[, c(target, setdiff(colnames(data_clean), target_vars))]
  colnames(data_model)[1] <- "y"
  
  set.seed(123)
  fit <- train(
    y ~ ., 
    data = data_model,
    method = "monmlp",  # monmlpが使えない場合は "nnet"
    trControl = trainControl(method = "cv", number = 10),
    preProcess = c("center", "scale")
  )
  
  pred <- predict(fit, data_model)
  actual <- data_model$y
  
  r2 <- cor(actual, pred)^2
  rmse_val <- rmse(actual, pred)
  mae_val <- mae(actual, pred)
  
  cat(sprintf("Model: %s | R² = %.3f | RMSE = %.3f | MAE = %.3f\n", target, r2, rmse_val, mae_val))
  
  # プロット
  gg <- ggplot(data.frame(Actual = actual, Predicted = pred), aes(x = Actual, y = Predicted)) +
    geom_point(color = "blue", alpha = 0.7) +
    geom_abline(intercept = 0, slope = 1, linetype = "dashed", color = "red") +
    coord_fixed() +
    labs(title = paste("monmlp Prediction -", target),
         subtitle = sprintf("R² = %.3f, RMSE = %.3f", r2, rmse_val)) +
    theme_minimal()
  
  print(gg)
}
```

```r
# 必要なパッケージの読み込み
library(caret)
library(Metrics)
library(ggplot2)
library(monmlp)  # 明示的に必要
library(lattice)

# 対象ファイル名一覧（WM用）
file_names <- c(
  "n_base.csv", "n_base_OH.csv", "n_base_FP.csv",
  "n_all.csv", "n_all_OH.csv", "n_all_FP.csv",
  "m_base.csv", "m_base_OH.csv", "m_base_FP.csv",
  "m_all.csv", "m_all_OH.csv", "m_all_FP.csv"
)

# データパス
base_path <- "/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/20220127_PLS 202201 ALL/"

# 日付
today <- format(Sys.Date(), "%Y%m%d")

# 結果格納行列
result_matrix <- matrix(nrow = 5, ncol = length(file_names))
rownames(result_matrix) <- c("Best_hidden", "R2_PCEmax", "RMSE_PCEmax", "MAE_PCEmax", "RPD_PCEmax")
colnames(result_matrix) <- file_names

# 軸スケール決定関数
get_axis_limit <- function(values) {
  max_val <- max(values, na.rm = TRUE)
  if (max_val <= 1.5) {
    return(ceiling(max_val / 0.1) * 0.1)
  } else if (max_val <= 5) {
    return(ceiling(max_val / 0.5) * 0.5)
  } else if (max_val <= 30) {
    return(ceiling(max_val / 2) * 2)
  } else {
    return(ceiling(max_val / 5) * 5)
  }
}

#--- メインループ ---#
for (file in file_names) {
  filepath <- paste0(base_path, file)
  cat("\n=== Processing:", file, "===\n")
  df_all <- read.csv(filepath)
  cat("Final dataset size:", dim(df_all)[1], dim(df_all)[2], "\n")

  target_vars <- c("Jsc", "Voc", "FF", "PCEmax")
  feature_vars <- setdiff(colnames(df_all), target_vars)

  for (target_var in target_vars) {
    cat("\n---\nTraining model for:", target_var, "on", file, "\n")
    df <- df_all[complete.cases(df_all[[target_var]]), c(feature_vars, target_var)]

    # ハイパーパラメータチューニング：隠れ層のノード数を1層で調整
    tune_grid <- expand.grid(hidden1 = 1:5, hidden2 = 0)

    ctrl <- trainControl(method = "cv", number = 10)

    model <- train(
      formula(paste(target_var, "~ .")),
      data = df,
      method = "monmlp",
      metric = "RMSE",
      trControl = ctrl,
      tuneGrid = tune_grid
    )

    pred <- predict(model, df)
    obs <- df[[target_var]]

    R2 <- round(cor(pred, obs)^2, 3)
    RMSE_val <- round(rmse(obs, pred), 3)
    MAE_val <- round(mae(obs, pred), 3)
    RPD_val <- round(sd(obs) / RMSE_val, 3)

    best_hidden <- model$bestTune$hidden1
    cat("Best hidden1:", best_hidden, "\n")
    cat(file, target_var, ": R2 =", R2, ", RMSE =", RMSE_val, ", MAE =", MAE_val, ", RPD =", RPD_val, "\n")

    if (target_var == "PCEmax") {
      result_matrix["Best_hidden", file] <- best_hidden
      result_matrix["R2_PCEmax", file] <- R2
      result_matrix["RMSE_PCEmax", file] <- RMSE_val
      result_matrix["MAE_PCEmax", file] <- MAE_val
      result_matrix["RPD_PCEmax", file] <- RPD_val
    }

    # プロット
    range_max <- get_axis_limit(c(obs, pred))

    p <- ggplot(data.frame(Predicted = pred, Observed = obs), aes(x = Observed, y = Predicted)) +
      geom_point(color = "blue", alpha = 0.7) +
      geom_abline(slope = 1, intercept = 0, linetype = "dashed", color = "red") +
      scale_x_continuous(limits = c(0, range_max)) +
      scale_y_continuous(limits = c(0, range_max)) +
      coord_fixed(ratio = 1) +
      labs(
        title = paste0(target_var, " Prediction (", file, ")"),
        x = "Observed",
        y = "Predicted",
        subtitle = paste0("R² = ", R2, ", RMSE = ", RMSE_val, ", MAE = ", MAE_val, ", RPD = ", RPD_val)
      ) +
      theme_minimal()

    print(p)
  }
}

#--- 結果保存 ---#
write.csv(result_matrix, paste0("monmlp_summary_", today, ".csv"), row.names = TRUE)
cat("\nSummary saved as monmlp_summary_", today, ".csv\n")
```
