```python
# === 設定 ===

from pathlib import Path
from datetime import datetime

# ルート
BASE = Path("/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/cal_cluster_No")   # 変更可
RUN_ROOT = BASE / "20251026_under_25clusters"                                # 変更可

# 出力フォルダ名（未指定なら自動命名）
OUT_NAME = None  # 例: "STEP3.3_onlyMat_20251027"
# OUT_NAME = "STEP3.3_onlyMat_20251027"   # ← 明示したい場合はコメントアウト解除

# 変数リストの与え方を選択：
USE_INLINE_LISTS = False   # True: 下の OH_INLINE / FP_INLINE を使用, False: CSVから読む

# --- インライン指定する場合（USE_INLINE_LISTS=True にした上で編集）---
OH_INLINE = "SMILESsnameinM_M1, SMILESsnamep1M_M2"  # 例；空でもOK
FP_INLINE = "123, 456.1, 789.4"                     # 例；空でもOK

# --- CSVから読む場合（USE_INLINE_LISTS=False のとき使用）---
OH_LIST_CSV = BASE / "features_OH_onlyMat.csv"
FP_LIST_CSV = BASE / "features_FP_onlyMat.csv"

# 変数名の一致モード（"exact" or "contains"）
MATCH_MODE = "exact"  # 通常は "exact" 推奨。不一致が続くときだけ "contains" を試す
```

```python
import pandas as pd
import numpy as np

NAME_HINTS = [
    "variable","var","name","feature","feat","column","col","field",
    "descriptor","smiles","sname","id","xid","token","label"
]

def norm_series(s: pd.Series) -> pd.Series:
    return (
        s.astype(str)
         .str.strip()
         .str.replace(r"\s+", " ", regex=True)
         .str.replace("\u3000", " ")  # 全角空白対策
    )

def parse_csv_list_text(s: str) -> set:
    if not s or not str(s).strip():
        return set()
    items = [x.strip() for x in str(s).split(",")]
    items = [x for x in items if x != ""]
    return set(items)

def candidate_name_columns(df: pd.DataFrame):
    cols = list(df.columns)
    hinted = [c for c in cols if any(h in str(c).lower() for h in NAME_HINTS)]
    if df.columns[0] not in hinted:
        hinted = [df.columns[0]] + hinted
    out = []
    for c in hinted + cols:
        if c not in out:
            out.append(c)
    return out

def choose_best_name_col(df: pd.DataFrame, keep_set: set, match_mode: str = "exact"):
    best_col, best_hits = None, -1
    stats = []
    for c in candidate_name_columns(df):
        s = norm_series(df[c])
        if match_mode == "exact":
            hits = int(s.isin(keep_set).sum())
        else:
            # containsモードは簡易評価（速度重視）
            ks = list(keep_set)
            sample = s if len(s) <= 2000 else s.sample(2000, random_state=0)
            hits = 0
            for v in sample:
                if any(k in v for k in ks):
                    hits += 1
            # 簡易スケール補正
            hits = int(hits * (len(s) / max(1, len(sample))))
        stats.append((c, hits))
        if hits > best_hits:
            best_col, best_hits = c, hits
    stats.sort(key=lambda x: x[1], reverse=True)
    return best_col, best_hits, stats

def read_keep_set_from_csv(p: Path) -> set:
    df = pd.read_csv(p)
    # 最もユニーク数が多い列を“名前列”として採用（素性CSVの列名が不明でもOK）
    best_col, best_u = None, -1
    for c in df.columns:
        u = df[c].nunique(dropna=True)
        if u > best_u:
            best_col, best_u = c, u
    return set(norm_series(df[best_col]).tolist())
```

```python
# keepセットの準備
if USE_INLINE_LISTS:
    keep_OH = set(norm_series(pd.Series(list(parse_csv_list_text(OH_INLINE)))))
    keep_FP = set(norm_series(pd.Series(list(parse_csv_list_text(FP_INLINE)))))
else:
    keep_OH = read_keep_set_from_csv(OH_LIST_CSV)
    keep_FP = read_keep_set_from_csv(FP_LIST_CSV)

print(f"keep_OH: {len(keep_OH)} features")
print(f"keep_FP: {len(keep_FP)} features")

# 出力ルート
if OUT_NAME is None or str(OUT_NAME).strip() == "":
    OUT_NAME = f"STEP3.3_onlyMat_{datetime.now().strftime('%Y%m%d_%H%M%S')}"
OUT_ROOT = RUN_ROOT / OUT_NAME
OUT_OH = OUT_ROOT / "OH" / "04_assignments_onlyMat"
OUT_FP = OUT_ROOT / "FP" / "04_assignments_onlyMat"
OUT_LOGS = OUT_ROOT / "logs"

for p in [OUT_OH, OUT_FP, OUT_LOGS]:
    p.mkdir(parents=True, exist_ok=True)

OUT_ROOT, OUT_OH, OUT_FP, OUT_LOGS
```

```python
from pathlib import Path

def filter_one_file(in_path: Path, keep_set: set, out_dir: Path, match_mode: str = "exact"):
    df = pd.read_csv(in_path)
    if df.empty:
        return {
            "input_file": str(in_path),
            "output_file": "",
            "name_col_used": "",
            "match_mode": match_mode,
            "n_rows_before": 0,
            "n_rows_after": 0,
            "n_dropped": 0,
            "overlap_hits_on_used_col": 0,
            "side": "",
            "note": "empty_input_df"
        }
    if len(keep_set) == 0:
        return {
            "input_file": str(in_path),
            "output_file": "",
            "name_col_used": "",
            "match_mode": match_mode,
            "n_rows_before": len(df),
            "n_rows_after": 0,
            "n_dropped": len(df),
            "overlap_hits_on_used_col": 0,
            "side": "",
            "note": "empty_keep_set"
        }

    name_col, hits, stats = choose_best_name_col(df, keep_set, match_mode)
    s = norm_series(df[name_col])
    if match_mode == "exact":
        mask = s.isin(keep_set)
    else:
        mask = s.apply(lambda v: any(k in v for k in keep_set))

    before = len(df)
    df2 = df.loc[mask].copy()
    after = len(df2)

    out_path = out_dir / (in_path.stem + "_onlyMat" + in_path.suffix)
    df2.to_csv(out_path, index=False, encoding="utf-8-sig")

    rec = {
        "input_file": str(in_path),
        "output_file": str(out_path),
        "name_col_used": name_col,
        "match_mode": match_mode,
        "n_rows_before": before,
        "n_rows_after": after,
        "n_dropped": before - after,
        "overlap_hits_on_used_col": int(hits),
        "top_namecol_candidates": [{"col": c, "hits": h} for c,h in stats[:5]]
    }
    return rec
```

```python
# 入力探索（再帰）
files_oh = sorted(RUN_ROOT.glob("**/OH/04_assignments/ClusterAssign_*_OH_*.csv"))
files_fp = sorted(RUN_ROOT.glob("**/FP/04_assignments/ClusterAssign_*_FP_*.csv"))

logs = []

# OH 側
if len(keep_OH) > 0:
    for f in files_oh:
        try:
            rec = filter_one_file(f, keep_OH, OUT_OH, MATCH_MODE)
            rec["side"] = "OH"
            logs.append(rec)
        except Exception as e:
            logs.append({"side":"OH","input_file":str(f),"error":str(e)})

# FP 側
if len(keep_FP) > 0:
    for f in files_fp:
        try:
            rec = filter_one_file(f, keep_FP, OUT_FP, MATCH_MODE)
            rec["side"] = "FP"
            logs.append(rec)
        except Exception as e:
            logs.append({"side":"FP","input_file":str(f),"error":str(e)})

log_df = pd.DataFrame(logs)
log_df.to_csv(OUT_LOGS / "filter_onlyMat_index_ALL.csv", index=False, encoding="utf-8-sig")
log_df.to_json(OUT_LOGS / "filter_onlyMat_index_ALL.json", orient="records", force_ascii=False, indent=2)

# 空出力の注意喚起
empty = log_df[(log_df.get("n_rows_after").fillna(0) == 0)]
if not empty.empty:
    empty.to_csv(OUT_LOGS / "ATTN_empty_outputs.csv", index=False, encoding="utf-8-sig")

print("=== Done ===")
print("OUT_ROOT:", OUT_ROOT)
print("OH out  :", OUT_OH)
print("FP out  :", OUT_FP)
print("Logs    :", OUT_LOGS)

display(log_df.head(10))
```
