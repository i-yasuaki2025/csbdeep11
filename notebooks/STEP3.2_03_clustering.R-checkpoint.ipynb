```r
############################################################
# STEP3.2_03_clustering.R
#
# - 入力: STEP3.2_02_mds.R で作成された MDS 座標
#     sub/02_mds_STEP3.2_signlessCorr/run_<run_id>/<unit>/<dataset>/
#         MDScoords_linear_top3_<unit>_<dataset>_<run_id>.csv
#         MDScoords_linear_cumeig_<unit>_<dataset>_<run_id>.csv
#         MDScoords_nonlinear_top3_<unit>_<dataset>_<run_id>.csv
#         MDScoords_nonlinear_cumeig_<unit>_<dataset>_<run_id>.csv
#
# - 出力: クラスタリング結果
#     sub/03_clustering_STEP3.2_signlessCorr/run_<run_id>/<unit>/<dataset>/
#         labels/
#           ClusterAssign_<mode>_<dim>_<index>_<unit>_<dataset>_<run_id>.csv
#           ClusterAssign_<mode>_<dim>_k10_<unit>_<dataset>_<run_id>.csv など
#         indices/
#           k_score_<mode>_<dim>_<index>_<unit>_<dataset>_<run_id>.csv
#         plots/
#           NbClust_AllIndex_<mode>_<dim>_<index>_<unit>_<dataset>_<run_id>.png
#           MDS12_scatter_<mode>_<dim>_<index>_<unit>_<dataset>_<run_id>.png
############################################################

## パッケージ読み込み -------------------------------------------------------
if (!require(NbClust)) { install.packages("NbClust"); library(NbClust) }
if (!require(ggplot2)) { install.packages("ggplot2"); library(ggplot2) }
if (!require(ggrepel)) { install.packages("ggrepel"); library(ggrepel) }
if (!require(scales))  { install.packages("scales");  library(scales) }
if (!require(dplyr))   { install.packages("dplyr");   library(dplyr) }

set.seed(42)

############################################################
# 0. パス設定 & 基本設定
############################################################

root_base <- "/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/cal_cluster_No/20251026_under_25clusters_program_and_result/for_checking_20251127"
sub_base  <- file.path(root_base, "sub")

step02_base <- file.path(sub_base, "02_mds_STEP3.2_signlessCorr")
step03_base <- file.path(sub_base, "03_clustering_STEP3.2_signlessCorr")
dir.create(step03_base, showWarnings = FALSE, recursive = TRUE)

# 対象データセット
datasets_variables <- c("OH", "FP")
datasets_samples   <- c("A_OH_plus_others",
                        "B_FP_plus_others",
                        "C_OH_only",
                        "D_FP_only")

# NbClust 指標
index_list <- c("silhouette","dunn","gap","ch","db","ptbiserial")

############################################################
# 1. run_id の取得
############################################################

# ★ Jupyter から使うときはここを書き換える：
#    02_mds で使用した run_id を指定する
# 例: override_run_id <- "20251128_193916"
override_run_id <- NA_character_

get_run_id <- function() {
  # 1) override_run_id があればそれを最優先
  if (!is.na(override_run_id) &&
      is.character(override_run_id) &&
      grepl("^\\d{8}_\\d{6}$", override_run_id)) {
    message("[INFO] Using override_run_id = ", override_run_id)
    return(override_run_id)
  }
  # 2) Rscript 実行時の引数を使う (Rscript STEP3.2_03_clustering.R 20251128_193916)
  args <- commandArgs(trailingOnly = TRUE)
  if (length(args) >= 1 && grepl("^\\d{8}_\\d{6}$", args[1])) {
    return(args[1])
  }
  # 3) どちらも無い場合はエラー
  stop("Usage: Rscript STEP3.2_03_clustering.R <run_id (YYYYMMDD_HHMMSS)>\n",
       "  - or set override_run_id <- \"YYYYMMDD_HHMMSS\" at the top of this script.")
}

run_id <- get_run_id()
cat(">>> STEP3.2_03_clustering — run_id:", run_id, "\n")

step02_run_dir <- file.path(step02_base, paste0("run_", run_id))
step03_run_dir <- file.path(step03_base, paste0("run_", run_id))
dir.create(step03_run_dir, showWarnings = FALSE, recursive = TRUE)

cat(">>> STEP3.2_03_clustering input dir:  ", normalizePath(step02_run_dir), "\n")
cat(">>> STEP3.2_03_clustering output dir: ", normalizePath(step03_run_dir), "\n")

############################################################
# 2. ヘルパー関数
############################################################

# NbClust ラッパー
run_nbclust <- function(X, index) {
  tryCatch({
    NbClust(
      data = X, diss = NULL, distance = "euclidean",
      min.nc = 2, max.nc = min(25, nrow(X) - 1),
      method = "ward.D2", index = index
    )
  }, error = function(e) {
    warning(paste("NbClust failed:", index, "-", e$message))
    NULL
  })
}

# NbClust の k vs score CSV & プロット
save_k_score_plot <- function(res, cond, idx, unit, dataset, out_dir, run_id) {
  if (is.null(res$All.index)) return()
  
  idx_vals <- as.numeric(res$All.index)
  k_labels <- names(res$All.index)
  k_seq <- if (is.null(k_labels) || any(k_labels == "")) {
    seq_along(idx_vals)
  } else {
    suppressWarnings(as.numeric(k_labels))
  }
  if (any(is.na(k_seq))) k_seq <- seq_along(idx_vals)
  
  df_k_score <- data.frame(k = k_seq, score = idx_vals)
  
  csv_file <- file.path(
    out_dir,
    sprintf("k_score_%s_%s_%s_%s_%s.csv", cond, unit, dataset, idx, run_id)
  )
  write.csv(df_k_score, csv_file, row.names = FALSE)
  
  df_idx <- data.frame(k = k_seq, value = idx_vals)
  p_idx <- ggplot(df_idx, aes(k, value)) +
    geom_line(linewidth = 1) + geom_point(size = 2) +
    labs(
      title    = paste0("NbClust All.index — ", idx),
      subtitle = paste0(cond, " | ", unit, " / ", dataset),
      x = "k", y = "Index value"
    ) +
    theme_minimal(base_size = 12) +
    theme(panel.grid.minor = element_blank(),
          axis.title = element_text(face = "bold"),
          plot.title = element_text(face = "bold"))
  
  png_file <- file.path(
    out_dir,
    sprintf("NbClust_AllIndex_%s_%s_%s_%s_%s.png", cond, unit, dataset, idx, run_id)
  )
  ggsave(png_file, p_idx, width = 7, height = 5, dpi = 300)
}

# MDS 2次元散布図（クラスタ着色）
save_mds_scatter <- function(X, grp, mode, dim_mode, label_tag,
                             unit, dataset, k_info, out_dir, run_id,
                             method_tag = "NbClust") {
  df_plot <- data.frame(
    MDS1    = X[, 1],
    MDS2    = X[, 2],
    Cluster = as.factor(grp),
    ID      = rownames(X)
  )
  rng <- range(c(df_plot$MDS1, df_plot$MDS2), na.rm = TRUE)
  pad <- diff(rng) * 0.05
  lims <- c(min(rng) - pad, max(rng) + pad)
  
  title_main <- sprintf("MDS (%s/%s) — Ward.D2 × %s", mode, dim_mode, method_tag)
  subtitle_text <- sprintf("%s | %s | %s / %s",
                           k_info, label_tag, unit, dataset)
  
  p_sc <- ggplot(df_plot, aes(MDS1, MDS2, color = Cluster)) +
    stat_ellipse(aes(group = Cluster), type = "norm", level = 0.95,
                 linewidth = 0.6, alpha = 0.25) +
    geom_point(size = 2.2, alpha = 0.8) +
    coord_equal(xlim = lims, ylim = lims, expand = TRUE) +
    scale_color_hue(h = c(0, 360), l = 55, c = 90) +
    labs(
      title    = title_main,
      subtitle = subtitle_text,
      x = "Dim 1", y = "Dim 2", color = "Cluster"
    ) +
    theme_minimal(base_size = 12) +
    theme(panel.grid.major = element_line(linewidth = 0.3, linetype = "dashed"),
          panel.grid.minor = element_blank(),
          axis.title = element_text(face = "bold"),
          plot.title = element_text(face = "bold"))
  
  png_file <- file.path(
    out_dir,
    sprintf("MDS12_scatter_%s_%s_%s_%s_%s.png",
            paste0(mode, "_", dim_mode, "_", label_tag),
            unit, dataset, method_tag, run_id)
  )
  ggsave(png_file, p_sc, width = 7, height = 6, dpi = 300)
}

############################################################
# 3. 1データセットのクラスタリング処理
############################################################

process_one_dataset_clust <- function(unit, dataset) {
  message("\n[STEP3.2_03] unit = ", unit, ", dataset = ", dataset)
  
  # MDS 座標の入力ディレクトリ
  mds_dir_ds <- file.path(step02_run_dir, unit, dataset)
  if (!dir.exists(mds_dir_ds)) {
    stop("[ERROR] MDS dir not found: ", mds_dir_ds)
  }
  
  # 出力ディレクトリ
  out_dir_ds <- file.path(step03_run_dir, unit, dataset)
  dir.create(out_dir_ds, showWarnings = FALSE, recursive = TRUE)
  labels_dir <- file.path(out_dir_ds, "labels")
  idx_dir    <- file.path(out_dir_ds, "indices")
  plots_dir  <- file.path(out_dir_ds, "plots")
  dir.create(labels_dir, showWarnings = FALSE, recursive = TRUE)
  dir.create(idx_dir,    showWarnings = FALSE, recursive = TRUE)
  dir.create(plots_dir,  showWarnings = FALSE, recursive = TRUE)
  
  # 4モードの MDS 座標を読み込み
  mode_dims <- c("linear_top3", "linear_cumeig",
                 "nonlinear_top3", "nonlinear_cumeig")
  coords_list <- list()
  
  for (mode_dim in mode_dims) {
    coord_file <- file.path(
      mds_dir_ds,
      sprintf("MDScoords_%s_%s_%s_%s.csv", mode_dim, unit, dataset, run_id)
    )
    if (!file.exists(coord_file)) {
      stop("[ERROR] MDS coord file not found: ", coord_file)
    }
    X <- as.matrix(read.csv(coord_file, row.names = 1, check.names = FALSE))
    coords_list[[mode_dim]] <- X
  }
  
  # 代表 ID
  ids <- rownames(coords_list[[1]])
  
  # NbClust & k固定クラスタリング
  for (mode_dim in names(coords_list)) {
    X <- coords_list[[mode_dim]]
    if (nrow(X) < 3) {
      warning("[WARN] nrow(X) < 3 for ", unit, "/", dataset,
              " (", mode_dim, "); skipping this mode_dim.")
      next
    }
    
    parts   <- strsplit(mode_dim, "_")[[1]]
    mode    <- parts[1]  # "linear" or "nonlinear"
    dim_mode <- parts[2] # "top3" or "cumeig"
    
    cond_tag <- sprintf("%s_%s", mode, dim_mode)
    
    message("  [INFO] Clustering for mode_dim = ", cond_tag)
    
    ## 3-1. NbClust 6指標ごと
    for (idx in index_list) {
      message("    [NbClust] index = ", idx)
      
      res <- run_nbclust(X, idx)
      if (is.null(res)) {
        warning("    [WARN] NbClust failed for ", cond_tag, " / ", idx)
        next
      }
      
      # cluster labels
      grp <- as.factor(res$Best.partition)
      grp <- grp[ids]  # ID順に揃える
      
      # 推奨クラスタ数 k の抽出
      k_val <- NA_integer_
      if (!is.null(res$Best.nc)) {
        if (is.list(res$Best.nc) && !is.null(res$Best.nc[["Number_clusters"]])) {
          k_val <- as.integer(res$Best.nc[["Number_clusters"]])
        } else if (is.matrix(res$Best.nc) &&
                   "Number_clusters" %in% rownames(res$Best.nc)) {
          k_val <- as.integer(res$Best.nc["Number_clusters", 1])
        } else if (is.numeric(res$Best.nc)) {
          k_val <- as.integer(res$Best.nc[1])
        }
      }
      
      # ラベル CSV
      lab_file <- file.path(
        labels_dir,
        sprintf("ClusterAssign_%s_%s_%s_%s_%s.csv",
                mode, dim_mode, idx, unit, dataset, run_id)
      )
      write.csv(
        data.frame(ID = ids, Cluster = grp),
        lab_file, row.names = FALSE
      )
      
      # k vs score & NbClust AllIndex プロット
      save_k_score_plot(
        res, cond = cond_tag, idx = idx,
        unit = unit, dataset = dataset,
        out_dir = idx_dir, run_id = run_id
      )
      
      # MDS 2次元散布図 (NbClust結果)
      save_mds_scatter(
        X, grp,
        mode = mode, dim_mode = dim_mode,
        label_tag = idx,
        unit = unit, dataset = dataset,
        k_info = sprintf("k = %s", ifelse(is.na(k_val), "NA", k_val)),
        out_dir = plots_dir, run_id = run_id,
        method_tag = "NbClust"
      )
    } # end index loop
    
    ## 3-2. k固定クラスタリング (10, 25, 50)
    message("    [fixed-k] k = 10, 25, 50")
    hc <- hclust(dist(X), method = "ward.D2")
    for (k_fixed in c(10, 25, 50)) {
      if (k_fixed >= nrow(X)) {
        message("      [INFO] skip k_fixed = ", k_fixed,
                " (>= nrow(X) = ", nrow(X), ")")
        next
      }
      grp_k <- cutree(hc, k = k_fixed)
      grp_k <- grp_k[ids]
      
      # ラベル CSV
      lab_file_k <- file.path(
        labels_dir,
        sprintf("ClusterAssign_%s_%s_k%d_%s_%s_%s.csv",
                mode, dim_mode, k_fixed, unit, dataset, run_id)
      )
      write.csv(
        data.frame(ID = ids, Cluster = grp_k),
        lab_file_k, row.names = FALSE
      )
      
      # MDS 2次元散布図 (fixed-k)
      save_mds_scatter(
        X, grp_k,
        mode = mode, dim_mode = dim_mode,
        label_tag = paste0("k", k_fixed),
        unit = unit, dataset = dataset,
        k_info = sprintf("k = %d", k_fixed),
        out_dir = plots_dir, run_id = run_id,
        method_tag = "fixed-k"
      )
    }
  } # end mode_dim loop
  
  message("[DONE] Clustering finished for unit = ", unit,
          ", dataset = ", dataset)
}

############################################################
# 4. メインループ
############################################################

cat("\n===== STEP3.2_03_clustering: Variables (OH/FP) =====\n")
for (ds in datasets_variables) {
  process_one_dataset_clust("variables", ds)
}

cat("\n===== STEP3.2_03_clustering: Samples (A/B/C/D) =====\n")
for (ds in datasets_samples) {
  process_one_dataset_clust("samples", ds)
}

cat("\n✅ STEP3.2_03_clustering finished. run_id = ", run_id, "\n")
cat("   Output: ", normalizePath(step03_run_dir), "\n")
```
