```r
# CSVを読み込み
df <- read.csv("removed_perfect_corr_groups_FP.csv", stringsAsFactors = FALSE)

# Removed列が空でない場合に、; 区切りで分割してカウント
df$Removed_count <- sapply(df$Removed, function(x) {
  if (is.na(x) || x == "") {
    return(0)
  } else {
    return(length(strsplit(x, ";")[[1]]))
  }
})

# 各グループの削除数を確認
print(df[, c("Component", "Representative", "Removed_count")])

# 全体の削除変数の総数
total_removed <- sum(df$Removed_count)
cat("✅ 削除された変数の総数:", total_removed, "\n")

# 削除された変数名をすべて集めて ", " 区切りで出力
removed_vars <- unlist(strsplit(df$Removed[df$Removed != ""], ";"))
if (length(removed_vars) > 0) {
  removed_vars_str <- paste(removed_vars, collapse = ", ")
  cat("📝 削除された変数名リスト:\n", removed_vars_str, "\n")
} else {
  cat("📝 削除された変数はありません。\n")
}
```

```r
# 読み込み
df <- read.csv("removed_perfect_corr_groups_FP.csv", stringsAsFactors = FALSE)

# Removed列を "; " 区切りから ", " 区切りへ整形
df$Removed_fmt <- sapply(df$Removed, function(x) {
  if (is.na(x) || x == "") {
    return("")
  } else {
    paste(strsplit(x, ";")[[1]], collapse = ", ")
  }
})

# Removedが空でない行のみ抽出
out_df <- df[df$Removed_fmt != "", c("Representative", "Removed_fmt")]

# 列名を「Kept」「Removed」に変更
colnames(out_df) <- c("Kept", "Removed")

# CSV保存
write.csv(out_df, "removed_vars_pairs_FP.csv", row.names = FALSE)

cat("✅ 出力しました: removed_vars_pairs_FP.csv\n")
```
