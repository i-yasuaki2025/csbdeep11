```r
# ===== Encoding-robust header v2 =====
# 1) 端末に存在するロケールだけを設定（Win/Mac/Linux で分岐）
.pick_and_set_locale <- function(){
  os <- Sys.info()[["sysname"]]
  if (is.na(os)) os <- .Platform$OS.type
  cand <- if (grepl("Windows", os, ignore.case = TRUE) || os == "Windows") {
    c("Japanese_Japan.65001",   # WinのUTF-8（新しめ）
      "Japanese_Japan.932",     # CP932（Shift_JIS）
      "English_United States.65001",
      "C")
  } else {
    c("ja_JP.UTF-8", "C.UTF-8", "en_US.UTF-8", "C")
  }
  for (lc in cand) {
    ok <- try(Sys.setlocale("LC_CTYPE", lc), silent = TRUE)
    if (!inherits(ok, "try-error") && !is.na(ok)) return(invisible(ok))
  }
}
.pick_and_set_locale()

# 2) この端末の“表示用エンコーディング”を推定
.encoding_target <- local({
  if (l10n_info()[["UTF-8"]]) return("UTF-8")
  lc <- Sys.getlocale("LC_CTYPE")
  # 例: "Japanese_Japan.932" → "CP932" / "English_United States.65001" → "UTF-8"
  enc <- sub(".*\\.", "", lc)
  if (grepl("65001|UTF-?8", enc, ignore.case = TRUE)) "UTF-8"
  else if (enc %in% c("932","CP932")) "CP932"
  else if (nzchar(enc)) enc else "ASCII"
})

# 3) 表示前に安全変換（UTF-8→端末ネイティブへ）
to_console <- function(x){
  x <- as.character(x)
  if (.encoding_target == "UTF-8") return(enc2utf8(x))
  out <- iconv(x, from = "UTF-8", to = paste0(.encoding_target, "//TRANSLIT"), sub = "?")
  out[is.na(out)] <- iconv(x[is.na(out)], from = "UTF-8", to = "ASCII//TRANSLIT", sub = "?")
  out
}

# 4) UTF-8/BOM 優先の安全CSV読込
safe_read_csv <- function(path){
  if (is.na(path) || is.null(path)) return(NULL)
  tryCatch(utils::read.csv(path, check.names = FALSE, fileEncoding = "UTF-8"),
           error = function(e)
             tryCatch(utils::read.csv(path, check.names = FALSE, fileEncoding = "UTF-8-BOM"),
                      error = function(e2) NULL))
}

# 5) 日本語/英語切替（まだ化けるなら TRUE）
use_ascii_headers <- FALSE
jp_label <- function(jp, en) if (use_ascii_headers) en else jp

# 6) 安全プリンタ
catJP      <- function(...) base::cat(to_console(paste0(..., collapse="")), "\n")
messageJP  <- function(...) base::message(to_console(paste0(..., collapse="")))
print_dfJP <- function(df, n = Inf, row.names = FALSE){
  df2 <- df
  names(df2) <- to_console(names(df2))
  for (j in seq_along(df2)) if (is.character(df2[[j]])) df2[[j]] <- to_console(df2[[j]])
  if (is.finite(n)) df2 <- utils::head(df2, n)
  print(df2, row.names = row.names)
}

# 動作確認（必要なら）
# catJP("日本語見出しテスト：寄与のクイック要約")
# print_dfJP(data.frame(項目=c("第1軸","第2軸"), 寄与率=c(0.12,0.09)))
# ===== /header v2 =====
```

```r
# ---- 環境：UTF-8安全な出力 ----
invisible(try(Sys.setlocale("LC_CTYPE", "ja_JP.UTF-8"), silent=TRUE))
say <- function(..., sep="") {
  txt <- paste0(..., collapse=sep)
  cat(enc2native(txt)); if (!grepl("\n$", txt)) cat("\n")
}
# ============================================================
# STEP3_* 結果の一括比較（コンソール要約 & ステップ間 ARI）
# - ルート: i-yasuaki/tree/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/cal_cluster_No
# - 対象フォルダ:
#     STEP3_new_MDS_20250904/
#     STEP3.2_signlessCorr_MDS_20250904/
#     STEP3.3_olddata_MDS_20250904/
#   （それぞれ配下に OH/ FP サブフォルダあり）
# - 読み取り:
#     MDS_eigen_contributions_*.csv, cluster_labels_*.csv, cluster_k_summary_*.csv
# - 出力:
#     コンソールに要約テーブルと ARI 比較（new vs signless, new vs olddata）
# ============================================================

suppressPackageStartupMessages({
  if (!require(mclust))  install.packages("mclust");  library(mclust)   # ARI
  if (!require(dplyr))   install.packages("dplyr");   library(dplyr)
  if (!require(tools))   install.packages("tools");   library(tools)
})

# ---- パス設定 ----
root_base <- "i-yasuaki/tree/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/cal_cluster_No"
steps <- list(
  new      = "STEP3_new_MDS_20250904",
  signless = "STEP3.2_signlessCorr_MDS_20250904",
  olddata  = "STEP3.3_olddata_MDS_20250904"
)
datasets <- c("OH", "FP")

# ---- ユーティリティ ----
latest_file <- function(dir_path, pattern){
  cand <- list.files(dir_path, pattern = pattern, full.names = TRUE)
  if (length(cand) == 0) return(NA_character_)
  cand[order(file.info(cand)$mtime, decreasing = TRUE)][1]
}

safe_read_csv <- function(path){
  if (is.na(path)) return(NULL)
  tryCatch(utils::read.csv(path, check.names = FALSE), error = function(e) NULL)
}

# signless の列名を new/olddata と比較可能な「標準名」に揃える
#   例: "linear_top3_silhouette" -> "top3_silhouette"
#   ※ nonlinear_* は new/olddata には無いので、比較対象から除外する方針
normalize_label_cols <- function(df, step_key){
  if (is.null(df)) return(NULL)
  if (!"id" %in% names(df)) return(NULL)
  out <- df
  cols <- setdiff(names(out), "id")
  if (length(cols) == 0) return(out)

  if (step_key == "signless"){
    # linear_* のみを採用、nonlinear_* は除外
    keep <- grepl("^linear_", cols)
    cols_keep <- cols[keep]
    if (length(cols_keep) == 0) {
      # 何も一致しない場合は NULL
      return(data.frame(id = out$id, check.names = FALSE))
    }
    # "linear_" を外した標準名へ
    std_names <- sub("^linear_", "", cols_keep)
    # 既存に同名があれば後勝ち上書き
    for (i in seq_along(cols_keep)) {
      nm_old <- cols_keep[i]; nm_new <- std_names[i]
      out[[nm_new]] <- out[[nm_old]]
    }
    # 余分な列（nonlinear_*や元のlinear_*）を削除
    drop_cols <- cols[!keep]
    out <- out[, setdiff(c("id", names(out)), c(cols, drop_cols)), drop = FALSE]
  } else {
    # new / olddata はそのまま（"top3_*", "cumeig_*" など前提）
    # 余計な空列や全NA列を落とす
    drop_mask <- sapply(out[, cols, drop=FALSE], function(v) all(is.na(v)))
    out <- cbind(out["id"], out[, cols[!drop_mask], drop = FALSE])
  }
  out
}

# kサマリの正規化（stepによって列違いに対応）
normalize_ksum <- function(df, step_key){
  if (is.null(df)) return(NULL)
  cn <- names(df)
  df2 <- df

  # signless: mode(=linear/nonlinear), dim_mode(=top3/cumeig), index, k
  # new/olddata: mode(=top3/cumeig) だけの場合がある
  if ("dim_mode" %in% cn && "mode" %in% cn) {
    names(df2)[names(df2)=="dim_mode"] <- "dim_mode"
    names(df2)[names(df2)=="mode"] <- "mode_type"
  } else if ("mode" %in% cn) {
    # "mode" が top3/cumeig を指しているケース
    names(df2)[names(df2)=="mode"] <- "dim_mode"
    df2$mode_type <- "linear"  # 比較都合上、new/olddataは linear 相当とみなす
  } else {
    df2$dim_mode <- NA_character_
    df2$mode_type <- NA_character_
  }
  if (!"index" %in% names(df2)) df2$index <- NA_character_
  if (!"k" %in% names(df2)) df2$k <- NA_integer_

  df2[, c("mode_type","dim_mode","index","k")]
}

# 寄与CSVから 80% 到達次元（正の固有値基準）と top3累積を計算
eigen_quick_summary <- function(eig_df){
  if (is.null(eig_df)) return(list(pos_k80 = NA_integer_, top3_cum_pos = NA_real_))
  # 正の固有値寄与列の名称は new: Share_pos / CumShare_pos、signless もほぼ同様
  # あれば使う。無いときは NA。
  share_pos <- eig_df$Share_pos
  cum_pos   <- eig_df$CumShare_pos
  if (is.null(share_pos) || is.null(cum_pos)) return(list(pos_k80 = NA_integer_, top3_cum_pos = NA_real_))
  pos_idx <- which(!is.na(share_pos) & share_pos > 0)
  if (length(pos_idx) == 0) return(list(pos_k80 = NA_integer_, top3_cum_pos = NA_real_))
  k80 <- which(cum_pos[pos_idx] >= 0.8)[1]
  if (is.na(k80)) k80 <- length(pos_idx)
  top3 <- sum(share_pos[pos_idx][1:min(3,length(pos_idx))], na.rm = TRUE)
  list(pos_k80 = k80, top3_cum_pos = top3)
}

# ステップ間の ARI を列ごとに計算（共通の id & 共通の列名のみ）
#   signless は linear_* を標準化して new/olddata と比較可能にしている前提
pairwise_ari <- function(dfA, dfB){
  if (is.null(dfA) || is.null(dfB)) return(NULL)
  if (!("id" %in% names(dfA) && "id" %in% names(dfB))) return(NULL)

  colsA <- setdiff(names(dfA), "id")
  colsB <- setdiff(names(dfB), "id")
  common_cols <- intersect(colsA, colsB)
  if (length(common_cols) == 0) return(NULL)

  merged <- merge(dfA, dfB, by = "id", suffixes = c(".A",".B"))
  out <- lapply(common_cols, function(col){
    a <- merged[[paste0(col,".A")]]
    b <- merged[[paste0(col,".B")]]
    ok <- !(is.na(a) | is.na(b))
    if (sum(ok) < 2) {
      data.frame(condition = col, n = sum(ok), ARI = NA_real_)
    } else {
      data.frame(condition = col, n = sum(ok),
                 ARI = mclust::adjustedRandIndex(as.factor(a[ok]), as.factor(b[ok])))
    }
  })
  dplyr::bind_rows(out)
}

# ---- データ読み込み＆要約 ----
all_eigen_summary <- list()
all_k_summary     <- list()
all_labels        <- list()

for (step_key in names(steps)) {
  for (ds in datasets) {
    dir_ds <- file.path(root_base, steps[[step_key]], ds)

    # 1) eigen contributions
    f_eig <- latest_file(dir_ds, "^MDS_eigen_contributions_.*\\.csv$")
    eig_df <- safe_read_csv(f_eig)
    es <- eigen_quick_summary(eig_df)
    all_eigen_summary[[paste(step_key, ds, sep=":")]] <- data.frame(
      step = step_key, dataset = ds,
      pos_k80 = es$pos_k80, top3_cum_pos = es$top3_cum_pos,
      rows_in_file = if (!is.null(eig_df)) nrow(eig_df) else NA_integer_,
      file = f_eig,
      stringsAsFactors = FALSE
    )

    # 2) cluster_k_summary
    f_ksum <- latest_file(dir_ds, "^cluster_k_summary_.*\\.csv$")
    ksum_df <- safe_read_csv(f_ksum)
    ksum_norm <- normalize_ksum(ksum_df, step_key)
    all_k_summary[[paste(step_key, ds, sep=":")]] <- cbind(
      data.frame(step = step_key, dataset = ds, stringsAsFactors = FALSE),
      ksum_norm %||% data.frame(mode_type=NA, dim_mode=NA, index=NA, k=NA)
    )

    # 3) cluster_labels（比較の主役）
    f_lbl <- latest_file(dir_ds, "^cluster_labels_.*\\.csv$")
    lbl_df <- safe_read_csv(f_lbl)
    lbl_std <- normalize_label_cols(lbl_df, step_key)
    all_labels[[paste(step_key, ds, sep=":")]] <- lbl_std
  }
}

# ---- 出力：寄与とkの要約（コンソール）----
cat("\n================ 寄与のクイック要約（正の固有値基準） ================\n")
eig_summary_tbl <- dplyr::bind_rows(all_eigen_summary)
if (nrow(eig_summary_tbl)) {
  eig_summary_tbl <- eig_summary_tbl %>%
    dplyr::mutate(top3_cum_pos_pct = round(100*top3_cum_pos, 2)) %>%
    dplyr::select(step, dataset, pos_k80, top3_cum_pos_pct, rows_in_file, file)
  print(eig_summary_tbl, row.names = FALSE)
} else {
  cat("（該当ファイルが見つかりませんでした）\n")
}

cat("\n================ クラスタ数（k）サマリ ================\n")
ksum_tbl <- dplyr::bind_rows(all_k_summary)
if (nrow(ksum_tbl)) {
  # 見やすく集計： step×dataset×dim_mode×index×mode_type で代表値（NA除く）
  ksum_view <- ksum_tbl %>%
    dplyr::group_by(step, dataset, mode_type, dim_mode, index) %>%
    dplyr::summarise(k = if (all(is.na(k))) NA_integer_ else as.integer(stats::median(k, na.rm=TRUE)),
                     .groups = "drop") %>%
    dplyr::arrange(dataset, step, mode_type, dim_mode, index)
  print(ksum_view, n=Inf)
} else {
  cat("（該当ファイルが見つかりませんでした）\n")
}

# ---- ステップ間 ARI（new vs signless、new vs olddata）----
for (ds in datasets){
  cat("\n================ ARI（", ds, "） ================\n", sep="")
  lbl_new      <- all_labels[[paste("new", ds, sep=":")]]
  lbl_signless <- all_labels[[paste("signless", ds, sep=":")]]
  lbl_olddata  <- all_labels[[paste("olddata", ds, sep=":")]]
  # signless は linear_* を標準名化済み。nonlinear_*は比較対象外。

    # （dir_ds, f_eig, f_ksum, f_lbl を求めた直後あたりに挿入）
    say("[USE] ", step_key, " / ", ds)
    say("  eigen : ", ifelse(is.na(f_eig),  "NA (not found)", f_eig))
    say("  ksum  : ", ifelse(is.na(f_ksum), "NA (not found)", f_ksum))
    say("  label : ", ifelse(is.na(f_lbl),  "NA (not found)", f_lbl))

  if (!is.null(lbl_new) && !is.null(lbl_signless)) {
    cat("[new  vs signless (linear 部分のみ)]\n")
    ari_ns <- pairwise_ari(lbl_new, lbl_signless)
    if (!is.null(ari_ns)) print(ari_ns[order(ari_ns$condition), ], row.names = FALSE) else cat("  比較できる共通列がありません。\n")
  } else {
    cat("[new  vs signless]  いずれかの labels が見つかりません。\n")
  }

  if (!is.null(lbl_new) && !is.null(lbl_olddata)) {
    cat("\n[new  vs olddata]\n")
    ari_no <- pairwise_ari(lbl_new, lbl_olddata)
    if (!is.null(ari_no)) print(ari_no[order(ari_no$condition), ], row.names = FALSE) else cat("  比較できる共通列がありません。\n")
  } else {
    cat("\n[new  vs olddata]  いずれかの labels が見つかりません。\n")
  }
}

cat("\n==== 完了：上の表が、寄与要約・kサマリ・ステップ間ARIです。====\n")

`%||%` <- function(a,b) if(!is.null(a)) a else b
```

```r
# --- 置き換え版: ClusterAssign_* から labels を組み立てる ---

# ファイル名→(mode_type, dim_mode, index) を抽出
parse_cond_from_fname <- function(fn, step_key){
  bn <- basename(fn)
  if (step_key == "signless") {
    m <- regmatches(bn, regexec("^ClusterAssign_(linear|nonlinear)_(top3|cumeig)_([a-z]+)_.*\\.csv$", bn))[[1]]
    if (length(m) == 4) return(list(mode_type = m[2], dim_mode = m[3], index = m[4]))
  } else {
    m <- regmatches(bn, regexec("^ClusterAssign_(top3|cumeig)_([a-z]+)_.*\\.csv$", bn))[[1]]
    if (length(m) == 3) return(list(mode_type = "linear", dim_mode = m[2], index = m[3]))
  }
  NULL
}

# ClusterAssign_* を全走査して、mode_type（linear/nonlinear）ごとに wide へ統合
build_labels_from_assign <- function(dir_ds, step_key){
  pats <- if (step_key == "signless")
            "^ClusterAssign_(linear|nonlinear)_(top3|cumeig)_([a-z]+)_.+\\.csv$"
          else
            "^ClusterAssign_(top3|cumeig)_([a-z]+)_.+\\.csv$"
  fs <- list.files(dir_ds, pattern = pats, full.names = TRUE)
  if (!length(fs)) return(list(linear = NULL, nonlinear = NULL, file = NA_character_))

  buckets <- list(linear = list(), nonlinear = list())

  for (f in fs) {
    info <- parse_cond_from_fname(f, step_key); if (is.null(info)) next
    df <- tryCatch(read.csv(f, check.names = FALSE), error = function(e) NULL); if (is.null(df)) next
    id  <- if ("Variable" %in% names(df)) df$Variable else df[[1]]
    lab <- df$Cluster
    colname <- paste0(info$dim_mode, "_", info$index)  # 例: "top3_ch", "cumeig_silhouette"
    one <- data.frame(id = id, tmp = as.character(lab), stringsAsFactors = FALSE)
    names(one)[2] <- colname
    buckets[[info$mode_type]] <- append(buckets[[info$mode_type]], list(one))
  }

  merge_list <- function(lst){
    if (!length(lst)) return(NULL)
    Reduce(function(x, y) merge(x, y, by = "id", all = TRUE), lst)
  }
  list(
    linear    = merge_list(buckets$linear),
    nonlinear = merge_list(buckets$nonlinear),
    file      = paste(length(fs), "ClusterAssign_* files")
  )
}

# ラベル読み込み：まず cluster_labels_*.csv を試し、無ければ ClusterAssign_* から構築
load_labels <- function(dir_ds, step_key){
  latest_file <- function(dir_path, pattern){
    cand <- list.files(dir_path, pattern = pattern, full.names = TRUE)
    if (length(cand) == 0) return(NA_character_)
    cand[order(file.info(cand)$mtime, decreasing = TRUE)][1]
  }
  safe_read_csv <- function(path){
    if (is.na(path)) return(NULL)
    tryCatch(utils::read.csv(path, check.names = FALSE), error = function(e) NULL)
  }

  # 1) cluster_labels_* があれば優先（signless は linear_/nonlinear_ をはがす）
  f_lbl <- latest_file(dir_ds, "^cluster_labels_.*\\.csv$")
  df <- safe_read_csv(f_lbl)
  if (!is.null(df) && "id" %in% names(df)) {
    if (step_key == "signless") {
      lin_cols <- grep("^linear_", names(df), value = TRUE)
      non_cols <- grep("^nonlinear_", names(df), value = TRUE)
      mk <- function(cols, pref){
        if (!length(cols)) return(NULL)
        out <- df[c("id", cols)]
        names(out) <- c("id", sub(paste0("^", pref, "_"), "", cols))
        out
      }
      return(list(linear = mk(lin_cols, "linear"),
                  nonlinear = mk(non_cols, "nonlinear"),
                  file = f_lbl))
    } else {
      other <- setdiff(names(df), "id")
      return(list(linear = df[c("id", other)], nonlinear = NULL, file = f_lbl))
    }
  }

  # 2) 無ければ ClusterAssign_* から組み立て（new/olddata はこちらになるはず）
  build_labels_from_assign(dir_ds, step_key)
}
```

```r
# ============================================================
# 軸セット別にクラスタリング → ARI（軸間比較）を出すツール
# - 既存の結果フォルダから MDScoords_*.csv / MDS_eigen_contributions_*.csv を読む
# - 軸セット（例：D1-D2, D1-D3, D2-D3, Top3(1:3), K80(累積80%)）ごとに NbClust でラベル作成
# - 同一 step/dataset/mode_type/index 内で、軸セット間の ARI を出力（ペア全通り）
# - signless は linear / nonlinear の両方に対応、new/olddata は linear のみ
# ============================================================

suppressPackageStartupMessages({
  if (!require(NbClust)) install.packages("NbClust"); library(NbClust)
  if (!require(mclust))  install.packages("mclust");  library(mclust)
  if (!require(dplyr))   install.packages("dplyr");   library(dplyr)
})

# -------- ルートとステップ定義（必要なら書き換え） --------
root_base <- "/Volumes/csbdeep11/_yasu-i/20250303_rebuiled/cal_cluster_No"
steps <- list(
  new      = "STEP3_new_MDS_20250904",
  signless = "STEP3.2_signlessCorr_MDS_20250904",
  olddata  = "STEP3.3_olddata_MDS_20250904"
)
datasets  <- c("OH", "FP")

# -------- ここを編集：比較したい「軸セット」定義 --------
# キー名（表に出る名前） = 取り出す次元ベクトル。K80 は後で動的解決するので NULL 固定。
axis_sets <- list(
  `D1-D2` = c(1, 2),
  `D1-D3` = c(1, 3),
  `D2-D3` = c(2, 3),
  `Top3`  = 1:3,
  `K80`   = NULL   # ← 累積80%（正の固有値基準）の「1:K」を自動採用
)

# -------- ユーティリティ --------
latest_file <- function(dir_path, pattern){
  ff <- list.files(dir_path, pattern = pattern, full.names = TRUE)
  if (!length(ff)) return(NA_character_)
  ff[order(file.info(ff)$mtime, decreasing = TRUE)][1]
}

read_csv_quiet <- function(path) {
  if (is.na(path)) return(NULL)
  tryCatch(utils::read.csv(path, check.names = FALSE), error = function(e) NULL)
}

# 寄与CSVから K80（正の固有値累積>=0.8での次元数）を取得
get_k80_from_eigen <- function(dir_ds){
  f <- latest_file(dir_ds, "^MDS_eigen_contributions_.*\\.csv$")
  df <- read_csv_quiet(f)
  if (is.null(df) || !"Share_pos" %in% names(df)) return(NA_integer_)
  sp <- df$Share_pos
  idx <- which(!is.na(sp) & sp > 0)
  if (!length(idx)) return(NA_integer_)
  csp <- cumsum(sp)
  k <- which(csp[idx] >= 0.8)[1]
  if (is.na(k)) k <- length(idx)
  k
}

# 該当 step/dataset/mode_type における「使える座標行列」を取得
# mode_type: "linear" / "nonlinear"
#   - new, olddata: linear のみ（ファイル名: MDScoords_top3_*, MDScoords_cumeig_*）
#   - signless:     linear / nonlinear（ファイル名に mode_type が入る）
load_coords <- function(step_key, dataset, mode_type, dir_ds){
  if (step_key == "signless") {
    f_top3   <- latest_file(dir_ds, paste0("^MDScoords_", mode_type, "_top3_.*\\.csv$"))
    f_cumeig <- latest_file(dir_ds, paste0("^MDScoords_", mode_type, "_cumeig_.*\\.csv$"))
  } else {
    if (mode_type != "linear") return(NULL)  # new/olddata は nonlin 無し
    f_top3   <- latest_file(dir_ds, "^MDScoords_top3_.*\\.csv$")
    f_cumeig <- latest_file(dir_ds, "^MDScoords_cumeig_.*\\.csv$")
  }
  X_top3   <- read_csv_quiet(f_top3)
  X_cumeig <- read_csv_quiet(f_cumeig)

  # 行名を揃える（最初の列が名前になっていない場合への保険）
  fix_mat <- function(df){
    if (is.null(df)) return(NULL)
    rn <- df[[1]]
    if (!is.character(rn)) rn <- as.character(rn)
    df <- df[, -1, drop = FALSE]
    rownames(df) <- rn
    as.matrix(df)
  }
  list(top3 = fix_mat(X_top3), cumeig = fix_mat(X_cumeig))
}

# 指定座標＆軸で NbClust → ラベル
label_by_index <- function(X, dims, index){
  # 利用可能次元でクリップ
  dims <- dims[dims <= ncol(X)]
  if (!length(dims)) return(NULL)
  XX <- X[, dims, drop = FALSE]
  est <- tryCatch(
    NbClust(data = XX, diss = NULL, distance = "euclidean",
            min.nc = 2, max.nc = 100, method = "ward.D2", index = index),
    error = function(e) NULL
  )
  if (is.null(est)) return(NULL)
  as.factor(est$Best.partition)
}

# 軸セットごとにラベルを作成（存在する軸のみ）、その全ペアで ARI を計算
axis_ari_table <- function(X_top3, X_cum, k80, index, axis_sets){
  # 軸セットごとに「使う行列」と「実際の dims」を決める
  labs <- list(); n_used <- NA_integer_
  for (nm in names(axis_sets)) {
    dims <- axis_sets[[nm]]
    if (identical(nm, "K80")) {
      if (is.na(k80) || is.null(X_cum)) next
      dims <- 1:k80
      Xuse <- X_cum
    } else {
      # 3次元までなら top3 の座標でOK。>3 が来たら cum を使う。
      Xuse <- if (!is.null(dims) && max(dims) <= 3 && !is.null(X_top3)) X_top3 else X_cum
      if (is.null(Xuse)) next
      if (!is.null(dims)) dims <- dims[dims <= ncol(Xuse)]
    }
    lab <- label_by_index(Xuse, dims, index)
    if (!is.null(lab)) {
      labs[[nm]] <- lab
      n_used <- length(lab)
    }
  }
  if (length(labs) < 2) return(NULL)

  # 全ペア ARI
  nmv <- names(labs)
  out <- list()
  for (i in 1:(length(nmv)-1)) {
    for (j in (i+1):length(nmv)) {
      a <- labs[[ nmv[i] ]]; b <- labs[[ nmv[j] ]]
      ok <- !(is.na(a) | is.na(b))
      ari <- if (sum(ok) >= 2) mclust::adjustedRandIndex(a[ok], b[ok]) else NA_real_
      out[[length(out)+1]] <- data.frame(
        from = nmv[i], to = nmv[j], n = sum(ok), ARI = ari, stringsAsFactors = FALSE
      )
    }
  }
  dplyr::bind_rows(out)
}

# -------- ここから実行 --------
index_list <- c("silhouette","dunn","gap","ch","db","ptbiserial")

for (step_key in names(steps)) {
  for (ds in datasets) {

    dir_ds <- file.path(root_base, steps[[step_key]], ds)
    if (!dir.exists(dir_ds)) {
      cat(sprintf("⚠️ パスが見つかりません: %s\n", dir_ds))
      next
    }

    # signless は linear / nonlinear、他は linear のみ
    mode_types <- if (step_key == "signless") c("linear","nonlinear") else "linear"

    for (mode_type in mode_types) {

      # 座標を読み込み
      coords <- load_coords(step_key, ds, mode_type, dir_ds)
      X_top3 <- coords$top3
      X_cum  <- coords$cumeig
      if (is.null(X_top3) && is.null(X_cum)) {
        cat(sprintf("[SKIP] %s / %s / %s: 座標CSVが見つかりません。\n", step_key, ds, mode_type))
        next
      }

      # K80 決定（正の固有値寄与 80%）
      k80 <- get_k80_from_eigen(dir_ds)

      cat(sprintf("\n==== 軸間 ARI（step=%s, dataset=%s, mode=%s, K80=%s）====\n",
                  step_key, ds, mode_type, ifelse(is.na(k80), "NA", as.character(k80))))

      for (ix in index_list) {
        tab <- axis_ari_table(X_top3, X_cum, k80, ix, axis_sets)
        if (is.null(tab)) {
          cat(sprintf("  - %s: 比較可能な軸セットが不足（座標や次元が無い）\n", ix))
        } else {
          tab <- tab %>% arrange(desc(ARI))
          cat(sprintf("  ■ index=%s\n", ix))
          print(tab, row.names = FALSE)
        }
      }
    }
  }
}
cat("\n✅ 完了：軸セット間の ARI を出力しました（step×dataset×mode×indexごと）。\n")
```
