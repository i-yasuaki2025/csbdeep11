```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

"""
batch_eval_after_onlyMat.py

目的:
- run_root 配下を再帰探索し、OH/FP の ClusterAssign_* をペアリング
- 各CSVが「サンプル割当（ID列あり）」か「変数割当（特徴名列のみ）」かを自動判定
- サンプル割当なら: OH×FP を同 (mode,index) で突合し ARI/NMI/AMI を計算
- 変数割当なら: materials-only 抽出後の監査レポート（keep/drop統計）を作成
- 出力は run_root/<out_name>/reports/ 以下に集約

使い方例:
python batch_eval_after_onlyMat.py \
  --base /Volumes/csbdeep11/_yasu-i/20250303_rebuiled/cal_cluster_No \
  --run-root 20251026_under_25clusters \
  --onlymat-root STEP3.3_onlyMat_20251027
"""

import argparse
from pathlib import Path
import pandas as pd
import numpy as np
import re

# sklearn が無い環境でも最低限 ARI/NMI/AMI が動くフォールバックを用意
try:
    from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score, adjusted_mutual_info_score
    SKLEARN_OK = True
except Exception:
    SKLEARN_OK = False

ID_CANDIDATES = {"id","sample","sample_id","sno","sid","index","name","SampleID","ID"}
VAR_CANDIDATES = {"variable","var","name","feature","Feature","column","col","field"}

def is_sample_assignment(df: pd.DataFrame) -> bool:
    # ID候補列があり、その列のユニーク比率が高め（>0.3）なら「サンプル割当」とみなす
    for c in df.columns:
        if str(c).strip() in ID_CANDIDATES or str(c).strip().lower() in {x.lower() for x in ID_CANDIDATES}:
            u = df[c].nunique(dropna=True)
            if len(df) == 0: 
                return False
            if u / max(1, len(df)) >= 0.3:
                return True
    return False

def detect_id_col(df):
    for c in df.columns:
        lc = str(c).strip().lower()
        if lc in {x.lower() for x in ID_CANDIDATES}:
            return c
    return df.columns[0]

def detect_label_col(df, id_col):
    priors = ["cluster","label","assign","assignment","cluster_label","clusterid","cluster_id","k_label","k"]
    for p in priors:
        for c in df.columns:
            if str(c).strip().lower() == p:
                return c
    # ヒューリスティック
    not_id = [c for c in df.columns if c != id_col]
    n = len(df)
    best, score_best = None, -1
    for c in not_id:
        u = df[c].nunique(dropna=True)
        is_intlike = pd.api.types.is_integer_dtype(df[c]) or pd.api.types.is_categorical_dtype(df[c])
        score = (1 if is_intlike else 0) + (1 if 2 <= u <= max(2, n//2) else 0)
        if score > score_best:
            best, score_best = c, score
    return best

def parse_mode_index(fname: str):
    # ClusterAssign_{mode}_{index}_{OH|FP}_*.csv から mode/index をざっくり抽出（フォールバック付き）
    m = re.search(r"ClusterAssign_([^_]+)_([^_]+)_(OH|FP)_", fname)
    if m:
        return m.group(1), m.group(2), m.group(3)
    # フォールバック: _OH_/_FP_ の直前2トークンを mode/index とみなす
    m2 = re.search(r"ClusterAssign_(.*)_(OH|FP)_", fname)
    if m2:
        mid = m2.group(1)
        toks = mid.split("_")
        if len(toks) >= 2:
            return toks[-2], toks[-1], m2.group(2)
    return "", "", ("OH" if "_OH_" in fname else "FP")

def compute_pair_metrics(df_oh, df_fp):
    id_oh = detect_id_col(df_oh)
    id_fp = detect_id_col(df_fp)
    lab_oh = detect_label_col(df_oh, id_oh)
    lab_fp = detect_label_col(df_fp, id_fp)
    ids = sorted(set(df_oh[id_oh]).intersection(set(df_fp[id_fp])))
    if len(ids) == 0:
        return None
    y_oh = df_oh.set_index(id_oh).loc[ids][lab_oh].astype(str).values
    y_fp = df_fp.set_index(id_fp).loc[ids][lab_fp].astype(str).values
    kA = int(pd.Series(y_oh).nunique()); kB = int(pd.Series(y_fp).nunique())
    if kA < 2 or kB < 2:
        return None

    if SKLEARN_OK:
        ARI = float(adjusted_rand_score(y_oh, y_fp))
        NMI = float(normalized_mutual_info_score(y_oh, y_fp))
        AMI = float(adjusted_mutual_info_score(y_oh, y_fp))
    else:
        # 簡易フォールバック: ARI は手実装、NMI/AMIは近似
        def _ari(y_true, y_pred):
            y_true = np.asarray(y_true); y_pred = np.asarray(y_pred)
            _, lt = np.unique(y_true, return_inverse=True)
            _, lp = np.unique(y_pred, return_inverse=True)
            cont = pd.crosstab(pd.Series(lt), pd.Series(lp)).values
            nij = cont
            ai = nij.sum(axis=1); bj = nij.sum(axis=0); n = nij.sum()
            def comb2(x): x = np.asarray(x); return (x*(x-1))//2
            sum_c = comb2(nij).sum(); sum_a = comb2(ai).sum(); sum_b = comb2(bj).sum(); total = comb2(n)
            exp = (sum_a*sum_b)/total if total>0 else 0.0
            maxv = 0.5*(sum_a+sum_b); denom = (maxv-exp); 
            return float((sum_c-exp)/denom) if denom!=0 else 0.0
        def _entropy(p):
            p = np.asarray(p, dtype=float); p = p[p>0]; 
            return -float(np.sum(p*np.log(p)))
        def _nmi(y_true, y_pred):
            y_true = np.asarray(y_true); y_pred = np.asarray(y_pred)
            _, lt = np.unique(y_true, return_inverse=True)
            _, lp = np.unique(y_pred, return_inverse=True)
            n = len(lt); cont = pd.crosstab(pd.Series(lt), pd.Series(lp)).values.astype(float)
            pi = cont.sum(axis=1)/n; pj = cont.sum(axis=0)/n; pij = cont/n
            with np.errstate(divide='ignore', invalid='ignore'):
                logt = np.log(pij/(pi[:,None]*pj[None,:])); logt[~np.isfinite(logt)] = 0.0
                mi = float(np.nansum(pij*logt))
            h_t = _entropy(pi); h_p = _entropy(pj); denom = (h_t+h_p)/2.0 if h_t+h_p>0 else 1.0
            return float(mi/denom) if denom>0 else 0.0
        ARI = _ari(y_oh, y_fp); NMI = _nmi(y_oh, y_fp); AMI = NMI

    return dict(n_overlap=len(ids), kA_OH=kA, kB_FP=kB, ARI=ARI, NMI=NMI, AMI=AMI)

def main():
    ap = argparse.ArgumentParser()
    ap.add_argument("--base", required=True)
    ap.add_argument("--run-root", required=True)
    ap.add_argument("--onlymat-root", required=True, help="onlyMat 出力フォルダ名（run-root直下）")
    args = ap.parse_args()

    base = Path(args.base).resolve()
    run_root = (base / args.run_root).resolve()
    only_root = (run_root / args.onlymat_root).resolve()
    report_root = only_root / "reports"
    report_root.mkdir(parents=True, exist_ok=True)

    # onlyMat 抽出後のフォルダを対象（OH/FP ともに）
    in_oh = list((only_root / "OH" / "04_assignments_onlyMat").glob("ClusterAssign_*_OH_*_onlyMat.csv"))
    in_fp = list((only_root / "FP" / "04_assignments_onlyMat").glob("ClusterAssign_*_FP_*_onlyMat.csv"))

    # ファイル一覧（監査用）
    pd.DataFrame({"OH_files": [str(p) for p in in_oh]}).to_csv(report_root/"listing_OH.csv", index=False)
    pd.DataFrame({"FP_files": [str(p) for p in in_fp]}).to_csv(report_root/"listing_FP.csv", index=False)

    # 変数割当 vs サンプル割当の自動判定と集計
    audit_rows = []
    for p in in_oh + in_fp:
        try:
            df = pd.read_csv(p, nrows=1000)  # 先頭だけで判定
            kind = "sample" if is_sample_assignment(df) else "variable"
        except Exception as e:
            kind = f"read_error: {e}"
        audit_rows.append({"file": str(p), "kind": kind, "n_cols": (df.shape[1] if 'df' in locals() else np.nan)})
    pd.DataFrame(audit_rows).to_csv(report_root/"file_kinds_audit.csv", index=False)

    # サンプル割当が存在する場合のみ、同 mode/index で ARI/NMI/AMI を計算
    # ペアリング辞書： (mode,index) -> latest OH/FP ファイル
    def latest_by_mode_index(paths):
        d = {}
        for p in sorted(paths):
            mode, index, side = parse_mode_index(p.name)
            d.setdefault((mode,index,side), p)
        return d

    oh_map = latest_by_mode_index(in_oh)
    fp_map = latest_by_mode_index(in_fp)

    results = []
    for (mode,index,side), oh_path in oh_map.items():
        if side != "OH": 
            continue
        fp_key = (mode,index,"FP")
        fp_path = fp_map.get(fp_key, None)
        if fp_path is None:
            continue
        # 読み込み
        df_oh = pd.read_csv(oh_path)
        df_fp = pd.read_csv(fp_path)
        # サンプル割当のときだけ
        if is_sample_assignment(df_oh) and is_sample_assignment(df_fp):
            met = compute_pair_metrics(df_oh, df_fp)
            if met is not None:
                results.append({
                    "mode": mode, "index": index,
                    "OH_file": str(oh_path), "FP_file": str(fp_path),
                    **met
                })

    if results:
        df_res = pd.DataFrame(results).sort_values(["mode","index"])
        out_sum = report_root / "ARI_NMI_AMI_summary_onlyMat_pairs.csv"
        df_res.to_csv(out_sum, index=False, encoding="utf-8-sig")
        print("Saved:", out_sum)
    else:
        print("No sample-assignment pairs detected under onlyMat. Skipped ARI/NMI/AMI.")

    # 変数割当だった場合の監査：keep/drop の集計（onlyMat 抽出後の件数を集める）
    # 元ログを同フォルダから拾う
    log_all = only_root / "logs" / "filter_onlyMat_index_ALL.csv"
    if log_all.exists():
        df_log = pd.read_csv(log_all)
        # 集約：side×mode×index で after/before の合計
        def get_mi(name):
            m = re.search(r"ClusterAssign_([^_]+)_([^_]+)_(OH|FP)_", name)
            if m: return m.group(1), m.group(2), m.group(3)
            return "", "", "OH" if "_OH_" in name else "FP"
        recs=[]
        for _, r in df_log.iterrows():
            mode, index, side = get_mi(Path(r["input_file"]).name)
            recs.append({
                "side": side, "mode": mode, "index": index,
                "n_rows_before": r.get("n_rows_before", np.nan),
                "n_rows_after": r.get("n_rows_after", np.nan),
                "n_dropped": r.get("n_dropped", np.nan)
            })
        df_agg = pd.DataFrame(recs).groupby(["side","mode","index"], dropna=False).sum(numeric_only=True).reset_index()
        out_audit = report_root / "onlyMat_keep_drop_summary.csv"
        df_agg.to_csv(out_audit, index=False, encoding="utf-8-sig")
        print("Saved:", out_audit)

if __name__ == "__main__":
    main()
```
